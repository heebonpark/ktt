```python
import os
import subprocess
import pandas as pd
import glob

# AutoIt 실행 파일 경로
autoit_path = r"C:\Program Files (x86)\AutoIt3\AutoIt3.exe"

# AutoIt 스크립트 파일 경로
script_path = r"D:\util\autoit-v3-setup\login_script.au3"
log_file_path = r"D:\util\autoit-v3-setup\log.txt"

# 병합 폴더 경로 설정
daily_folder_path = r'd:\신규P값총량제\통합'
cumulative_folder_path = r'd:\신규P값총량제\누계통합'
output_excel_path = r'd:\신규P값총량제\신규P값병합_통합.xlsx'

# AutoIt 스크립트 내용 작성
autoit_script_content = '''
Func LogMessage($message)
    FileWrite("D:\\util\\autoit-v3-setup\\log.txt", $message & @CRLF)
    ConsoleWrite($message & @CRLF)
EndFunc

; 1. Microsoft Edge 실행
LogMessage("1단계: Microsoft Edge 실행 시작")
Run('"C:\\Program Files (x86)\\Microsoft\\Edge\\Application\\msedge.exe" http://10.200.2.100/nims')
If @error Then
    LogMessage("1단계: Microsoft Edge 실행 실패")
    Exit
EndIf

; 2. 브라우저 창 활성화
LogMessage("2단계: 브라우저 창 활성화 대기")
If Not WinWaitActive("[CLASS:Chrome_WidgetWin_1]", "", 20) Then
    LogMessage("2단계: 브라우저 창 활성화 실패")
    Exit
EndIf

; 3. 로그인 정보 입력
LogMessage("3단계: 로그인 정보 입력")
MouseClick("left", 1408, 730) ; 아이디 필드 클릭
Sleep(500)
Send("03313") ; 아이디 입력
Sleep(500)
ClipPut("plus0303!!") ; 비밀번호 복사
MouseClick("left", 1360, 755) ; 비밀번호 필드 클릭
Sleep(500)
Send("^v") ; 비밀번호 붙여넣기
Sleep(500)
MouseClick("left", 1446, 751) ; 로그인 버튼 클릭
Sleep(5000)

; 4. 닫기 버튼 클릭
LogMessage("4단계: 닫기 버튼 클릭")
MouseClick("left", 1027, 590)
Sleep(2000)

; 5. 현황/통계 클릭
LogMessage("5단계: 현황/통계 클릭")
MouseClick("left", 626, 101)
Sleep(1000)

; 6. 실적현황 클릭
LogMessage("6단계: 실적현황 클릭")
MouseClick("left", 655, 320)
Sleep(1000)

; 7. 데이터 다운로드 순서대로 실행
Local $파일목록[8][2] = [ _
    ["채널별, 월", 783, 362], _
    ["기관별, 월", 118, 271], _
    ["강북_월", 281, 186], _
    ["강남_월", 262, 200], _
    ["부산", 272, 208], _
    ["대구", 273, 219], _
    ["전남", 265, 234], _
    ["충남", 270, 245] _
]

For $i = 0 To UBound($파일목록) - 1
    LogMessage("7단계: " & $파일목록[$i][0] & " 데이터 다운로드 시작")
   
    ; 메뉴 클릭
    MouseClick("left", $파일목록[$i][1], $파일목록[$i][2])
    Sleep(2000)

    ; 검색 클릭
    LogMessage("검색 클릭")
    MouseClick("left", 996, 164)
    Sleep(5000)

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(2000)

    ; 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 필드 클릭
    Sleep(500)
    ClipPut($파일목록[$i][0])
    Send("^v") ; 파일 이름 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(2000)
Next

LogMessage("7단계 완료: 모든 데이터 다운로드 완료")
MsgBox(0, "완료", "데이터 다운로드가 완료되었습니다.")
'''

# AutoIt 스크립트 저장
with open(script_path, 'w', encoding='utf-8') as script_file:
    script_file.write(autoit_script_content)
print(f"AutoIt 스크립트가 {script_path}에 저장되었습니다.")

# AutoIt 스크립트 실행
try:
    if os.path.exists(autoit_path):
        result = subprocess.run(
            [autoit_path, script_path],
            capture_output=True,
            text=True,
        )
        print("AutoIt 스크립트 실행 성공")
        print("출력 메시지:", result.stdout)
        print("에러 메시지:", result.stderr)

        # 로그 파일 출력
        if os.path.exists(log_file_path):
            with open(log_file_path, 'r', encoding='utf-8') as log_file:
                print("\n로그 파일 내용:")
                print(log_file.read())
    else:
        print(f"AutoIt 실행 파일을 찾을 수 없습니다: {autoit_path}")
except subprocess.CalledProcessError as e:
    print(f"AutoIt 실행 실패: {e}")
    print(f"에러 메시지:", e.stderr)
except Exception as e:
    print(f"예외가 발생했습니다: {e}")

# 병합 작업
file_order = [
    "0. 채널별, 월", "1.기관별, 월",
    "2.강북_월", "2.강남_월", "2.부산",
    "2.전남", "2.대구", "2.충남"
]

def get_file_priority(file_name):
    for i, keyword in enumerate(file_order):
        if keyword in file_name:
            return i
    return len(file_order)

def merge_files(folder_path):
    all_files = glob.glob(folder_path + "/*.xls")
    all_files.sort(key=lambda x: get_file_priority(os.path.basename(x)))
    all_data = pd.DataFrame()

    for file in all_files:
        try:
            df_list = pd.read_html(file)
            df = df_list[0]
            all_data = pd.concat([all_data, df], ignore_index=True, sort=False)
        except Exception as e:
            print(f"파일 읽기 중 오류: {file}, 오류: {e}")
    return all_data

try:
    daily_data = merge_files(daily_folder_path)
    cumulative_data = merge_files(cumulative_folder_path)

    with pd.ExcelWriter(output_excel_path, engine='xlsxwriter') as writer:
        daily_data.to_excel(writer, index=False, sheet_name='일일 데이터')
        cumulative_data.to_excel(writer, index=False, sheet_name='누계 데이터')

    print(f"병합된 파일이 저장되었습니다: {output_excel_path}")
except Exception as e:
    print(f"병합 작업 중 오류 발생: {e}")
```
