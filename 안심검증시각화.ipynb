```python
import pandas as pd

# 파일 경로 설정
file_path1 = '/content/drive/My Drive/data/월별평가_본부.xlsx'
file_path2 = '/content/drive/My Drive/data/안심검증_월_지사평가.xlsx'

# Excel 파일 불러오기
df1 = pd.read_excel(file_path1)
df2 = pd.read_excel(file_path2)

# 데이터 확인
print(df1.head())
print(df2.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'  # Linux에서 한글 폰트 경로 설정
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

# 스타일 설정
sns.set(style="whitegrid")

# 시각화: 본부별 월별 평가 점수 (df1 데이터)
plt.figure(figsize=(14, 8))
sns.barplot(data=df1, x='월', y='평가점수', hue='본부', palette='muted')
sns.lineplot(data=df1, x='월', y='평가점수', hue='본부', marker='o', sort=False, linewidth=2.5, legend=False, palette='muted')

plt.title('본부별 월별 평가 점수', fontsize=16)
plt.xlabel('월', fontsize=14)
plt.ylabel('평가점수', fontsize=14)
plt.legend(title='본부', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# 시각화: 지사별 안심검증 평가 점수 (df2 데이터)
plt.figure(figsize=(14, 8))
sns.barplot(data=df2, x='지사', y='평가점수', hue='월', palette='deep')
sns.lineplot(data=df2, x='지사', y='평가점수', hue='월', marker='o', sort=False, linewidth=2.5, legend=False, palette='deep')

plt.title('지사별 안심검증 평가 점수', fontsize=16)
plt.xlabel('지사', fontsize=14)
plt.ylabel('평가점수', fontsize=14)
plt.legend(title='월', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드 (이미 로드된 상태로 가정)
# df1, df2 = pd.read_excel(...)

# 데이터를 길게 변환
df1_melted = df1.melt(id_vars=['ID', '구분'],
                      value_vars=['4월평균', '5월평균', '6월평균', '7월평균'],
                      var_name='월',
                      value_name='평균점수')

df2_melted = df2.melt(id_vars=['ID', '구분'],
                      value_vars=['4월평균', '5월평균', '6월평균', '7월평균'],
                      var_name='월',
                      value_name='평균점수')

# 시각화: 본부별 월별 평가 점수 (df1_melted 데이터)
plt.figure(figsize=(14, 8))
sns.barplot(data=df1_melted, x='월', y='평균점수', hue='구분', palette='muted')
sns.lineplot(data=df1_melted, x='월', y='평균점수', hue='구분', marker='o', sort=False, linewidth=2.5, legend=False, palette='muted')

plt.title('본부별 월별 평가 점수', fontsize=16)
plt.xlabel('월', fontsize=14)
plt.ylabel('평균점수', fontsize=14)
plt.legend(title='구분', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# 시각화: 지사별 안심검증 평가 점수 (df2_melted 데이터)
plt.figure(figsize=(14, 8))
sns.barplot(data=df2_melted, x='월', y='평균점수', hue='구분', palette='deep')
sns.lineplot(data=df2_melted, x='월', y='평균점수', hue='구분', marker='o', sort=False, linewidth=2.5, legend=False, palette='deep')

plt.title('지사별 안심검증 평가 점수', fontsize=16)
plt.xlabel('월', fontsize=14)
plt.ylabel('평균점수', fontsize=14)
plt.legend(title='구분', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드 및 변환 (이미 로드된 상태로 가정)
# df1_melted, df2_melted = ...

# 시각화: 본부별 월별 평가 점수 (df1_melted 데이터)
plt.figure(figsize=(14, 8))
sns.barplot(data=df1_melted, x='월', y='평균점수', hue='구분', palette='muted', alpha=0.8)
sns.lineplot(data=df1_melted, x='월', y='평균점수', hue='구분', marker='o', sort=False, linewidth=2.5, legend=False, palette='muted')

plt.title('본부별 월별 평가 점수', fontsize=16)
plt.xlabel('월', fontsize=14)
plt.ylabel('평균점수', fontsize=14)
plt.legend(title='구분', fontsize=12, loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()

# 시각화: 지사별 안심검증 평가 점수 (df2_melted 데이터)
plt.figure(figsize=(14, 8))
sns.barplot(data=df2_melted, x='월', y='평균점수', hue='구분', palette='deep', alpha=0.8)
sns.lineplot(data=df2_melted, x='월', y='평균점수', hue='구분', marker='o', sort=False, linewidth=2.5, legend=False, palette='deep')

plt.title('지사별 안심검증 평가 점수', fontsize=16)
plt.xlabel('월', fontsize=14)
plt.ylabel('평균점수', fontsize=14)
plt.legend(title='구분', fontsize=12, loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()
```
