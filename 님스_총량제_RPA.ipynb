````python


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import time

# 설정
url = "http://10.200.2.100/nims"  # 내부망 URL
username = "03313"  # 사용자 아이디
password = "plus0303!!"  # 사용자 비밀번호

# Edge WebDriver 경로 설정
edge_driver_path = "D:\\util\\edgedriver_win64\\msedgedriver.exe"  # WebDriver 경로
edge_options = Options()
edge_options.add_argument("--disable-gpu")
edge_options.add_argument("--no-sandbox")
# 필요시 Headless 모드 사용
# edge_options.add_argument("--headless")

# Edge WebDriver 초기화
service = Service(edge_driver_path)
driver = webdriver.Edge(service=service, options=edge_options)

try:
    # URL 열기
    driver.get(url)
    time.sleep(2)  # 페이지 로딩 대기

    # 로그인 폼 찾기
    username_field = driver.find_element(By.NAME, "03313")  # 아이디 입력 필드
    password_field = driver.find_element(By.NAME, "plus0303!!")  # 비밀번호 입력 필드
    login_button = driver.find_element(By.ID, "loginButton")  # 로그인 버튼

    # 로그인 정보 입력
    username_field.send_keys(username)
    password_field.send_keys(password)

    # 로그인 버튼 클릭
    login_button.click()

    # 로그인 성공 여부 확인
    time.sleep(3)  # 대기 후 로그인 확인
    print("로그인 성공!")

except Exception as e:
    print(f"에러 발생: {e}")

finally:
    # 브라우저 닫기
    driver.quit()
````

````python
import os
import subprocess

# AutoIt 실행 파일 경로
autoit_path = r"C:\Program Files (x86)\AutoIt3\AutoIt3.exe"

# AutoIt 스크립트 파일 경로
script_path = r"D:\util\autoit-v3-setup\login_script.au3"

# AutoIt 스크립트 내용 작성
autoit_script = '''
Func LogMessage($message)
    ConsoleWrite($message & @CRLF)
EndFunc

; Microsoft Edge 실행
LogMessage("1단계: Microsoft Edge 실행 시작")
Run('"C:\\Program Files (x86)\\Microsoft\\Edge\\Application\\msedge.exe" http://10.200.2.100/nims')
If @error Then
    LogMessage("1단계: Microsoft Edge 실행 실패")
    Exit
EndIf

; 브라우저 창 대기
LogMessage("2단계: 브라우저 창 활성화 대기")
If Not WinWaitActive("[CLASS:Chrome_WidgetWin_1]", "", 20) Then
    LogMessage("2단계: 브라우저 창 활성화 실패")
    Exit
EndIf

; 창 활성화 및 맨 앞으로 가져오기
LogMessage("3단계: 브라우저 창 맨 앞으로 가져오기")
WinSetState("[CLASS:Chrome_WidgetWin_1]", "", @SW_SHOW)
WinActivate("[CLASS:Chrome_WidgetWin_1]")
Sleep(2000)

; 로그인 GUI 활성화 확인
LogMessage("4단계: 로그인 GUI 활성화 확인")
If Not WinExists("[CLASS:Chrome_WidgetWin_1]") Then
    LogMessage("4단계: 로그인 창 활성화 실패")
    Exit
EndIf

; 창 최소화 버튼 클릭
LogMessage("5단계: 창 최소화 클릭")
MouseClick("left", 2435, 15) ; 최소화 버튼 클릭
Sleep(2000)

; 로그인 정보 입력
LogMessage("6단계: 로그인 정보 입력")
MouseClick("left", 1408, 730) ; 아이디 입력 필드 클릭
Sleep(500)
Send("03313") ; 아이디 입력
Sleep(500)
ClipPut("plus0303!!") ; 비밀번호 복사
MouseClick("left", 1360, 755) ; 비밀번호 입력 필드 클릭
Sleep(500)
Send("^v") ; 비밀번호 붙여넣기
Sleep(500)
MouseClick("left", 1446, 751) ; 로그인 버튼 클릭
Sleep(5000)

; 작업 시작
LogMessage("7단계: 닫기 선택 필드 클릭")
MouseClick("left", 1027, 590)
Sleep(2000)

; 현황/통계 선택 필드
LogMessage("8단계: 현황/통계 선택")
MouseClick("left", 626, 101)
Sleep(2000)

; 실적현황 선택 필드
LogMessage("9단계: 실적현황 선택")
MouseClick("left", 655, 320)
Sleep(2000)

; 경로 설정 (최초 1회)
LogMessage("10단계: 저장 경로 설정")
Send("^l") ; 주소창 포커스
Sleep(500)
ClipPut("D:\\신규P값총량제\\병합") ; 경로 복사
Send("^v") ; 붙여넣기
Send("{ENTER}") ; 경로 이동
Sleep(2000)

; 반복 작업 정의
LogMessage("11단계: 반복 작업 시작")
Local $파일목록[8][2] = [ _
    ["기관별", "1.기관별, 월.xls"], _
    ["사원별", "3.사원.xls"], _
    ["강북", "2.강북_월.xls"], _
    ["강남", "2.강남_월.xls"], _
    ["부산", "2.부산.xls"], _
    ["대구", "2.대구.xls"], _
    ["전남", "2.전남.xls"], _
    ["충남", "2.충남.xls"], _
]

For $i = 0 To UBound($파일목록) - 1
    ; 전체목록 클릭
    LogMessage("11단계: " & $파일목록[$i][0] & " 선택 시작")
    MouseClick("left", 371, 149)
    Sleep(2000)

    ; 본부 클릭 (조건부)
    If $i > 1 Then
        Switch $파일목록[$i][0]
            Case "강북"
                MouseClick("left", 323, 175)
            Case "강남"
                MouseClick("left", 301, 190)
            Case "부산"
                MouseClick("left", 319, 202)
            Case "대구"
                MouseClick("left", 311, 221)
            Case "전남"
                MouseClick("left", 303, 227)
            Case "충남"
                MouseClick("left", 304, 237)
        EndSwitch
        Sleep(2000)
    EndIf

    ; 검색 클릭
    LogMessage("11단계: 검색 클릭")
    MouseClick("left", 1024, 155)
    Sleep(5000)

    ; 엑셀 다운로드 클릭
    LogMessage("11단계: 엑셀 다운로드 클릭")
    MouseClick("left", 1034, 713)
    Sleep(5000)

    ; 파일 저장
    LogMessage("11단계: 파일 저장")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut($파일목록[$i][1]) ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 830, 527) ; 저장 버튼 클릭
    Sleep(5000)
Next

LogMessage("12단계: 모든 작업 완료")
MsgBox(0, "완료", "모든 작업이 완료되었습니다.")
'''

# AutoIt 스크립트 저장
with open(script_path, 'w', encoding='utf-8') as script_file:
    script_file.write(autoit_script)
print(f"AutoIt 스크립트가 {script_path}에 저장되었습니다.")

# AutoIt 스크립트 실행
try:
    if os.path.exists(autoit_path):
        result = subprocess.run(
            [autoit_path, script_path],
            capture_output=True,
            text=True
        )
        print("AutoIt 스크립트 실행 성공")
        print("출력 메시지:", result.stdout)
        print("에러 메시지:", result.stderr)
        result.check_returncode()  # 반환 코드가 0이 아니면 예외 발생
    else:
        print(f"AutoIt 실행 파일을 찾을 수 없습니다: {autoit_path}")
except subprocess.CalledProcessError as e:
    print(f"AutoIt 실행 실패: {e}")
    print("에러 메시지:", e.stderr)
except Exception as e:
    print(f"예상치 못한 오류 발생: {e}")
````

````python
import os
import subprocess

# AutoIt 실행 파일 경로
autoit_path = r"C:\Program Files (x86)\AutoIt3\AutoIt3.exe"

# AutoIt 스크립트 파일 경로
script_path = r"D:\util\autoit-v3-setup\login_script.au3"
log_file_path = r"D:\util\autoit-v3-setup\log.txt"

# 로그 파일 초기화
if os.path.exists(log_file_path):
    os.remove(log_file_path)

# AutoIt 스크립트 내용 작성
autoit_script_content = '''
Func LogMessage($message)
    FileWrite("D:\\util\\autoit-v3-setup\\log.txt", $message & @CRLF)
    ConsoleWrite($message & @CRLF)
EndFunc

; Microsoft Edge 실행
LogMessage("1단계: Microsoft Edge 실행 시작")
Run('"C:\\Program Files (x86)\\Microsoft\\Edge\\Application\\msedge.exe" http://10.200.2.100/nims')
If @error Then
    LogMessage("1단계: Microsoft Edge 실행 실패")
    Exit
EndIf

; 브라우저 창 대기
LogMessage("2단계: 브라우저 창 활성화 대기")
If Not WinWaitActive("[CLASS:Chrome_WidgetWin_1]", "", 20) Then
    LogMessage("2단계: 브라우저 창 활성화 실패")
    Exit
EndIf

; 창 활성화 및 맨 앞으로 가져오기
LogMessage("3단계: 브라우저 창 맨 앞으로 가져오기")
WinSetState("[CLASS:Chrome_WidgetWin_1]", "", @SW_SHOW)
WinActivate("[CLASS:Chrome_WidgetWin_1]")
Sleep(2000)

; 로그인 GUI 활성화 확인
LogMessage("4단계: 로그인 GUI 활성화 확인")
If Not WinExists("[CLASS:Chrome_WidgetWin_1]") Then
    LogMessage("4단계: 로그인 창 활성화 실패")
    Exit
EndIf

; 창 최소화 버튼 클릭
LogMessage("5단계: 창 최소화 클릭")
MouseClick("left", 2435, 15) ; 최소화 버튼 클릭
Sleep(2000)

; 로그인 정보 입력
LogMessage("6단계: 로그인 정보 입력")
MouseClick("left", 1408, 730) ; 아이디 입력 필드 클릭
Sleep(500)
Send("03313") ; 아이디 입력
Sleep(500)
ClipPut("plus0303!!") ; 비밀번호 복사
MouseClick("left", 1360, 755) ; 비밀번호 입력 필드 클릭
Sleep(500)
Send("^v") ; 비밀번호 붙여넣기
Sleep(500)
MouseClick("left", 1446, 751) ; 로그인 버튼 클릭
Sleep(5000)

 ; 작업 수행
    ; 닫기 선택 필드
    MouseClick("left", 1027, 590)
    Sleep(1000)

 ; 현황/통계 선택 필드
    MouseClick("left", 626, 101)
    Sleep(1000)

     ; 실적현황 선택 필드
    MouseClick("left", 655, 320)
    Sleep(1000)

    ; 신규P값총량제 선택 필드
    MouseClick("left", 783, 362)
    Sleep(1000)

    ; 조회기간목록 선택 필드
    MouseClick("left", 223, 189) ; 조회기간목록 클릭
    Sleep(500)

    ; 조회기간 "월" 선택 필드
    MouseClick("left", 155, 214) ; "월" 클릭
    Sleep(500)

    ; 달력1 입력 필드
    MouseClick("left", 335, 183) ; 달력1 클릭
    Sleep(500)

    ; 날짜1 입력 필드
    MouseClick("left", 276, 252) ; 날짜1 클릭
    Sleep(500)

    ; 달력2 입력 필드
    MouseClick("left", 462, 186) ; 달력2 클릭
    Sleep(500)

    ; 날짜2 입력 필드
    MouseClick("left", 426, 328) ; 날짜2 클릭
    Sleep(500)

    ; 검색 선택 필드
    MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(1000) ; 검색 결과 대기

    ; 엑셀 선택 필드
    MouseClick("left", 982, 715) ; 엑셀 다운로드 클릭
    Sleep(1000) ; 다운로드 대기

    
    ; 13. 경로 이동 (D:\) ; 초기화
    Send("^l") ; 주소창 포커스
    Sleep(500)
    ; 13. 경로 이동 (D:\신규P값총량제\통합)
    Send("^l") ; 주소창 포커스
    Sleep(500)
    ClipPut("D:\\신규P값총량제\\통합") ; 경로 복사
    Send("^v") ; 붙여넣기
    Send("{ENTER}") ; 경로 이동
    Sleep(700)

    ; no1. 파일 저장
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("0. 채널별, 월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)

    ; 저장 버튼 클릭
    LogMessage("저장 클릭")
    MouseClick("left", 793, 540) ; 저장 버튼 클릭
    Sleep(700) ; 저장 완료 대기

    ; 기관별_클릭
    LogMessage("기관별 클릭")
    MouseClick("left", 118, 271) ; 저장 버튼 클릭

     
   ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

   
    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)
          
    ; no2. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("1.기관별, 월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)

; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 강북/강원 _클릭
    LogMessage(" 강북/강원 클릭")
    MouseClick("left", 281, 186) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

     
    ; no3. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.강북_월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 강남/서부 _클릭
    LogMessage(" 강남/서부 클릭")
    MouseClick("left", 262, 200) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.강남_월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------


; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 부산 _클릭
    LogMessage(" 부산/경남 클릭")
    MouseClick("left", 272, 208) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.부산") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------


; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 대구 _클릭
    LogMessage(" 대구/경북 클릭")
    MouseClick("left", 273, 219) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.대구") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------

; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 전남 _클릭
    LogMessage(" 전남/전북 클릭")
    MouseClick("left", 265, 234) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.전남") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------
  ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 충남 _클릭
    LogMessage(" 충남/충북 클릭")
    MouseClick("left", 270, 245) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.충남") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)

; -------


   ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 전체_클릭
    LogMessage("전체 클릭")
    MouseClick("left", 242, 175) ; 저장 버튼 클릭


  ; 사원별_클릭
    LogMessage("사원별 클릭")
    MouseClick("left", 181, 270) ; 저장 버튼 클릭

  ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(700) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)


    
   ; no4. 사원 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("3.사원") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)



    
   ;---- 누계통합--

   ; 채널별 선택 필드
    MouseClick("left", 57, 273) ; 채널별선택
    Sleep(500)  
           
   ; 달력1 입력 필드
    MouseClick("left", 335, 183) ; 달력1 클릭
    Sleep(500)

    MouseClick("left", 242, 212) ; 날짜 11월 클릭
    Sleep(500)

    MouseClick("left", 242, 212) ; 날짜 10월 클릭
    Sleep(700)

    MouseClick("left", 242, 212) ; 날짜 09월 클릭
    Sleep(700)

    MouseClick("left", 243, 214) ; 날짜 08월 클릭
    Sleep(700)

     
    MouseClick("left", 354, 249) ; 날짜 6일자 클릭
    Sleep(700)  

   
    ; 검색 선택 필드
    MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(2000) ; 검색 결과 대기

    ; 엑셀 선택 필드
    MouseClick("left", 982, 715) ; 엑셀 다운로드 클릭
    Sleep(1000) ; 다운로드 대기

    ; 13. 경로 이동 (D:\) ; 초기화
    Send("^l") ; 주소창 포커스
    Sleep(1000)
    ; 13. 경로 이동 (D:\신규P값총량제\누계통합)
    Send("^l") ; 주소창 포커스
    Sleep(800)
    ClipPut("D:\\신규P값총량제\\누계통합") ; 경로 복사
    Send("^v") ; 붙여넣기
    Send("{ENTER}") ; 경로 이동
    Sleep(900)

    ; no1. 파일 저장
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(900)
    ClipPut("0. 채널별, 월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(2000)

    ; 저장 버튼 클릭
    LogMessage("저장 클릭")
    MouseClick("left", 793, 540) ; 저장 버튼 클릭
    Sleep(700) ; 저장 완료 대기

    ; 기관별_클릭
    LogMessage("기관별 클릭")
    MouseClick("left", 118, 271) ; 저장 버튼 클릭

     
   ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(9000) ; 검색 결과 대기

   
    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(900)
          
    ; no2. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(600)
    ClipPut("1.기관별, 월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(600)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(900)

; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 강북/강원 _클릭
    LogMessage(" 강북/강원 클릭")
    MouseClick("left", 281, 186) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

     
    ; no3. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.강북_월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 강남/서부 _클릭
    LogMessage(" 강남/서부 클릭")
    MouseClick("left", 262, 200) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(800)
    ClipPut("2.강남_월") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(800)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(2000)
; -------

; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭
    Sleep(900)

    ; 부산 _클릭
    LogMessage(" 부산/경남 클릭")
    MouseClick("left", 262, 211) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 975, 162) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.부산") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------


; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 대구 _클릭
    LogMessage(" 대구/경북 클릭")
    MouseClick("left", 273, 219) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.대구") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------

; -------
    ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 전남 _클릭
    LogMessage(" 전남/전북 클릭")
    MouseClick("left", 265, 234) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.전남") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)
; -------
  ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 충남 _클릭
    LogMessage(" 충남/충북 클릭")
    MouseClick("left", 270, 245) ; 저장 버튼 클릭


    ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("2.충남") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)

; -------

   ; 목록_클릭
    LogMessage("전체 옆 목록버튼 클릭")
    MouseClick("left", 333, 158) ; 저장 버튼 클릭

    ; 전체_클릭
    LogMessage("전체 클릭")
    MouseClick("left", 242, 175) ; 저장 버튼 클릭


  ; 사원별_클릭
    LogMessage("사원별 클릭")
    MouseClick("left", 181, 270) ; 저장 버튼 클릭

  ; 검색 클릭
    LogMessage("검색 클릭")
     MouseClick("left", 996, 164) ; 검색 클릭
    Sleep(5000) ; 검색 결과 대기

    ; 엑셀 다운로드 클릭
    LogMessage("엑셀 다운로드 클릭")
    MouseClick("left", 986, 720)
    Sleep(700)

      
    ; no4. 파일 저장
    LogMessage("파일 저장 시작")
    MouseClick("left", 605, 543) ; 파일 이름 입력 필드 클릭
    Sleep(500)
    ClipPut("3.사원") ; 파일 이름 복사
    Send("^v") ; 붙여넣기
    Sleep(500)
    MouseClick("left", 802, 540) ; 저장 버튼 클릭
    Sleep(700)

LogMessage("7단계: 테스트 완료")
MsgBox(0, "완료", "테스트 완료!")
'''

# AutoIt 스크립트 저장
with open(script_path, 'w', encoding='utf-8') as script_file:
    script_file.write(autoit_script_content)
print(f"AutoIt 스크립트가 {script_path}에 저장되었습니다.")

# AutoIt 스크립트 실행
try:
    if os.path.exists(autoit_path):
        result = subprocess.run(
            [autoit_path, script_path],
            capture_output=True,
            text=True,
        )
        print("AutoIt 스크립트 실행 성공")
        print("출력 메시지:", result.stdout)
        print("에러 메시지:", result.stderr)

        # 로그 파일 출력
        if os.path.exists(log_file_path):
            with open(log_file_path, 'r', encoding='utf-8') as log_file:
                print("\n로그 파일 내용:")
                print(log_file.read())
    else:
        print(f"AutoIt 실행 파일을 찾을 수 없습니다: {autoit_path}")
except subprocess.CalledProcessError as e:
    print(f"AutoIt 실행 실패: {e}")
    print(f"에러 메시지:", e.stderr)
except Exception as e:
    print(f"예외가 발생했습니다: {e}")



import pandas as pd
import glob
import os

# 폴더 경로 설정 (일일 데이터와 누계 데이터 경로)
daily_folder_path = r'd:\신규P값총량제\통합'
cumulative_folder_path = r'd:\신규P값총량제\누계통합'

# 파일명에 따른 정렬 기준
file_order = [
    "0. 채널별, 월", "1.기관별, 월",  # "기관별" 파일을 채널 다음으로 정렬
    "2.강북_월", "2.강남_월", "2.부산",
    "2.전남", "2.대구", "2.충남",
    "사원별"
]

# 파일명 정렬 함수
def get_file_priority(file_name):
    for i, keyword in enumerate(file_order):
        if keyword in file_name:
            return i
    return len(file_order)  # 키워드가 없으면 가장 뒤로 정렬

# "제거할 특정 숫자 + 공백 + 건" 패턴을 포함하는 행을 필터링하기 위한 함수 (A열에만 적용)
def filter_out_specific_gun_rows(df):
    remove_patterns = [r'8\s건$', r'12\s건$', r'6\s건$', r'9\s건$', r'7\s건$']
    pattern = '|'.join(remove_patterns)
    return df[~df.iloc[:, 0].astype(str).str.contains(pattern)]

# 폴더에서 파일을 읽어 병합하고 저장하는 함수
def merge_and_save_files(folder_path, output_filename):
    # 모든 파일 경로 가져오기
    all_files = glob.glob(folder_path + "/*.xls")

    # 파일명에 따라 정렬
    all_files.sort(key=lambda x: get_file_priority(os.path.basename(x)))

    # 전체 데이터프레임 초기화
    all_data = pd.DataFrame()
    for i, file in enumerate(all_files):
        try:
            # HTML 형식으로 파일 읽기
            df_list = pd.read_html(file)
            df = df_list[0]  # 첫 번째 테이블만 사용

            # 특정 "숫자 + 공백 + 건" 패턴을 포함하는 행 제거 (A열만)
            df = filter_out_specific_gun_rows(df)

            # 파일명 출력 및 데이터 미리보기
            if i == 0:
                print(f"{file} 데이터 미리보기:\n", df.head())
            elif i == len(all_files) - 1 or "2.기관별_" in file:
                df = df.iloc[1:]  # 헤더 제외
                print(f"{file} 데이터 미리보기:\n 헤더제외 ->", df.head())
            else:
                print(f"{file} 데이터 미리보기:\n", df.head())

            # 첫 번째 파일에서만 헤더를 사용하고, 이후 파일들은 헤더 없이 추가
            if all_data.empty:
                all_data = df
            else:
                # 기존 열들과 현재 파일의 열을 맞추어 병합
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)

        except Exception as e:
            print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")

    # 병합된 데이터를 csv 파일로 저장 (CP949 인코딩 적용)
    if not all_data.empty:
        csv_output_path = os.path.join(folder_path, output_filename)
        all_data.to_csv(csv_output_path, index=False, encoding='cp949')
        print(f"병합된 파일이 {csv_output_path}에 저장되었습니다.")
    else:
        print("병합된 데이터가 비어 있습니다. 엑셀 파일들을 확인하세요.")

# 일일 데이터 처리
merge_and_save_files(daily_folder_path, "신규P값병합11_일일.csv")

# 누계 데이터 처리
merge_and_save_files(cumulative_folder_path, "신규P값병합08월_11월_11_누계.csv")
````

# 병합 작업
def merge_files(folder_path):
    all_files = glob.glob(folder_path + "/*.xls")
    all_data = pd.DataFrame()

    for file in all_files:
        try:
            df_list = pd.read_html(file)
            df = df_list[0]
            all_data = pd.concat([all_data, df], ignore_index=True, sort=False)
        except Exception as e:
            print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")
    return all_data
