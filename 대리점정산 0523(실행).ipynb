```python
import pandas as pd
import os
import chardet
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication
import openpyxl
from openpyxl.styles import PatternFill, Font, Border, Side
import re
import tkinter as tk
from tkinter import simpledialog, messagebox, ttk

class EmailApp:
    def __init__(self, root):
        self.root = root
        self.root.title("대리점 및 이메일 설정")
        self.dealer_info = {
            "(주)미래넷": "82265391@kt.com",
            "구정례": "82265391@kt.com",
            "케이티텔보안네트워크": "82265391@kt.com",
            "커넥션네트워크": "82265391@kt.com",
            "애드라인": "82265391@kt.com",
            "제제컴퍼니": "82265391@kt.com",
            "아이티인프라": "82265391@kt.com"
        }
        self.setup_ui()

    def setup_ui(self):
        frame = tk.Frame(self.root)
        frame.pack(pady=10)

        self.tree = ttk.Treeview(frame, columns=('상호', '이메일'), show='headings')
        self.tree.heading('상호', text='상호')
        self.tree.heading('이메일', text='이메일')
        self.tree.pack(side=tk.LEFT)

        scrollbar = ttk.Scrollbar(frame, orient=tk.VERTICAL, command=self.tree.yview)
        self.tree.configure(yscroll=scrollbar.set)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

        self.load_dealers()

        form_frame = tk.Frame(self.root)
        form_frame.pack(pady=10)

        tk.Label(form_frame, text="상호").grid(row=0, column=0)
        tk.Label(form_frame, text="이메일").grid(row=1, column=0)
        self.entry_name = tk.Entry(form_frame)
        self.entry_name.grid(row=0, column=1)
        self.entry_email = tk.Entry(form_frame)
        self.entry_email.grid(row=1, column=1)

        button_frame = tk.Frame(self.root)
        button_frame.pack(pady=10)
        tk.Button(button_frame, text="추가", command=self.add_dealer).pack(side=tk.LEFT, padx=5)
        tk.Button(button_frame, text="수정", command=self.edit_dealer).pack(side=tk.LEFT, padx=5)
        tk.Button(button_frame, text="삭제", command=self.delete_dealer).pack(side=tk.LEFT, padx=5)
        tk.Button(button_frame, text="저장 및 닫기", command=self.save_and_close).pack(side=tk.LEFT, padx=5)

    def load_dealers(self):
        for name, email in self.dealer_info.items():
            self.tree.insert('', tk.END, values=(name, email))

    def add_dealer(self):
        name = self.entry_name.get().strip()
        email = self.entry_email.get().strip()
        if name and email:
            self.tree.insert('', tk.END, values=(name, email))
            self.dealer_info[name] = email
            self.entry_name.delete(0, tk.END)
            self.entry_email.delete(0, tk.END)

    def edit_dealer(self):
        selected_item = self.tree.selection()
        if selected_item:
            item = self.tree.item(selected_item)
            old_name = item['values'][0]
            old_email = item['values'][1]
            new_name = self.entry_name.get().strip()
            new_email = self.entry_email.get().strip()
            if new_name and new_email:
                self.tree.item(selected_item, values=(new_name, new_email))
                del self.dealer_info[old_name]
                self.dealer_info[new_name] = new_email
                self.entry_name.delete(0, tk.END)
                self.entry_email.delete(0, tk.END)

    def delete_dealer(self):
        selected_item = self.tree.selection()
        if selected_item:
            item = self.tree.item(selected_item)
            name = item['values'][0]
            self.tree.delete(selected_item)
            del self.dealer_info[name]

    def save_and_close(self):
        with open('dealers.csv', 'w') as f:
            for name, email in self.dealer_info.items():
                f.write(f"{name},{email}\n")
        self.root.quit()

# 파일 경로 설정
base_path = r"D:\대리점자동화연습\5월 원본"
output_base_path = os.path.join(base_path, "5월 정산 전송")
os.makedirs(output_base_path, exist_ok=True)

summary_path = os.path.join(base_path, "2024.05월_대리점정산_표(요약).cs")
new_detail_path = os.path.join(base_path, "2024.05월_대리점정산_신규상세.cs")
cancel_detail_path = os.path.join(base_path, "2024.05월_대리점정산_해지상세.cs")

# 파일 인코딩 감지 함수
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

# 파일 존재 여부 확인 및 인코딩 감지
def read_file_with_encoding(file_path, **kwargs):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{file_path} 파일이 존재하지 않습니다.")
    encoding = detect_encoding(file_path)
    return pd.read_csv(file_path, encoding=encoding, **kwargs)

# Function to find header
def find_header(file_path):
    for header in [0, 1, 2]:
        try:
            df = read_file_with_encoding(file_path, header=header)
            if '본부' in df.columns:
                print(f"'본부' column found with header at row {header}")
                return df
            else:
                print(f"'본부' column not found with header at row {header}")
        except Exception as e:
            print(f"Failed to read with header at row {header}: {e}")
    raise ValueError("Unable to find a suitable header containing '본부'")

def main():
    root = tk.Tk()
    app = EmailApp(root)
    root.mainloop()

    sender_email = simpledialog.askstring("Input", "이메일 주소를 입력하세요:")
    sender_password = simpledialog.askstring("Input", "비밀번호를 입력하세요:", show='*')

    df_summary = find_header(summary_path)
    df_new_detail = find_header(new_detail_path)
    df_cancel_detail = find_header(cancel_detail_path)

    df_summary = df_summary.loc[:, ~df_summary.columns.str.contains('Unnamed')]
    df_summary.columns = df_summary.columns.str.strip().str.replace('\n', '')

    required_columns = [
        '정산년월(YYMM)', '지사', '추천자사번', '대표자', '상호명', '사업자등록번호',
        '신규월정료', '신규건수', '모집수수료', '성사율수수료', '완결비중수수료', '완결영업정책수수료', '합계',
        '유지수수료', '판매형', '소급액', '전월미지급액', '전월미환수', '공급가액', '부가세(공급가액10%)', '최종지급액', '해지환수금액(별도세금계산서발행)', '승인번호'
    ]

    if len(df_summary.columns) < len(required_columns):
        for col in required_columns:
            if col not in df_summary.columns:
                df_summary[col] = None

    df_summary = df_summary.fillna('')
    df_new_detail = df_new_detail.fillna('')
    df_cancel_detail = df_cancel_detail.fillna('')

    def fill_missing_values(df, target_row, rows_to_check, cols):
        for col in cols:
            for row in rows_to_check:
                if pd.isna(df.at[target_row, col]):
                    if not pd.isna(df.at[row, col]):
                        df.at[target_row, col] = df.at[row, col]
                        break

    columns_to_copy = df_summary.columns
    fill_missing_values(df_summary, 6, [5, 4], columns_to_copy)

    # 여기서 열 이름을 출력하여 '본부' 열이 존재하는지 확인합니다.
    print(df_summary.columns)

    try:
        df_summary = df_summary[df_summary['본부'].str.contains("강북/강원본부", na=False)]
        df_summary = df_summary[~df_summary['본부'].str.contains('충남/충북본부|강남/서부본부|대구/경북본부|부산/경남본부|전남/전북본부', na=False)]
        df_summary = df_summary.drop(columns=['본부', '프로모션', '비고'], errors='ignore')
    except KeyError as e:
        print(f"Error: {e}")
        print("데이터프레임에 '본부' 열이 존재하지 않습니다. 데이터 구조를 확인해주세요.")
        return

    total_row_index = df_summary[df_summary['상호명'].str.contains("총계", na=False)].index
    if not total_row_index.empty:
        df_summary = df_summary.drop(total_row_index)
        df_summary = df_summary.drop(total_row_index + 1, errors='ignore')

    df_new_detail['서비스번호'] = df_new_detail['서비스번호'].astype(str).str.split('.').str[0]
    df_new_detail['고객번호'] = df_new_detail['고객번호'].astype(str).str.split('.').str[0]
    df_new_detail['계약번호'] = df_new_detail['계약번호'].astype(str).str.split('.').str[0]

    df_cancel_detail['서비스번호'] = df_cancel_detail['서비스번호'].astype(str).str.split('.').str[0]
    df_cancel_detail['고객번호'] = df_cancel_detail['고객번호'].astype(str).str.split('.').str[0]
    df_cancel_detail['계약번호'] = df_cancel_detail['계약번호'].astype(str).str.split('.').str[0]
    agencies = ["(주)미래넷", "구정례", "케이티텔보안네트워크", "커넥션네트워크", "애드라인", "제제컴퍼니", "아이티인프라"]

    dealer_info = {
       "(주)미래넷": "82265391@kt.com",
            "구정례": "82265391@kt.com",
            "케이티텔보안네트워크": "82265391@kt.com",
            "커넥션네트워크": "82265391@kt.com",
            "애드라인": "82265391@kt.com",
            "제제컴퍼니": "82265391@kt.com",
            "아이티인프라": "82265391@kt.com"
    }

    smtp_server = "smtp.naver.com"
    smtp_port = 587

    header_fill = PatternFill(start_color="4F4F4F", end_color="4F4F4F", fill_type="solid")
    header_font = Font(color="FFFFFF", bold=True)
    blue_font = Font(color="0000FF", bold=True)
    red_font = Font(color="FF0000", bold=True)
    thin_border = Border(left=Side(style='thin', color='000000'),
                         right=Side(style='thin', color='000000'),
                         top=Side(style='thin', color='000000'),
                         bottom=Side(style='thin', color='000000'))
    thick_border = Border(left=Side(style='thick', color='000000'),
                          right=Side(style='thick', color='000000'),
                          top=Side(style='thick', color='000000'),
                          bottom=Side(style='thick', color='000000'))

    for agency in agencies:
        summary_filtered = df_summary[df_summary['상호명'].str.contains(re.escape(agency), na=False)].copy()
       
        if summary_filtered.empty:
            print(f"{agency}에 대한 데이터가 없습니다.")
            continue

        summary_filtered = summary_filtered.drop_duplicates(subset=required_columns)
       
        new_detail_filtered = df_new_detail[df_new_detail['유통망대분류'].str.contains(re.escape(agency), na=False)].copy()
        new_detail_filtered = new_detail_filtered.drop(columns=['관리본부명', '서비스번호', '서비스(대)', '서비스(중)', '영업자사번', '추천본부명', '추천자사번', '추천자유형', '비고'], errors='ignore')

        cancel_detail_filtered = df_cancel_detail[df_cancel_detail['유통망대분류'].str.contains(re.escape(agency), na=False)].copy()
        cancel_detail_filtered = cancel_detail_filtered.drop(columns=["대리점 해지 여부"], errors='ignore')

        if summary_filtered['승인번호'].isnull().all():
            cancel_detail_filtered = pd.DataFrame(columns=cancel_detail_filtered.columns)

        summary_filtered['해지환수금액(별도세금계산서발행)'] = summary_filtered['해지환수금액(별도세금계산서발행)'].replace(' - ', '0')
       
        if agency == "(주)미래넷":
            excel_output_path = os.path.join(output_base_path, "(주)미래넷_정산.xlsx")
        else:
            excel_output_path = os.path.join(output_base_path, f"{agency}_정산.xlsx")

        with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
            summary_filtered.to_excel(writer, sheet_name="표(요약)", index=False)
            new_detail_filtered.to_excel(writer, sheet_name="신규상세", index=False)
            if not cancel_detail_filtered.empty:
                cancel_detail_filtered.to_excel(writer, sheet_name="해지상세", index=False)
       
        wb = openpyxl.load_workbook(excel_output_path)
       
        for sheet in wb.sheetnames:
            ws = wb[sheet]
           
            for cell in ws[1]:
                cell.fill = header_fill
                cell.font = header_font
                cell.border = thin_border
           
            if sheet == "표(요약)":
                ws.insert_rows(1)
                ws['A1'] = '합계'
                ws['A1'].font = Font(bold=True)
                ws['U1'] = summary_filtered['최종지급액'].replace({',': ''}, regex=True).astype(float).sum()
                ws['U1'].font = Font(bold=True, color='0000FF')
                ws['U1'].number_format = '#,##0'
                ws['V1'] = summary_filtered['해지환수금액(별도세금계산서발행)'].replace({',': ''}, regex=True).astype(float).sum()
                ws['V1'].font = Font(bold=True, color='FF0000')
                ws['V1'].number_format = '#,##0'

            if sheet == "신규상세":
                ws.insert_rows(1)
                ws['A1'] = '합계'
                ws['A1'].font = Font(bold=True)
                ws['C1'] = new_detail_filtered['계약번호'].nunique()
                ws['C1'].font = Font(bold=True)
                ws['C1'].number_format = '#,##0'
                ws['AB1'] = new_detail_filtered['지급수수료'].replace({',': ''}, regex=True).astype(float).sum()
                ws['AB1'].font = Font(bold=True, color='0000FF')
                ws['AB1'].number_format = '#,##0'

            if sheet == "해지상세":
                ws.insert_rows(1)
                ws['A1'] = '합계'
                ws['A1'].font = Font(bold=True)
                ws['E1'] = cancel_detail_filtered['계약번호'].nunique()
                ws['E1'].font = Font(bold=True)
                ws['E1'].number_format = '#,##0'
                ws['AF1'] = cancel_detail_filtered['환수금액'].replace({',': ''}, regex=True).astype(float).sum()
                ws['AF1'].font = Font(bold=True, color='FF0000')
                ws['AF1'].number_format = '#,##0'
       
        wb.save(excel_output_path)

        receiver_email = dealer_info[agency]
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = receiver_email
        msg['Subject'] = f"4월 정산실적_{agency}_대리점_(표(요약). 신규상세, 해지상세)"

        summary_html = f"""
            <p><b>표(요약):</b></p>
            <p>ㅇ 합계액: {summary_filtered['합계'].replace({',': ''}, regex=True).astype(float).sum():,.0f}</p>
            <p>ㅇ 최종지급액: <span style="color: blue; font-weight: bold;">{summary_filtered['최종지급액'].replace({',': ''}, regex=True).astype(float).sum():,.0f}</span></p>
            {f"<p>ㅇ 해지환수금액: <span style='color: red; font-weight: bold;'>{summary_filtered['해지환수금액(별도세금계산서발행)'].replace({',': ''}, regex=True).astype(float).sum():,.0f}</span></p>" if summary_filtered['해지환수금액(별도세금계산서발행)'].replace({',': ''}, regex=True).astype(float).sum() > 0 else ''}
            {f"<p>ㅇ 승인번호: <span style='color: green; font-weight: bold;'>{', '.join(summary_filtered['승인번호'].dropna().astype(str).unique())}</span></p>" if not summary_filtered['승인번호'].dropna().empty else ''}
            <table border="1" cellspacing="0" cellpadding="5" style="font-size:12px;">
                <tr style="background-color: #4F4F4F; color: #FFFFFF; font-weight: bold;">
                    {''.join([f'<th>{col}</th>' for col in summary_filtered.columns])}
                </tr>
                {''.join([f'<tr>{"".join([f"<td>{cell}</td>" for cell in row])}</tr>' for row in summary_filtered.values])}
            </table>
        """

        new_detail_html = f"""
            <p><b>신규상세:</b> <span style="color: blue; font-weight: bold;">지급수수료 합계: {new_detail_filtered['지급수수료'].replace({',': ''}, regex=True).astype(float).sum():,.0f}</span></p>
            <table border="1" cellspacing="0" cellpadding="5" style="font-size:12px;">
                <tr style="background-color: #4F4F4F; color: #FFFFFF; font-weight: bold;">
                    {''.join([f'<th>{col}</th>' for col in new_detail_filtered.columns])}
                </tr>
                  {''.join([f'<tr>{"".join([f"<td>{cell}</td>" for cell in row])}</tr>' for row in new_detail_filtered.values])}
            </table>
        """

        if not cancel_detail_filtered.empty:
            cancel_detail_html = f"""
                <p><b>[해지상세]</b></p>
                <p><b>ㅇ 환수금액 합계:</b> <span style="color:red; font-weight:bold;">{cancel_detail_filtered["환수금액"].replace({',': ''}, regex=True).astype(float).sum():,.0f}</span></p>
                <table border="1" cellspacing="0" cellpadding="5" style="font-size:12px;">
                    <tr style="background-color: #4F4F4F; color: #FFFFFF; font-weight: bold;">
                        {''.join([f'<th>{col}</th>' for col in cancel_detail_filtered.columns])}
                    </tr>
                    {''.join([f'<tr>{"".join([f"<td>{cell}</td>" for cell in row])}</tr>' for row in cancel_detail_filtered.values])}
                </table>
                """
        else:
            cancel_detail_html = ""

        body = f"""
        <!DOCTYPE html>
        <html>
        <body>
            <p style="font-size:12px;">{agency}님, 안녕하세요.</p>
            <p style="font-size:12px;">kt텔레캅 사외채널 담당자 장재혁입니다.</p>
            <p style="font-size:12px;">대리점 정산실적을 아래와 같이 송부드리니 6월 5일 18시까지 세금계산서 발행 요청드리며 검토 후 특이사항 있으실 경우 연락주시기 바랍니다.</p>
            {summary_html}
            {new_detail_html}
            {cancel_detail_html}
        </body>
        </html>
        """

        msg.attach(MIMEText(body, 'html'))

        with open(excel_output_path, "rb") as attachment:
            part = MIMEApplication(attachment.read(), _subtype="octet-stream")
            part.add_header('Content-Disposition', 'attachment', filename=os.path.basename(excel_output_path))
            msg.attach(part)

        # 이메일 전송 함수
        def send_email():
            try:
                with smtplib.SMTP(smtp_server, smtp_port) as server:
                    server.starttls()
                    server.login(sender_email, sender_password)
                    server.sendmail(sender_email, receiver_email, msg.as_string())
                messagebox.showinfo("성공", f"{agency}에 대한 이메일을 {receiver_email}로 전송했습니다.")
            except smtplib.SMTPAuthenticationError:
                messagebox.showerror("오류", "이메일 또는 비밀번호가 잘못되었습니다. 다시 시도해주세요.")
            except Exception as e:
                messagebox.showerror("오류", f"이메일 전송 중 오류가 발생했습니다: {str(e)}")

        send_email()

    print("모든 파일이 처리되고 이메일이 전송되었습니다.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
import chardet
import re
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication
import openpyxl
from openpyxl.styles import PatternFill, Font, Border, Side
import tkinter as tk
from tkinter import simpledialog, Listbox, MULTIPLE, messagebox, Scrollbar, RIGHT, Y, END, Toplevel, StringVar, Entry, Label, Button
from tkinter.ttk import Progressbar
import threading

# Sample data to represent the initial agent information
agent_data = [
    {"관리지사명": "강북지사", "구역번호": "G000101", "사번": "7864", "사원명": "성진수", "사번1": "7864", "담당 SP": "성진수", "이메일번호": "seong.js@kt.com"},
    {"관리지사명": "강북지사", "구역번호": "G000102", "사번": "85669", "사원명": "김태규", "사번1": "85669", "담당 SP": "김태규", "이메일번호": "#N/A"},
    {"관리지사명": "강북지사", "구역번호": "G000103", "사번": "4159", "사원명": "박영택", "사번1": "4159", "담당 SP": "박영택", "이메일번호": "youngtaek.park@kt.com"},
    {"관리지사명": "강북지사", "구역번호": "G000104", "사번": "5940", "사원명": "김의겸", "사번1": "5940", "담당 SP": "김의겸", "이메일번호": "uikyum.kim@kt.com"},
    {"관리지사명": "강북지사", "구역번호": "G000105", "사번": "4802", "사원명": "사공관", "사번1": "4802", "담당 SP": "사공관", "이메일번호": "ruga.sg@kt.com"},
    {"관리지사명": "강북지사", "구역번호": "G000106", "사번": "3208", "사원명": "김형선", "사번1": "3208", "담당 SP": "김형선", "이메일번호": "82004038@kt.com"},
    {"관리지사명": "서대문지사", "구역번호": "G000201", "사번": "5867", "사원명": "김용", "사번1": "5867", "담당 SP": "김용", "이메일번호": "#N/A"},
    {"관리지사명": "서대문지사", "구역번호": "G000202", "사번": "3646", "사원명": "안근영", "사번1": "3646", "담당 SP": "안근영", "이메일번호": "82004793@kt.com"},
    {"관리지사명": "서대문지사", "구역번호": "G000203", "사번": "6868", "사원명": "이정도", "사번1": "6868", "담당 SP": "이정도", "이메일번호": "jungdo.lee@kt.com"},
    {"관리지사명": "서대문지사", "구역번호": "G000204", "사번": "4161", "사원명": "서은석", "사번1": "4161", "담당 SP": "서은석", "이메일번호": "eunseok.seo@kt.com"},
    {"관리지사명": "서대문지사", "구역번호": "G000205", "사번": "3228", "사원명": "민유기", "사번1": "3228", "담당 SP": "민유기", "이메일번호": "82004091@kt.com"},
    {"관리지사명": "고양지사", "구역번호": "G000301", "사번": "5455", "사원명": "성형래", "사번1": "5455", "담당 SP": "성형래", "이메일번호": "hyeongrae.seong@kt.com"},
    {"관리지사명": "고양지사", "구역번호": "G000302", "사번": "4564", "사원명": "강경필", "사번1": "4564", "담당 SP": "강경필", "이메일번호": "82004653@kt.com"},
    {"관리지사명": "고양지사", "구역번호": "G000303", "사번1": "4643", "사원명": "유창준", "사번1": "4643", "담당 SP": "유창준", "이메일번호": "changjun.yoo@kt.com"},
    {"관리지사명": "고양지사", "구역번호": "G000304", "사번1": "6028", "사원명": "최우석", "사번1": "6028", "담당 SP": "최우석", "이메일번호": "useok.choi@kt.com"},
    {"관리지사명": "고양지사", "구역번호": "G000305", "사번1": "5002", "사원명": "김현교", "사번1": "5002", "담당 SP": "김현교", "이메일번호": "82002883@kt.com"},
    {"관리지사명": "고양지사", "구역번호": "G000306", "사번1": "5189", "사원명": "박민준", "사번1": "5189", "담당 SP": "박민준", "이메일번호": "82004316@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000401", "사번1": "4415", "사원명": "최태식", "사번1": "4415", "담당 SP": "최태식", "이메일번호": "82004907@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000402", "사번1": "4251", "사원명": "김병조", "사번1": "4251", "담당 SP": "김병조", "이메일번호": "82003769@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000403", "사번1": "4637", "사원명": "오광훈", "사번1": "4637", "담당 SP": "오광훈", "이메일번호": "82004734@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000404", "사번1": "5750", "사원명": "이재용", "사번1": "5750", "담당 SP": "이재용", "이메일번호": "jae.yong@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000405", "사번1": "3947", "사원명": "남기민", "사번1": "3947", "담당 SP": "남기민", "이메일번호": "kimin.nam@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000406", "사번1": "5360", "사원명": "권대호", "사번1": "5360", "담당 SP": "권대호", "이메일번호": "82005897@kt.com"},
    {"관리지사명": "중앙지사", "구역번호": "G000407", "사번1": "6870", "사원명": "문성진", "사번1": "6870", "담당 SP": "문성진", "이메일번호": "sj2.moon@kt.com"},
    {"관리지사명": "의정부지사", "구역번호": "G000501", "사번1": "3085", "사원명": "정성록", "사번1": "3085", "담당 SP": "정성록", "이메일번호": "#N/A"},
    {"관리지사명": "의정부지사", "구역번호": "G000503", "사번1": "4571", "사원명": "차현갑", "사번1": "4571", "담당 SP": "차현갑", "이메일번호": "82004661@kt.com"},
    {"관리지사명": "의정부지사", "구역번호": "G000504", "사번1": "4300", "사원명": "김진회", "사번1": "4300", "담당 SP": "김진회", "이메일번호": "jinhoe.kim@kt.com"},
    {"관리지사명": "의정부지사", "구역번호": "G000505", "사번1": "4318", "사원명": "임경석", "사번1": "4318", "담당 SP": "임경석", "이메일번호": "kyoungsuk.lim@kt.com"},
    {"관리지사명": "남양주지사", "구역번호": "G000601", "사번1": "3736", "사원명": "김성배", "사번1": "3736", "담당 SP": "김성배", "이메일번호": "soung.bae@kt.com"},
    {"관리지사명": "남양주지사", "구역번호": "G000602", "사번1": "5352", "사원명": "조형근", "사번1": "5352", "담당 SP": "조형근", "이메일번호": "#N/A"},
    {"관리지사명": "남양주지사", "구역번호": "G000603", "사번1": "6277", "사원명": "이한구", "사번1": "6277", "담당 SP": "이한구", "이메일번호": "82141316@kt.com"},
    {"관리지사명": "남양주지사", "구역번호": "G000604", "사번1": "4834", "사원명": "이현배", "사번1": "4834", "담당 SP": "이현배", "이메일번호": "hyounbae.lee@kt.com"},
    {"관리지사명": "남양주지사", "구역번호": "G000605", "사번1": "85408", "사원명": "김현돈", "사번1": "85408", "담당 SP": "김현돈", "이메일번호": "hyundon.kim@kt.com"},
    {"관리지사명": "강릉지사", "구역번호": "P000101", "사번1": "5916", "사원명": "최정훈", "사번1": "5916", "담당 SP": "최정훈", "이메일번호": "82077078@kt.com"},
    {"관리지사명": "강릉지사", "구역번호": "P000102", "사번1": "7447", "사원명": "권문석", "사번1": "7447", "담당 SP": "권문석", "이메일번호": "#N/A"},
    {"관리지사명": "강릉지사", "구역번호": "P000103", "사번1": "6086", "사원명": "양현성", "사번1": "6086", "담당 SP": "양현성", "이메일번호": "82107343@kt.com"},
    {"관리지사명": "강릉지사", "구역번호": "P000104", "사번1": "3540", "사원명": "김완용", "사번1": "3540", "담당 SP": "김완용", "이메일번호": "wanyong.kim@kt.com"},
    {"관리지사명": "원주지사", "구역번호": "P000201", "사번1": "85570", "사원명": "손용수", "사번1": "85570", "담당 SP": "손용수", "이메일번호": "sys7779.son@kt.com"},
    {"관리지사명": "원주지사", "구역번호": "P000202", "사번1": "7029", "사원명": "김상태", "사번1": "7029", "담당 SP": "김상태", "이메일번호": "sang_tae.kim@kt.com"},
    {"관리지사명": "원주지사", "구역번호": "P000206", "사번1": "3705", "사원명": "김광호", "사번1": "3705", "담당 SP": "김광호", "이메일번호": "kwangho12.kim@kt.com"},
    {"관리지사명": "원주지사", "구역번호": "P000207", "사번1": "5994", "사원명": "경규원", "사번1": "5994", "담당 SP": "경규원", "이메일번호": "82096643@kt.com"}
]

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

def read_file_with_encoding(file_path, header_row, **kwargs):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{file_path} 파일이 존재하지 않습니다.")
    encoding = detect_encoding(file_path)
    return pd.read_csv(file_path, encoding=encoding, header=header_row, **kwargs)

def send_emails(selected_agents, email_addresses, email_body_template, sender_email, sender_password, progress, root):
    smtp_server = "smtp.naver.com"
    smtp_port = 587

    header_fill = PatternFill(start_color="4F4F4F", end_color="4F4F4F", fill_type="solid")
    header_font = Font(color="FFFFFF", bold=True)
    thin_border = Border(left=Side(style='thin', color='000000'),
                         right=Side(style='thin', color='000000'),
                         top=Side(style='thin', color='000000'),
                         bottom=Side(style='thin', color='000000'))

    summary_path = 'D:\\체납\\이화영\\체납활동영업구역별 체납2회이상 활동요약.cs'
   
    header_row = 0

    df_summary = read_file_with_encoding(summary_path, header_row)

    df_summary = df_summary.loc[:, ~df_summary.columns.str.contains('Unnamed')]
    df_summary.columns = df_summary.columns.str.strip().str.replace('\n', '')

    missing_emails = []
    total_agents = len(selected_agents)
   
    for idx, agent in enumerate(selected_agents):
        agent_email = email_addresses.get(agent)
        if not agent_email or agent_email == "#N/A":
            missing_emails.append(agent)
            continue

        summary_filtered = df_summary[df_summary['영업구역담당자'] == agent].copy()

        if summary_filtered.empty:
            print(f"{agent}에 대한 데이터가 없습니다.")
            continue

        summary_html = summary_filtered.to_html(index=False)

        excel_output_path = f"D:\\체납\\이화영\\{agent}_체납활동보고서.xlsx"
       
        with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
            summary_filtered.to_excel(writer, sheet_name="체납활동요약", index=False)
       
        wb = openpyxl.load_workbook(excel_output_path)
       
        for sheet in wb.sheetnames:
            ws = wb[sheet]
            for cell in ws[1]:
                cell.fill = header_fill
                cell.font = header_font
                cell.border = thin_border
       
        wb.save(excel_output_path)
       
        email_body = f"""
        <!DOCTYPE html>
        <html>
        <body>
            <p style="font-size:12px;">{agent}님, 안녕하세요.</p>
            <p style="font-size:12px;">사업지원팀 체납 담당자입니다.</p>
            <p style="font-size:12px;">최우선 체납활동 대상(해지시설 등 일부 제외)을 아래와 같이 송부드리니</p>
            <p style="font-size:12px;">납부될수 있도록 활동해 주시고 체납활동일자, 수납금액, 특이사항 등 관리하여 주시고.</p>
            <p style="font-size:12px;">자료 요청(6월20일 요청예정)시 활동내역 자료 요청시 제출해주시기 바랍니다.</p>
            {summary_html}
        </body>
        </html>
        """
       
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = agent_email
        msg['Subject'] = f"{agent}님의 체납 활동 보고서"
        msg.attach(MIMEText(email_body, 'html', 'utf-8'))

        attach_files = [
            excel_output_path
        ]

        for file in attach_files:
            with open(file, 'rb') as f:
                part = MIMEApplication(f.read(), Name=os.path.basename(file))
            part['Content-Disposition'] = f'attachment; filename="{os.path.basename(file)}"'
            msg.attach(part)

        try:
            with smtplib.SMTP(smtp_server, smtp_port) as server:
                server.starttls()
                server.login(sender_email, sender_password)
                server.sendmail(sender_email, agent_email, msg.as_string())
                print(f"{agent}에게 이메일을 성공적으로 보냈습니다.")
        except smtplib.SMTPAuthenticationError as e:
            print(f"이메일 로그인 인증에 실패했습니다. 이메일 주소와 비밀번호를 확인하세요: {e}")
            break
        except Exception as e:
            print(f"{agent}에게 이메일 전송에 실패했습니다: {e}")

        # Update progress bar
        progress['value'] = ((idx + 1) / total_agents) * 100
        root.update_idletasks()

    if missing_emails:
        print("이메일 주소를 찾을 수 없는 담당자들:")
        for agent in missing_emails:
            print(f" - {agent}")

def start_sending(selected_agents, email_addresses, email_body_template, sender_email, sender_password, root):
    progress_window = Toplevel(root)
    progress_window.title("Progress")
    tk.Label(progress_window, text="Sending emails, please wait...").pack(pady=10)
    progress = Progressbar(progress_window, orient=tk.HORIZONTAL, length=300, mode='determinate')
    progress.pack(pady=20)
   
    # Start sending emails in a separate thread
    threading.Thread(target=send_emails, args=(selected_agents, email_addresses, email_body_template, sender_email, sender_password, progress, root)).start()

def main():
    root = tk.Tk()
    root.withdraw()

    sender_email = "bough38@naver.com"
    sender_password = simpledialog.askstring("Input", "이메일 비밀번호를 입력하세요:", show='*')

    agents = [agent["사원명"] for agent in agent_data]
    email_addresses = {agent["사원명"]: agent["이메일번호"] for agent in agent_data}

    def add_agent():
        new_window = Toplevel(root)
        new_window.title("추가할 담당자의 정보 입력")

        def save_agent():
            agent_info = {
                "관리지사명": entry_office.get(),
                "구역번호": entry_zone.get(),
                "사번": entry_id.get(),
                "사원명": entry_name.get(),
                "사번1": entry_id1.get(),
                "담당 SP": entry_sp.get(),
                "이메일번호": entry_email.get()
            }
            agent_data.append(agent_info)
            agents.append(agent_info["사원명"])
            email_addresses[agent_info["사원명"]] = agent_info["이메일번호"]
            checklist.insert(tk.END, agent_info["사원명"])
            new_window.destroy()
            update_selected_agents()

        Label(new_window, text="관리지사명").grid(row=0, column=0)
        entry_office = Entry(new_window)
        entry_office.grid(row=0, column=1)

        Label(new_window, text="구역번호").grid(row=1, column=0)
        entry_zone = Entry(new_window)
        entry_zone.grid(row=1, column=1)

        Label(new_window, text="사번").grid(row=2, column=0)
        entry_id = Entry(new_window)
        entry_id.grid(row=2, column=1)

        Label(new_window, text="사원명").grid(row=3, column=0)
        entry_name = Entry(new_window)
        entry_name.grid(row=3, column=1)

        Label(new_window, text="사번1").grid(row=4, column=0)
        entry_id1 = Entry(new_window)
        entry_id1.grid(row=4, column=1)

        Label(new_window, text="담당 SP").grid(row=5, column=0)
        entry_sp = Entry(new_window)
        entry_sp.grid(row=5, column=1)

        Label(new_window, text="이메일번호").grid(row=6, column=0)
        entry_email = Entry(new_window)
        entry_email.grid(row=6, column=1)

        Button(new_window, text="저장", command=save_agent).grid(row=7, column=0, columnspan=2)

    def edit_agent():
        selected_indices = checklist.curselection()
        if not selected_indices:
            messagebox.showwarning("Warning", "수정할 담당자를 선택하세요.")
            return
        index = selected_indices[0]
        agent_name = checklist.get(index)
        agent_info = next(agent for agent in agent_data if agent["사원명"] == agent_name)

        edit_window = Toplevel(root)
        edit_window.title("수정할 담당자의 정보 입력")

        def save_changes():
            agent_info["관리지사명"] = entry_office.get()
            agent_info["구역번호"] = entry_zone.get()
            agent_info["사번"] = entry_id.get()
            agent_info["사원명"] = entry_name.get()
            agent_info["사번1"] = entry_id1.get()
            agent_info["담당 SP"] = entry_sp.get()
            agent_info["이메일번호"] = entry_email.get()

            agents[index] = agent_info["사원명"]
            email_addresses[agent_info["사원명"]] = agent_info["이메일번호"]
            checklist.delete(index)
            checklist.insert(index, agent_info["사원명"])
            edit_window.destroy()
            update_selected_agents()

        Label(edit_window, text="관리지사명").grid(row=0, column=0)
        entry_office = Entry(edit_window)
        entry_office.insert(0, agent_info["관리지사명"])
        entry_office.grid(row=0, column=1)

        Label(edit_window, text="구역번호").grid(row=1, column=0)
        entry_zone = Entry(edit_window)
        entry_zone.insert(0, agent_info["구역번호"])
        entry_zone.grid(row=1, column=1)

        Label(edit_window, text="사번").grid(row=2, column=0)
        entry_id = Entry(edit_window)
        entry_id.insert(0, agent_info["사번"])
        entry_id.grid(row=2, column=1)

        Label(edit_window, text="사원명").grid(row=3, column=0)
        entry_name = Entry(edit_window)
        entry_name.insert(0, agent_info["사원명"])
        entry_name.grid(row=3, column=1)

        Label(edit_window, text="사번1").grid(row=4, column=0)
        entry_id1 = Entry(edit_window)
        entry_id1.insert(0, agent_info["사번1"])
        entry_id1.grid(row=4, column=1)

        Label(edit_window, text="담당 SP").grid(row=5, column=0)
        entry_sp = Entry(edit_window)
        entry_sp.insert(0, agent_info["담당 SP"])
        entry_sp.grid(row=5, column=1)

        Label(edit_window, text="이메일번호").grid(row=6, column=0)
        entry_email = Entry(edit_window)
        entry_email.insert(0, agent_info["이메일번호"])
        entry_email.grid(row=6, column=1)

        Button(edit_window, text="저장", command=save_changes).grid(row=7, column=0, columnspan=2)

    def delete_agent():
        selected_indices = checklist.curselection()
        for index in reversed(selected_indices):
            agent_name = checklist.get(index)
            agents.remove(agent_name)
            email_addresses.pop(agent_name, None)
            agent_data[:] = [agent for agent in agent_data if agent["사원명"] != agent_name]
            checklist.delete(index)
        update_selected_agents()

    def update_selected_agents():
        selected_agents_listbox.delete(0, tk.END)
        selected_indices = checklist.curselection()
        for i in selected_indices:
            selected_agents_listbox.insert(tk.END, f"{checklist.get(i)} ({email_addresses.get(checklist.get(i), 'N/A')})")

    def select_all():
        checklist.select_set(0, tk.END)
        update_selected_agents()

    def deselect_all():
        checklist.select_clear(0, tk.END)
        update_selected_agents()

    def send_selected():
        selected_indices = checklist.curselection()
        selected_agents = [agents[i] for i in selected_indices]
        if selected_agents:
            email_body_template = email_body_var.get()
            start_sending(selected_agents, email_addresses, email_body_template, sender_email, sender_password, root)
        else:
            messagebox.showwarning("Warning", "적어도 한 명의 담당자를 선택하세요.")

    def preview_email():
        selected_indices = checklist.curselection()
        selected_agents = [agents[i] for i in selected_indices]
        if selected_agents:
            email_body_template = email_body_var.get()
            preview_window = Toplevel(root)
            preview_window.title("Email Preview")
            email_preview = tk.Text(preview_window, wrap=tk.WORD)
            email_preview.pack(expand=True, fill=tk.BOTH)
            for agent in selected_agents:
                email_body = email_body_template.replace("{agent}", agent)
                email_preview.insert(tk.END, f"To: {email_addresses.get(agent, 'N/A')}\n\n{email_body}\n\n{'-'*50}\n\n")
        else:
            messagebox.showwarning("Warning", "적어도 한 명의 담당자를 선택하세요.")

    root.deiconify()
    root.title("Select Agents to Send Emails")
    root.geometry("1000x600")
   
    left_frame = tk.Frame(root)
    left_frame.pack(side=tk.LEFT, padx=10, pady=10, fill=tk.BOTH, expand=True)
   
    right_frame = tk.Frame(root)
    right_frame.pack(side=tk.RIGHT, padx=10, pady=10, fill=tk.BOTH, expand=True)
   
    tk.Label(left_frame, text="메일을 보낼 영업구역담당자를 선택하세요:", font=("Helvetica", 12)).pack(pady=10)
   
    left_inner_frame = tk.Frame(left_frame)
    left_inner_frame.pack(pady=10, fill=tk.BOTH, expand=True)

    scrollbar = Scrollbar(left_inner_frame, orient=tk.VERTICAL)
    checklist = Listbox(left_inner_frame, selectmode=MULTIPLE, yscrollcommand=scrollbar.set, font=("Helvetica", 10))
    scrollbar.config(command=checklist.yview)
    scrollbar.pack(side=RIGHT, fill=Y)
    checklist.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
   
    for agent in agents:
        checklist.insert(tk.END, agent)

    button_frame = tk.Frame(left_frame)
    button_frame.pack(pady=5)
    tk.Button(button_frame, text="전체 선택", command=select_all, font=("Helvetica", 10)).pack(side=tk.LEFT, padx=5)
    tk.Button(button_frame, text="전체 취소", command=deselect_all, font=("Helvetica", 10)).pack(side=tk.LEFT, padx=5)

    tk.Button(left_frame, text="추가", command=add_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="수정", command=edit_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="삭제", command=delete_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="보내기", command=send_selected, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="미리보기", command=preview_email, font=("Helvetica", 10)).pack(pady=5)

    tk.Label(right_frame, text="실제 메일을 받을 담당자:", font=("Helvetica", 12)).pack(pady=10)
   
    selected_agents_listbox = Listbox(right_frame, font=("Helvetica", 10))
    selected_agents_listbox.pack(fill=tk.BOTH, expand=True)
   
    tk.Label(right_frame, text="메일 내용:", font=("Helvetica", 12)).pack(pady=10)
   
    email_body_var = StringVar()
    email_body_var.set("""<!DOCTYPE html>
<html>
<body>
    <p style="font-size:12px;">{agent}님, 안녕하세요.</p>
    <p style="font-size:12px;">사업지원팀 체납 담당자입니다.</p>
    <p style="font-size:12px;">최우선 체납활동 대상(해지시설 등 일부 제외)을 아래와 같이 송부드리니</p>
    <p style="font-size:12px;">납부될수 있도록 활동해 주시고 체납활동일자, 수납금액, 특이사항 등 관리하여 주시고.</p>
    <p style="font-size:12px;">자료 요청(6월20일 요청예정)시 활동내역 자료 요청시 제출해주시기 바랍니다.</p>
</body>
</html>
""")
    email_body_entry = tk.Entry(right_frame, textvariable=email_body_var, font=("Helvetica", 10), width=50)
    email_body_entry.pack(pady=5)

    checklist.bind("<<ListboxSelect>>", lambda event: update_selected_agents())

    # Exit after 10 minutes (600000 ms)
    root.after(600000, lambda: root.quit())
    root.mainloop()

if __name__ == "__main__":
    main()
```
