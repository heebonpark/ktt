````python
아래 코드는 Jupyter Notebook에서 수동으로 좌표를 찍고, 드래그 혹은 클릭을 통해 텍스트를 선택하고 복사한 뒤, 자동화를 적용하는 테스트입니다.

---

### **1. 좌표 찍기 및 수동 드래그**
아래 코드는 마우스 클릭 좌표를 기록한 뒤, 드래그를 수행하고 텍스트를 복사합니다.

#### **Jupyter Notebook 코드**

```python
import pyautogui
import time
import pyperclip

# 좌표 기록 함수
def record_mouse_position():
    print("마우스를 원하는 위치에 놓고 클릭하세요.")
    time.sleep(3)
    position = pyautogui.position()
    print(f"기록된 좌표: {position}")
    return position

# 드래그 및 복사
def drag_and_copy_text(start_pos, end_pos):
    print("드래그를 시작합니다.")
    pyautogui.moveTo(start_pos.x, start_pos.y, duration=1)
    pyautogui.dragTo(end_pos.x, end_pos.y, duration=1)  # 드래그
    print("드래그 완료.")

    print("텍스트를 복사합니다.")
    pyautogui.hotkey('ctrl', 'c')  # 복사
    time.sleep(1)

    # 클립보드 텍스트 확인
    copied_text = pyperclip.paste()
    print(f"복사된 텍스트: {copied_text}")
    return copied_text

# 1️⃣ 시작 좌표 기록
print("드래그 시작 지점을 클릭하세요.")
start_position = record_mouse_position()

# 2️⃣ 끝 좌표 기록
print("드래그 끝 지점을 클릭하세요.")
end_position = record_mouse_position()

# 3️⃣ 드래그 및 복사 실행
copied_text = drag_and_copy_text(start_position, end_position)
```

---

### **2. 자동화 적용 테스트**
드래그된 영역의 이미지를 OCR로 읽고, 텍스트를 자동으로 붙여넣기 합니다.

#### **OCR 및 자동화 코드**
아래 코드는 OCR로 텍스트를 인식한 뒤, 해당 텍스트를 자동으로 복사/붙여넣기합니다.

```python
import pytesseract
from PIL import ImageGrab

# Tesseract OCR 경로 설정 (Windows 사용자만 필요)
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# OCR로 텍스트 읽기 및 붙여넣기
def capture_ocr_and_paste(start_pos, end_pos):
    # 드래그된 영역 캡처
    bbox = (start_pos.x, start_pos.y, end_pos.x, end_pos.y)
    image = ImageGrab.grab(bbox)
    image.show()  # 캡처된 이미지 확인

    # OCR로 텍스트 추출
    extracted_text = pytesseract.image_to_string(image, lang='kor+eng')
    print(f"OCR 추출된 텍스트:\n{extracted_text}")

    # 텍스트 붙여넣기
    pyperclip.copy(extracted_text)
    pyautogui.click(1000, 500)  # 붙여넣을 위치로 이동 (수동으로 설정 필요)
    pyautogui.hotkey('ctrl', 'v')  # 붙여넣기
    print("텍스트가 자동으로 붙여넣어졌습니다.")

# OCR 및 붙여넣기 실행
capture_ocr_and_paste(start_position, end_position)
```

---

### **3. 전체 워크플로우**
다음은 전체 프로세스를 하나로 통합한 코드입니다.

```python
# 전체 자동화 워크플로우
def full_workflow():
    # 1️⃣ 시작 좌표 기록
    print("드래그 시작 지점을 클릭하세요.")
    start_position = record_mouse_position()

    # 2️⃣ 끝 좌표 기록
    print("드래그 끝 지점을 클릭하세요.")
    end_position = record_mouse_position()

    # 3️⃣ 드래그 및 복사
    copied_text = drag_and_copy_text(start_position, end_position)

    # 4️⃣ OCR 및 붙여넣기
    capture_ocr_and_paste(start_position, end_position)

# 전체 워크플로우 실행
full_workflow()
```

---

### **4. 사용 방법**
1. 코드를 실행하면 마우스 클릭 위치를 기다립니다.
2. 시작 지점과 끝 지점을 클릭하여 드래그 영역을 지정합니다.
3. 드래그된 영역의 텍스트를 복사하거나 OCR로 읽습니다.
4. 자동으로 읽은 텍스트를 지정된 위치에 붙여넣습니다.

---

### **5. 테스트 환경**
1. 텍스트가 포함된 화면에서 위 코드를 실행합니다.
2. 드래그를 수동으로 지정한 뒤, OCR 결과를 확인합니다.
3. 최종적으로 자동화된 텍스트 복사/붙여넣기를 수행합니다.

---

이 코드를 사용하여 첫 번째 수동 녹화 후 자동화 적용 테스트를 진행해 보세요. 필요한 부분이 있다면 추가적으로 조정해 드리겠습니다.
````
