```python
import win32com.client as win32
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# í˜„ì¬ ì‹¤í–‰ ì¤‘ì¸ ì—‘ì…€ ì¸ìŠ¤í„´ìŠ¤ ê°€ì ¸ì˜¤ê¸°
try:
    excel = win32.GetActiveObject("Excel.Application")
except Exception as e:
    print("âš ï¸ ì‹¤í–‰ ì¤‘ì¸ ì—‘ì…€ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤. ë¨¼ì € ì—‘ì…€ íŒŒì¼ì„ ì—´ì–´ì£¼ì„¸ìš”.")
    exit()

# í˜„ì¬ í™œì„±í™”ëœ ì—‘ì…€ì—ì„œ ì²« ë²ˆì§¸ ì›Œí¬ë¶ ê°€ì ¸ì˜¤ê¸°
wb = excel.ActiveWorkbook

# "ì·¨í•©" ì‹œíŠ¸ ê°€ì ¸ì˜¤ê¸°
try:
    ws_summary = wb.Sheets("ì·¨í•©")
except Exception as e:
    print("âš ï¸ 'ì·¨í•©' ì‹œíŠ¸ê°€ ì—†ìŠµë‹ˆë‹¤.")
    exit()

# "ì·¨í•©" ì‹œíŠ¸ì—ì„œ "ìƒí˜¸" (Eì—´) ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
last_row = ws_summary.Cells(ws_summary.Rows.Count, 5).End(-4162).Row  # Eì—´ ë§ˆì§€ë§‰ í–‰ ì°¾ê¸°
business_names = [ws_summary.Cells(row, 5).Value for row in range(2, last_row + 1) if ws_summary.Cells(row, 5).Value]

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(business_names, columns=["ìƒí˜¸"])

# ë¨¸ì‹ ëŸ¬ë‹ ê¸°ë°˜ ìœ ì‚¬ë„ ë¶„ì„ ëª¨ë¸ (BERT í™œìš©)
model = SentenceTransformer("paraphrase-MiniLM-L6-v2")

# ìƒí˜¸ëª…ì„ ì„ë² ë”© ë²¡í„°ë¡œ ë³€í™˜
embeddings = model.encode(df["ìƒí˜¸"].tolist(), convert_to_tensor=True)

# ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°
similarity_matrix = cosine_similarity(embeddings)

# ìœ ì‚¬í•œ ê·¸ë£¹ ì°¾ê¸° (ê¸°ì¤€: 0.7 ì´ìƒ ìœ ì‚¬í•œ í•­ëª©)
threshold = 0.7
grouped_data = []
visited = set()

for i in range(len(df)):
    if i in visited:
        continue
    similar_group = [df.iloc[i]["ìƒí˜¸"]]
    visited.add(i)
   
    for j in range(i + 1, len(df)):
        if j in visited:
            continue
        if similarity_matrix[i][j] > threshold:
            similar_group.append(df.iloc[j]["ìƒí˜¸"])
            visited.add(j)
   
    grouped_data.append(similar_group)

# ìƒˆë¡œìš´ ì‹œíŠ¸ ìƒì„±
new_sheet_name = "ìœ ì‚¬í•œ ìƒí˜¸"
try:
    ws_new = wb.Sheets(new_sheet_name)
except:
    ws_new = wb.Sheets.Add()
    ws_new.Name = new_sheet_name

# ê¸°ì¡´ ë°ì´í„° ì‚­ì œ
ws_new.Cells.Clear()

# ìœ ì‚¬í•œ ê·¸ë£¹ì„ ì—‘ì…€ì— ì €ì¥
ws_new.Cells(1, 1).Value = "ê·¸ë£¹ ë²ˆí˜¸"
ws_new.Cells(1, 2).Value = "ìœ ì‚¬í•œ ìƒí˜¸ëª…"

row_num = 2
for group_idx, group in enumerate(grouped_data, start=1):
    ws_new.Cells(row_num, 1).Value = f"ê·¸ë£¹ {group_idx}"
    for name in group:
        ws_new.Cells(row_num, 2).Value = name
        row_num += 1

print("âœ… ìœ ì‚¬í•œ ìƒí˜¸ëª…ì´ ìƒˆë¡œìš´ ì‹œíŠ¸ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤!")
```

```python
import win32com.client as win32
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import time

# ğŸš€ ì‹¤í–‰ ì‹œê°„ ì¸¡ì • ì‹œì‘
start_time = time.time()

# âœ… í˜„ì¬ ì‹¤í–‰ ì¤‘ì¸ ì—‘ì…€ ì¸ìŠ¤í„´ìŠ¤ ê°€ì ¸ì˜¤ê¸°
try:
    excel = win32.GetActiveObject("Excel.Application")
    excel.ScreenUpdating = False  # UI ì—…ë°ì´íŠ¸ ë¹„í™œì„±í™” (ì†ë„ í–¥ìƒ)
    excel.DisplayAlerts = False   # ê²½ê³  ë¹„í™œì„±í™”
except Exception as e:
    print("âš ï¸ ì‹¤í–‰ ì¤‘ì¸ ì—‘ì…€ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤. ë¨¼ì € ì—‘ì…€ íŒŒì¼ì„ ì—´ì–´ì£¼ì„¸ìš”.")
    exit()

# âœ… í˜„ì¬ í™œì„±í™”ëœ ì—‘ì…€ì—ì„œ ì²« ë²ˆì§¸ ì›Œí¬ë¶ ê°€ì ¸ì˜¤ê¸°
wb = excel.ActiveWorkbook

# âœ… "ì·¨í•©" ì‹œíŠ¸ ê°€ì ¸ì˜¤ê¸°
try:
    ws_summary = wb.Sheets("ì·¨í•©")
except Exception as e:
    print("âš ï¸ 'ì·¨í•©' ì‹œíŠ¸ê°€ ì—†ìŠµë‹ˆë‹¤.")
    exit()

# âœ… "ì·¨í•©" ì‹œíŠ¸ì—ì„œ "ìƒí˜¸" (Eì—´) ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
last_row = ws_summary.Cells(ws_summary.Rows.Count, 5).End(-4162).Row  # Eì—´ ë§ˆì§€ë§‰ í–‰ ì°¾ê¸°
business_names = [ws_summary.Cells(row, 5).Value for row in range(2, last_row + 1) if ws_summary.Cells(row, 5).Value]

# âœ… ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(business_names, columns=["ìƒí˜¸"])

# âœ… ë¨¸ì‹ ëŸ¬ë‹ ê¸°ë°˜ ìœ ì‚¬ë„ ë¶„ì„ ëª¨ë¸ (ë¹ ë¥¸ ëª¨ë¸ ì‚¬ìš©)
model = SentenceTransformer("all-MiniLM-L6-v2", device="cpu")

# âœ… ìƒí˜¸ëª…ì„ ë²¡í„° ë³€í™˜ (ë°°ì¹˜ ì²˜ë¦¬í•˜ì—¬ ì†ë„ ìµœì í™”)
embeddings = model.encode(df["ìƒí˜¸"].tolist(), convert_to_numpy=True, batch_size=32)

# âœ… ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚° (NumPy ë°°ì—´ ìµœì í™”)
similarity_matrix = cosine_similarity(embeddings)

# âœ… ìœ ì‚¬í•œ ê·¸ë£¹ ì°¾ê¸° (ê¸°ì¤€: 0.75 ì´ìƒ ìœ ì‚¬í•œ í•­ëª©ë§Œ ê·¸ë£¹í™”)
threshold = 0.75
grouped_data = []
visited = set()

for i in range(len(df)):
    if i in visited:
        continue
    similar_group = [df.iloc[i]["ìƒí˜¸"]]
    visited.add(i)
   
    for j in range(i + 1, len(df)):
        if j in visited:
            continue
        if similarity_matrix[i][j] > threshold:
            similar_group.append(df.iloc[j]["ìƒí˜¸"])
            visited.add(j)
   
    grouped_data.append(similar_group)

# âœ… ìƒˆë¡œìš´ ì‹œíŠ¸ ìƒì„± (ì—†ìœ¼ë©´ ì¶”ê°€)
new_sheet_name = "ìœ ì‚¬í•œ ìƒí˜¸"
try:
    ws_new = wb.Sheets(new_sheet_name)
except:
    ws_new = wb.Sheets.Add()
    ws_new.Name = new_sheet_name

# âœ… ê¸°ì¡´ ë°ì´í„° ì‚­ì œ
ws_new.Cells.Clear()

# âœ… ìœ ì‚¬í•œ ê·¸ë£¹ì„ ì—‘ì…€ì— ì €ì¥ (í•œ ë²ˆì— ì €ì¥)
data_to_write = [("ê·¸ë£¹ ë²ˆí˜¸", "ìœ ì‚¬í•œ ìƒí˜¸ëª…")]

row_num = 2
for group_idx, group in enumerate(grouped_data, start=1):
    for name in group:
        data_to_write.append((f"ê·¸ë£¹ {group_idx}", name))

# âœ… Excelì— í•œ ë²ˆì— ë°ì´í„° ì“°ê¸° (ì…€ ë‹¨ìœ„ ë°˜ë³µ ìµœì†Œí™”)
for row_idx, (group, name) in enumerate(data_to_write, start=1):
    ws_new.Cells(row_idx, 1).Value = group
    ws_new.Cells(row_idx, 2).Value = name

# âœ… Excel ì‘ì—… ì•ˆì •í™”
excel.ScreenUpdating = True
excel.DisplayAlerts = True

# ğŸš€ ì‹¤í–‰ ì‹œê°„ ì¶œë ¥
print(f"âœ… ìœ ì‚¬í•œ ìƒí˜¸ëª…ì´ ìƒˆë¡œìš´ ì‹œíŠ¸ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤! ì‹¤í–‰ ì‹œê°„: {time.time() - start_time:.2f}ì´ˆ")
```
