## 엑셀의 **"취합" 시트에서 "상호" 열(E열)을 기준으로 유사한 상호를 찾고, 딥러닝/머신러닝을 활용하여 자동으로 유사도를 분석한 후 새로운 시트에 저장**하는 파이썬 코드입니다.새 창으로 메일 보기 ## 📌 **방법**
1. **엑셀에서 "상호" 데이터 가져오기**
   - 엑셀에서 "취합" 시트의 **E열(상호명)**을 가져옵니다.

2. **딥러닝/머신러닝 기반 유사도 분석**
   - **`SentenceTransformers` (BERT)**를 사용하여 상호명을 임베딩 후 **유사한 상호를 찾음**.
   - **코사인 유사도(Cosine Similarity)**를 이용하여 상호명 간의 유사도를 계산.

3. **유사한 상호를 그룹화하여 새로운 시트에 저장**
   - 일정 유사도 이상인 상호명을 그룹으로 묶어 새로운 시트에 추가.

```python
import win32com.client as win32
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 현재 실행 중인 엑셀 인스턴스 가져오기
try:
    excel = win32.GetActiveObject("Excel.Application")
except Exception as e:
    print("⚠️ 실행 중인 엑셀 파일이 없습니다. 먼저 엑셀 파일을 열어주세요.")
    exit()

# 현재 활성화된 엑셀에서 첫 번째 워크북 가져오기
wb = excel.ActiveWorkbook

# "취합" 시트 가져오기
try:
    ws_summary = wb.Sheets("취합")
except Exception as e:
    print("⚠️ '취합' 시트가 없습니다.")
    exit()

# "취합" 시트에서 "상호" (E열) 데이터 가져오기
last_row = ws_summary.Cells(ws_summary.Rows.Count, 5).End(-4162).Row  # E열 마지막 행 찾기
business_names = [ws_summary.Cells(row, 5).Value for row in range(2, last_row + 1) if ws_summary.Cells(row, 5).Value]

# 데이터프레임 생성
df = pd.DataFrame(business_names, columns=["상호"])

# 머신러닝 기반 유사도 분석 모델 (BERT 활용)
model = SentenceTransformer("paraphrase-MiniLM-L6-v2")

# 상호명을 임베딩 벡터로 변환
embeddings = model.encode(df["상호"].tolist(), convert_to_tensor=True)

# 코사인 유사도 계산
similarity_matrix = cosine_similarity(embeddings)

# 유사한 그룹 찾기 (기준: 0.7 이상 유사한 항목)
threshold = 0.7
grouped_data = []
visited = set()

for i in range(len(df)):
    if i in visited:
        continue
    similar_group = [df.iloc[i]["상호"]]
    visited.add(i)
   
    for j in range(i + 1, len(df)):
        if j in visited:
            continue
        if similarity_matrix[i][j] > threshold:
            similar_group.append(df.iloc[j]["상호"])
            visited.add(j)
   
    grouped_data.append(similar_group)

# 새로운 시트 생성
new_sheet_name = "유사한 상호"
try:
    ws_new = wb.Sheets(new_sheet_name)
except:
    ws_new = wb.Sheets.Add()
    ws_new.Name = new_sheet_name

# 기존 데이터 삭제
ws_new.Cells.Clear()

# 유사한 그룹을 엑셀에 저장
ws_new.Cells(1, 1).Value = "그룹 번호"
ws_new.Cells(1, 2).Value = "유사한 상호명"

row_num = 2
for group_idx, group in enumerate(grouped_data, start=1):
    ws_new.Cells(row_num, 1).Value = f"그룹 {group_idx}"
    for name in group:
        ws_new.Cells(row_num, 2).Value = name
        row_num += 1

print("✅ 유사한 상호명이 새로운 시트에 저장되었습니다!")
```

```python
import win32com.client as win32
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import time

# 🚀 실행 시간 측정 시작
start_time = time.time()

# ✅ 현재 실행 중인 엑셀 인스턴스 가져오기
try:
    excel = win32.GetActiveObject("Excel.Application")
    excel.ScreenUpdating = False  # UI 업데이트 비활성화 (속도 향상)
    excel.DisplayAlerts = False   # 경고 비활성화
except Exception as e:
    print("⚠️ 실행 중인 엑셀 파일이 없습니다. 먼저 엑셀 파일을 열어주세요.")
    exit()

# ✅ 현재 활성화된 엑셀에서 첫 번째 워크북 가져오기
wb = excel.ActiveWorkbook

# ✅ "취합" 시트 가져오기
try:
    ws_summary = wb.Sheets("취합")
except Exception as e:
    print("⚠️ '취합' 시트가 없습니다.")
    exit()

# ✅ "취합" 시트에서 "상호" (E열) 데이터 가져오기
last_row = ws_summary.Cells(ws_summary.Rows.Count, 5).End(-4162).Row  # E열 마지막 행 찾기
business_names = [ws_summary.Cells(row, 5).Value for row in range(2, last_row + 1) if ws_summary.Cells(row, 5).Value]

# ✅ 데이터프레임 생성
df = pd.DataFrame(business_names, columns=["상호"])

# ✅ 머신러닝 기반 유사도 분석 모델 (빠른 모델 사용)
model = SentenceTransformer("all-MiniLM-L6-v2", device="cpu")

# ✅ 상호명을 벡터 변환 (배치 처리하여 속도 최적화)
embeddings = model.encode(df["상호"].tolist(), convert_to_numpy=True, batch_size=32)

# ✅ 코사인 유사도 계산 (NumPy 배열 최적화)
similarity_matrix = cosine_similarity(embeddings)

# ✅ 유사한 그룹 찾기 (기준: 0.75 이상 유사한 항목만 그룹화)
threshold = 0.75
grouped_data = []
visited = set()

for i in range(len(df)):
    if i in visited:
        continue
    similar_group = [df.iloc[i]["상호"]]
    visited.add(i)
   
    for j in range(i + 1, len(df)):
        if j in visited:
            continue
        if similarity_matrix[i][j] > threshold:
            similar_group.append(df.iloc[j]["상호"])
            visited.add(j)
   
    grouped_data.append(similar_group)

# ✅ 새로운 시트 생성 (없으면 추가)
new_sheet_name = "유사한 상호"
try:
    ws_new = wb.Sheets(new_sheet_name)
except:
    ws_new = wb.Sheets.Add()
    ws_new.Name = new_sheet_name

# ✅ 기존 데이터 삭제
ws_new.Cells.Clear()

# ✅ 유사한 그룹을 엑셀에 저장 (한 번에 저장)
data_to_write = [("그룹 번호", "유사한 상호명")]

row_num = 2
for group_idx, group in enumerate(grouped_data, start=1):
    for name in group:
        data_to_write.append((f"그룹 {group_idx}", name))

# ✅ Excel에 한 번에 데이터 쓰기 (셀 단위 반복 최소화)
for row_idx, (group, name) in enumerate(data_to_write, start=1):
    ws_new.Cells(row_idx, 1).Value = group
    ws_new.Cells(row_idx, 2).Value = name

# ✅ Excel 작업 안정화
excel.ScreenUpdating = True
excel.DisplayAlerts = True

# 🚀 실행 시간 출력
print(f"✅ 유사한 상호명이 새로운 시트에 저장되었습니다! 실행 시간: {time.time() - start_time:.2f}초")
```
