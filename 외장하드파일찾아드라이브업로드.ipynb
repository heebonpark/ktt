```python
import os
import pandas as pd
from datetime import datetime

# 검색할 디렉토리의 경로 설정
search_directory = '/mnt/ext_drive/Extreme SSD/phb_baeck/2020년 노트북백업/2020년 리텐션팀/해지/일일영업실적'

# 폴더 및 파일 정보를 가져오는 함수
def list_files_and_folders(search_directory):
    files_info = []
    for root, dirs, files in os.walk(search_directory):
        for name in files:
            file_path = os.path.join(root, name)
            file_size = os.path.getsize(file_path)
            modification_time = os.path.getmtime(file_path)
            human_readable_time = datetime.fromtimestamp(modification_time).strftime('%Y-%m-%d %H:%M:%S')

            files_info.append({
                "파일명": name,
                "경로": file_path,
                "파일 크기 (bytes)": file_size,
                "수정 날짜": human_readable_time
            })

    return files_info

# 파일 및 폴더 정보 가져오기
files_info = list_files_and_folders(search_directory)

# DataFrame으로 정리하여 보기
df_files_info = pd.DataFrame(files_info)

# 파일 정보 출력
if not df_files_info.empty:
    print("폴더 및 파일 정보:")
    print(df_files_info)
else:
    print("해당 디렉토리에 파일이 없습니다.")
```

```python
import os

# 외장하드의 폴더 및 파일 구조를 나열하는 함수
def list_directory_structure(search_directory):
    directory_structure = {}

    for root, dirs, files in os.walk(search_directory):
        # 상대 경로 계산
        relative_path = os.path.relpath(root, search_directory)

        # 최상위 폴더의 경우
        if relative_path == ".":
            directory_structure["Root"] = {"folders": dirs, "files": files}
        else:
            # 각 폴더별 구조 추가
            directory_structure[relative_path] = {"folders": dirs, "files": files}

    return directory_structure

# 디렉토리 구조 출력 함수
def print_directory_structure(directory_structure):
    for path, content in directory_structure.items():
        print(f"경로: {path}")
        print("  폴더들:")
        for folder in content["folders"]:
            print(f"    - {folder}")
        print("  파일들:")
        for file in content["files"]:
            print(f"    - {file}")
        print("-" * 50)

# 사용자가 선택한 파일 경로를 가져오는 함수
def get_full_file_path(directory_structure, search_directory, selected_file_name):
    for path, content in directory_structure.items():
        if selected_file_name in content['files']:
            return os.path.join(search_directory, path, selected_file_name)
    return None

# 사용자가 선택한 파일 열기 함수
def open_selected_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            print(content)
    except Exception as e:
        print(f"파일을 여는 중 오류가 발생다: {e}")

# 검색할 디렉토리의 경로 설정
search_directory = '/mnt/ext_drive/Extreme SSD/phb_baeck/2020년 노트북백업/2020년 리텐션팀/해지/일일영업실적'

# 디렉토리 구조 가져오기
directory_structure = list_directory_structure(search_directory)

# 디렉토리 구조 출력
print_directory_structure(directory_structure)

# 파일명 입력
selected_file_name = input("열고자 하는 파일의 이름을 입력하세요: ")

# 파일의 전체 경로 가져오기
full_file_path = get_full_file_path(directory_structure, search_directory, selected_file_name)

if full_file_path:
    # 선택한 파일 열기
    open_selected_file(full_file_path)
else:
    print(f"'{selected_file_name}' 파일을 찾을 수 없습니다.")
```

```python
import os
import ipywidgets as widgets
from IPython.display import display

# 검색할 디렉토리의 경로 설정
search_directory = '/mnt/ext_drive/Extreme SSD/phb_baeck'

# 디렉토리 구조를 탐색하여 트리 형태로 표현
def get_directory_structure(rootdir):
    dir_structure = {}
    for root, dirs, files in os.walk(rootdir):
        # 현재 디렉토리의 상대 경로
        relative_path = os.path.relpath(root, rootdir)
        if relative_path == ".":
            relative_path = "Root"
        dir_structure[relative_path] = {"folders": dirs, "files": files}
    return dir_structure

# 디렉토리와 파일을 선택하는 위젯 생성
def create_file_selector(directory_structure):
    folder_selector = widgets.Dropdown(options=["Root"] + list(directory_structure.keys()), description='폴더:')
    file_selector = widgets.Dropdown(description='파일:')

    def update_file_options(change):
        selected_folder = change['new']
        file_selector.options = directory_structure[selected_folder]['files']

    folder_selector.observe(update_file_options, names='value')

    return folder_selector, file_selector

# 파일을 열고 내용을 출력하는 함수
def open_selected_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            print(content)
    except Exception as e:
        print(f"파일을 여는 중 오류가 발생했습니다: {e}")

# 디렉토리 구조 가져오기
directory_structure = get_directory_structure(search_directory)

# 파일 선택 위젯 생성
folder_selector, file_selector = create_file_selector(directory_structure)

# 파일 열기 버튼 생성 및 동작 정의
open_button = widgets.Button(description="파일 열기")
def on_open_button_click(b):
    selected_folder = folder_selector.value
    selected_file = file_selector.value
    if selected_file:
        full_file_path = os.path.join(search_directory, selected_folder, selected_file)
        open_selected_file(full_file_path)
    else:
        print("파일이 선택되지 않았습니다.")

open_button.on_click(on_open_button_click)

# 인터페이스 표시
display(folder_selector)
display(file_selector)
display(open_button)
```

```python
import os

# 특정 디렉토리의 파일 및 폴더 정보를 나열하는 함수
def list_files_and_folders(search_directory):
    file_list = []
    for root, dirs, files in os.walk(search_directory):
        relative_path = os.path.relpath(root, search_directory)
        print(f"폴더: {relative_path}")

        if dirs:
            print("  하위 폴더:")
            for dir_name in dirs:
                print(f"    - {dir_name}")

        if files:
            print("  파일:")
            for file_name in files:
                print(f"    - {file_name}")
                file_list.append(os.path.join(root, file_name))
        print("-" * 50)

    return file_list

# 파일을 여는 함수
def open_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            print(content)
    except Exception as e:
        print(f"파일을 여는 중 오류가 발생했습니다: {e}")

# 검색할 디렉토리 경로 설정
search_directory = '/mnt/ext_drive/Extreme SSD/phb_baeck/2020년 노트북백업/2020년 리텐션팀/해지/일일영업실적'

# 디렉토리의 파일 및 폴더 목록을 나열
file_list = list_files_and_folders(search_directory)

# 사용자가 파일명 일부를 입력
partial_file_name = input("열고자 하는 파일명 일부를 입력하세요: ")

# 입력한 이름이 포함된 파일 검색 및 열기
matched_files = [file for file in file_list if partial_file_name in os.path.basename(file)]

if matched_files:
    if len(matched_files) == 1:
        print(f"파일을 엽니다: {matched_files[0]}")
        open_file(matched_files[0])
    else:
        print("여러 개의 파일이 일치합니다. 파일을 선택하세요:")
        for idx, file in enumerate(matched_files):
            print(f"{idx + 1}: {file}")

        file_index = int(input("파일 번호를 입력하세요: ")) - 1

        if 0 <= file_index < len(matched_files):
            print(f"파일을 엽니다: {matched_files[file_index]}")
            open_file(matched_files[file_index])
        else:
            print("잘못된 번호입니다.")
else:
    print("일치하는 파일을 찾을 수 없습니다.")
```

```python
import os

# 특정 디렉토리의 하위 폴더 및 파일을 나열하는 함수
def list_folders_and_files(directory):
    try:
        contents = os.listdir(directory)
        folders = [f for f in contents if os.path.isdir(os.path.join(directory, f))]
        files = [f for f in contents if os.path.isfile(os.path.join(directory, f))]
        return folders, files
    except Exception as e:
        print(f"디렉토리를 열 수 없습니다: {e}")
        return [], []

# 선택된 파일을 여는 함수
def open_selected_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            print(content)
    except Exception as e:
        print(f"파일을 여는 중 오류가 발생했습니다: {e}")

# 시작할 디렉토리 설정
current_directory = '/mnt/ext_drive/Extreme SSD/phb_baeck/2020년 노트북백업/2020년 리텐션팀/해지/일일영업실적'

while True:
    # 현재 디렉토리의 폴더 및 파일 목록 가져오기
    folders, files = list_folders_and_files(current_directory)

    # 폴더 및 파일 목록 출력
    print(f"\n현재 경로: {current_directory}")
    print("폴더 목록:")
    for i, folder in enumerate(folders):
        print(f"{i + 1}. {folder}")
    print("파일 목록:")
    for i, file in enumerate(files):
        print(f"{i + 1 + len(folders)}. {file}")

    # 폴더 또는 파일 선택
    choice = input("폴더를 선택하려면 번호를 입력하고, 파일을 선택하려면 'f<번호>' 형식으로 입력하세요 (종료: q): ")

    if choice.lower() == 'q':
        break
    elif choice.isdigit() and 1 <= int(choice) <= len(folders):
        # 폴더 선택 시, 그 폴더로 이동
        selected_folder = folders[int(choice) - 1]
        current_directory = os.path.join(current_directory, selected_folder)
    elif choice.startswith('f') and choice[1:].isdigit():
        file_index = int(choice[1:]) - 1 - len(folders)
        if 0 <= file_index < len(files):
            selected_file = files[file_index]
            file_path = os.path.join(current_directory, selected_file)
            print(f"\n파일을 엽니다: {file_path}")
            open_selected_file(file_path)
        else:
            print("잘못된 파일 번호입니다.")
    else:
        print("잘못된 입력입니다. 다시 시도하세요.")
```
