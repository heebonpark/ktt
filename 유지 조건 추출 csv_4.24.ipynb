```python
import pandas as pd
from openpyxl import Workbook
import os

# 파일 경로
csv_file_path = 'D:/시설/유지/24.3월유지가입자.csv'
xlsx_file_path = csv_file_path.replace('.csv', '.xlsx')

def load_csv_file():
    if not os.path.exists(csv_file_path):
        print("CSV 파일이 존재하지 않습니다. 경로를 확인해주세요.")
        return None
    else:
        print("CSV 파일 로드 중...")
        data = pd.read_csv(csv_file_path, encoding='cp949', sep=',', low_memory=False)
        print("CSV 파일 로드 성공")
        return data

def save_to_excel(data):
    print("Excel 파일 저장 중...")
    data.to_excel(xlsx_file_path, index=False, engine='openpyxl')
    print("Excel 파일 저장 성공")

def read_excel_file():
    print("Excel 파일 로드 중...")
    data_xlsx = pd.read_excel(xlsx_file_path, engine='openpyxl')
    print("Excel 파일 로드 성공")
    return data_xlsx

def get_user_input_and_filter(data_xlsx):
    print("사용자로부터 필터링 조건을 입력받습니다.")
    search_criteria = {
        'b': input("관리본부명을 입력하세요: "),
        'd': input(중앙지사),
        'g': input(51869050),
        'n': input("상호를 입력하세요: "),
        'fh': input("KTT월정료(조정) 금액을 입력하세요: ")
    }
    filtered_data = data_xlsx[
        (data_xlsx['b'].str.contains(search_criteria['b'], na=False)) &
        (data_xlsx['d'].str.contains(search_criteria['d'], na=False)) &
        (data_xlsx['g'].str.contains(search_criteria['g'], na=False)) &
        (data_xlsx['n'].str.contains(search_criteria['n'], na=False)) &
        (data_xlsx['fh'].astype(str).str.contains(search_criteria['fh'], na=False))
    ]
    return filtered_data

def preview_and_save(filtered_data):
    if filtered_data.empty:
        print("조회된 데이터가 없습니다.")
    else:
        print(f"조회된 행의 개수: {len(filtered_data)}")
        print("다음은 검색된 데이터의 첫 10개 항목입니다:")
        print(filtered_data.head(10))
        execute_save = input("검색된 결과를 Excel 파일로 저장하시겠습니까? (y/n): ")
        if execute_save.lower() == 'y':
            filtered_data.to_excel('filtered_data.xlsx', index=False)
            print("데이터가 'filtered_data.xlsx' 파일로 저장되었습니다.")
        else:
            print("데이터 저장이 취소되었습니다.")

# 메인 실행 루틴
data = load_csv_file()
if data is not None:
    save_to_excel(data)
    data_xlsx = read_excel_file()
    filtered_data = get_user_input_and_filter(data_xlsx)
    preview_and_save(filtered_data)
```

```python
import pandas as pd
import os

# 파일 경로
csv_file_path = 'D:/시설/유지/24.3월유지가입자.csv'

def load_and_filter_data(csv_file_path, contract_number):
    """CSV 파일을 로드하고 특정 계약번호에 따라 데이터를 필터링합니다."""
    if not os.path.exists(csv_file_path):
        print("CSV 파일이 존재하지 않습니다. 경로를 확인해주세요.")
        return None
    else:
        print("CSV 파일 로드 중...")
        data = pd.read_csv(csv_file_path, encoding='cp949', low_memory=False)
        print("CSV 파일 로드 성공")

        # 계약번호로 필터링
        print(f"계약번호 {contract_number}에 대한 데이터 필터링 중...")
        filtered_data = data.query('g == @contract_number')
       
        return filtered_data

def save_filtered_data_to_excel(filtered_data, output_file_path):
    """필터링된 데이터를 Excel 파일로 저장합니다."""
    print("Excel 파일 저장 중...")
    filtered_data.to_excel(output_file_path, index=False, engine='openpyxl')
    print("Excel 파일 저장 성공")

def print_filtered_data(filtered_data):
    """필터링된 데이터를 출력합니다."""
    if filtered_data.empty:
        print("조회된 데이터가 없습니다.")
    else:
        print("조회된 데이터:")
        print(filtered_data)

# 메인 실행 루틴
contract_number = "51869050"
filtered_data = load_and_filter_data(csv_file_path, contract_number)
if filtered_data is not None:
    print_filtered_data(filtered_data)
    # 필요한 경우만 Excel 파일로 저장
    save_filtered_data_to_excel(filtered_data, 'filtered_data.xlsx')
```
