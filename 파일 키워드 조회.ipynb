```python
import os
import glob
import subprocess

# 검색할 디렉토리 (D 드라이브)
search_directory = 'D:/'

# 검색할 키워드
search_keyword = 'your_keyword_here'  # 이 부분에 찾고 싶은 키워드를 입력하세요.

# 확장자별로 파일 검색
file_types = ['*.pptx', '*.pdf', '*.xlsx']

# 파일 검색 함수
def search_files(directory, keyword, file_types):
    found_files = []
   
    for file_type in file_types:
        # 파일 경로 패턴 설정 (D 드라이브 아래 모든 하위 폴더에서 검색)
        search_pattern = os.path.join(directory, '**', file_type)
       
        # 해당 확장자의 파일 목록 가져오기
        for file_path in glob.iglob(search_pattern, recursive=True):
            # 파일 이름에 키워드가 포함되었는지 확인
            if keyword.lower() in os.path.basename(file_path).lower():
                found_files.append(file_path)
   
    return found_files

# 파일 목록 출력 함수
def display_files(files):
    if files:
        print("Found files:")
        for idx, file in enumerate(files, start=1):
            print(f"{idx}. {file}")
    else:
        print("No files found with the given keyword.")

# 파일 열기 함수
def open_file(file_path):
    try:
        subprocess.run(['start', '', file_path], shell=True)
    except Exception as e:
        print(f"Error opening file: {file_path} - {e}")

# 메인 실행 함수
def main():
    files = search_files(search_directory, search_keyword, file_types)
    display_files(files)
   
    if files:
        # 사용자로부터 파일 번호 선택
        try:
            file_number = int(input("Enter the number of the file to open: "))
            if 1 <= file_number <= len(files):
                selected_file = files[file_number - 1]
                print(f"Opening file: {selected_file}")
                open_file(selected_file)
            else:
                print("Invalid number. Please run the program again and choose a valid number.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    else:
        print("No files to open.")

# 실행
main()
```

## 파일용량리스트추출

```python
import os

# D 드라이브 경로
search_directory = 'D:/'

# 파일 크기 정보 수집 및 리스트화
def get_files_with_size(directory):
    files_with_size = []
   
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                file_size = os.path.getsize(file_path)  # 파일 크기(바이트)
                files_with_size.append((file_path, file_size))
            except Exception as e:
                print(f"Error getting size for {file_path}: {e}")
   
    # 파일 크기에 따라 내림차순으로 정렬
    files_with_size.sort(key=lambda x: x[1], reverse=True)
   
    return files_with_size

# 파일 크기 정보 출력
def display_files(files_with_size, top_n=10):
    print(f"Top {top_n} largest files:")
    for i, (file_path, file_size) in enumerate(files_with_size[:top_n], start=1):
        print(f"{i}. {file_path} - {file_size / (1024 * 1024):.2f} MB")

# 실행
files_with_size = get_files_with_size(search_directory)
display_files(files_with_size, top_n=20)  # 상위 20개 파일 출력
```
