```python
import pandas as pd

def copy_selected_columns_to_new_sheet():
    # 원본 파일 경로
    source_file_path = "D:/시설/24.2월유지가입자.xlsx"
     
    # 필요한 열 목록
    selected_columns = ["B", "D", "E", "F", "CD", "G", "I", "V", "AA", "FH", "FI", "DQ", "AO", "AP", "AU", "AW", "DU", "EB", "EC", "EM", "ES"]
   
    # 새로운 시트의 이름
    target_sheet_name = "Sheet1"
   
    # 원본 파일 읽기
    df = pd.read_excel(source_file_path, engine='openpyxl') 

    # 필요한 열 선택
    selected_data = df[selected_columns]
   
    # 새로운 워크북 생성
    target_workbook = pd.ExcelWriter("D:/시설/24.2월유지가입자 조건요약_파이썬.xlsx", engine='xlsxwriter')
   
    # 선택한 데이터를 새 시트로 복사
    selected_data.to_excel(target_workbook, sheet_name=target_sheet_name, index=False)
   
    # 워크북 저장 및 닫기
    target_workbook.save()
    target_workbook.close()

    print("특정 열의 데이터를 새로운 시트 파일로 가져왔습니다.")

# 함수 호출
copy_selected_columns_to_new_sheet()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np

questions = ['직무 만족도', '도구 및 자원 충분성', '훈련 및 교육 충분성', '보상 체계 효과성', '팀 내 협력 및 지원', '영업 기술 및 지식 충분성']
scores = np.array([[4, 2, 2], [3, 3, 2], [4, 2, 2], [3, 3, 3], [3, 2, 2], [4, 2, 2]])

# 데이터에 따른 컬러 매핑 설정
norm = mcolors.Normalize(vmin=np.min(scores), vmax=np.max(scores))  # 데이터의 최솟값과 최댓값을 기준으로 정규화
color_map = cm.Reds  # Reds 컬러맵 사용

num_questions = len(questions)
num_scores_per_question = scores.shape[1]  # 한 질문당 점수의 개수
total_width, n = 0.8, num_scores_per_question  # 전체 막대 너비와 한 그룹당 막대 개수
width = total_width / n  # 각 막대의 너비

plt.figure(figsize=(10, 6))

# 각 점수에 대한 막대 그리기
for i in range(num_scores_per_question):
    inds = np.arange(num_questions)  # 질문별 x 위치
    plt.bar(inds + i*width, scores[:, i], width=width,
            color=[color_map(norm(score)) for score in scores[:, i]])

plt.xlabel('질문 항목')
plt.ylabel('점수')
plt.title('영업사원 설문조사 결과')
plt.xticks(inds + width / 2, questions, rotation=45)
plt.ylim(0, np.max(scores) + 1)
plt.colorbar(cm.ScalarMappable(norm=norm, cmap=color_map), label='점수에 따른 색상')  # 컬러바 추가
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

questions = ['직무 만족도', '도구 및 자원 충분성', '훈련 및 교육 충분성', '보상 체계 효과성', '팀 내 협력 및 지원', '영업 기술 및 지식 충분성']
scores = np.array([[4, 2, 2], [3, 3, 2], [4, 2, 2], [3, 3, 3], [3, 2, 2], [4, 2, 2]])

# 각 질문에 대한 점수의 평균과 표준편차 계산
means = np.mean(scores, axis=1)
stddevs = np.std(scores, axis=1)

# 평균 점수를 바탕으로 한 막대 그래프
x_pos = np.arange(len(questions))
plt.bar(x_pos, means, yerr=stddevs, capsize=5, color='skyblue', alpha=0.7)
plt.xlabel('질문 항목')
plt.ylabel('평균 점수')
plt.title('영업사원 설문조사 결과: 평균 및 표준편차')
plt.xticks(x_pos, questions, rotation=45)
plt.tight_layout()  # 그래프 레이아웃 조정

# 평균 점수 위에 표준편차를 표시하는 오차 막대 추가
plt.errorbar(x_pos, means, yerr=stddevs, fmt='o', color='r', alpha=0.5)

plt.show()
```

```python
import numpy as np
import pandas as pd  # Pandas 라이브러리 추가
import seaborn as sns
import matplotlib.pyplot as plt

# 폰트 설정
plt.rc('font', family='Malgun Gothic')

# 데이터 설정
questions = ['직무 만족도', '도구 및 자원\n충분성', '훈련 및 교육\n충분성', '보상 체계\n효과성', '팀 내 협력 및\n지원', '영업 기술 및\n지식 충분성']
scores = np.array([[4, 2, 2], [3, 3, 2], [4, 2, 2], [3, 3, 3], [3, 2, 2], [4, 2, 2]])

# 각 질문에 대한 평균과 표준편차 계산 (이미 계산된 것을 활용)
means = np.mean(scores, axis=1)
stddevs = np.std(scores, axis=1)

# Pandas DataFrame으로 변환
df = pd.DataFrame({'질문 항목': questions, '평균 점수': means, '표준편차': stddevs})

# Seaborn을 사용한 그래프 생성
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='평균 점수', y='질문 항목', data=df, xerr=stddevs, capsize=0.2, palette="Blues_d", orient='h')

# 막대에 값 라벨 추가
for index, row in df.iterrows():
    barplot.text(row['평균 점수'] + 0.1, index, f'{row["평균 점수"]:.2f}', color='black', va="center")

plt.xlabel('평균 점수', fontsize=14)
plt.ylabel('질문 항목', fontsize=14)
plt.title('영업사원 설문조사 결과 (Seaborn)', fontsize=16)
plt.xlim(0, 5)  # x축 범위 설정

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def score_to_color(score):
    # 최대 점수와 색상 범위 설정
    max_score = 100
    R_start, G_start, B_start = 255, 0, 0 # 점수가 낮을 때 색상 (빨간색)
    R_end, G_end, B_end = 204, 204, 204 # 점수가 높을 때 색상 (연한 빨간색)
    
    # 선형 보간법을 사용하여 RGB 값을 계산
    R = R_start + (R_end - R_start) * (score / max_score)
    G = G_start + (G_end - G_start) * (score / max_score)
    B = B_start + (B_end - B_start) * (score / max_score)
    
    # matplotlib에서 사용할 수 있는 형태로 변환 (0에서 1 사이의 값)
    return (R / 255, G / 255, B / 255)

# 점수 예시 (0부터 100까지 10점 간격)
scores = range(0, 101, 10)

# 점수에 해당하는 색상 계산
colors = [score_to_color(score) for score in scores]

# 색상 표시
plt.figure(figsize=(10, 4))
plt.bar(scores, [1]*len(scores), color=colors, edgecolor='none')
plt.xlabel('Score')
plt.yticks([])  # Y축 눈금 제거
plt.title('Score to Color Gradient')
plt.show()
```
