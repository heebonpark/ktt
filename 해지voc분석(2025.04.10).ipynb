```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
from wordcloud import WordCloud

# 한글 폰트 설정 (Colab용 나눔폰트 또는 맑은고딕)
!apt-get -qq install fonts-nanum
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# ▶️ 엑셀 파일 불러오기 (이미 불러온 경우 생략 가능)
file_path = '/content/drive/MyDrive/시설/해지/VOC/VOC정보조회_결과2.xlsx'
df = pd.read_excel(file_path, header=1)

# ▶️ 날짜 컬럼 정리 (예시: 계약종료일)
df['계약종료일'] = pd.to_datetime(df['계약종료일'], errors='coerce')

# ▶️ 가장 최근 VOC만 남기기 (계약번호 기준)
df = df.sort_values('계약종료일').drop_duplicates('계약번호', keep='last')

# ▶️ 주요 컬럼 EDA 시각화
eda_columns = ['관리지사', '영업구역정보', '서비스상태(중)', 'VOC유형']

for col in eda_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f'{col} 분포')
    plt.tight_layout()
    plt.show()

# ▶️ 등록내용 텍스트 분석
def extract_words(text, length):
    return re.findall(rf'\b\w{{{length}}}\b', text)

all_text = ' '.join(df['등록내용'].dropna().astype(str))
words_4 = extract_words(all_text, 4)
words_5 = extract_words(all_text, 5)

common_words_4 = Counter(words_4).most_common(10)
common_words_5 = Counter(words_5).most_common(10)

print("가장 많이 언급되는 4글자 단어:")
for word, freq in common_words_4:
    print(f'{word}: {freq}')

print("\n가장 많이 언급되는 5글자 단어:")
for word, freq in common_words_5:
    print(f'{word}: {freq}')

# ▶️ 워드클라우드 생성
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path='/usr/share/fonts/truetype/nanum/NanumGothic.ttf').generate_from_frequencies(dict(common_words_4 + common_words_5))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('가장 많이 언급되는 4~5글자 단어 워드클라우드')
plt.show()
```

```python
# 1. 기본 모듈 및 폰트 설정
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import re
from collections import Counter
import matplotlib.font_manager as fm

# Colab 한글 폰트 설정
!apt-get -qq install -y fonts-nanum > /dev/null
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# 2. Google 스프레드시트 연결
sheet_id = '1T63oteDqGdZS1d6v29UHYVlFm4YEoDzGNnauo-nTosY'
sheet_name = 'Sheet1'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'

# 데이터 불러오기
df = pd.read_csv(url)

# 3. 텍스트 컬럼 병합
text_columns = ['등록내용', '처리내용', '클레임구분']
merged_text = df[text_columns].fillna('').agg(' '.join, axis=1)

# 4. 한글 단어 추출 함수
def extract_korean_words(text):
    return re.findall(r'[가-힣]{2,}', text)

words = []
for line in merged_text:
    words.extend(extract_korean_words(line))

word_freq = Counter(words)
common_words = word_freq.most_common(100)

# 5. 워드클라우드 생성
wordcloud = WordCloud(
    font_path='/usr/share/fonts/truetype/nanum/NanumGothic.ttf',
    width=1000, height=500, background_color='white'
).generate_from_frequencies(dict(common_words))

plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('VOC 등록/처리/클레임 내용 기반 주요 키워드 워드클라우드')
plt.show()

# 6. 클레임구분 빈도 시각화
if '클레임구분' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y='클레임구분', order=df['클레임구분'].value_counts().index)
    plt.title('클레임 유형 분포')
    plt.tight_layout()
    plt.show()
```
