```python
# 라이브러리 설치 (자동으로 진행)
!pip install pandas fuzzywuzzy python-Levenshtein chardet openpyxl

import pandas as pd
from fuzzywuzzy import fuzz, process
import chardet

# 구글 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

# 파일 경로 설정
zip_file_path = '/content/drive/My Drive/회사업무/1.성과분석/'
file1_path = zip_file_path + '개시및청약.csv'
file2_path = zip_file_path + '검증매핑.csv'

# 인코딩 자동 감지 함수
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

# CSV 파일 읽기 (인코딩 자동 감지)
encoding1 = detect_encoding(file1_path)
encoding2 = detect_encoding(file2_path)

data1 = pd.read_csv(file1_path, encoding=encoding1)
data2 = pd.read_csv(file2_path, encoding=encoding2)

# 헤더명 설정
header1 = '설치주소'
header2 = '소재지전체주소'
header3 = '도로명전체주소'

# 주소를 시, 군, 구, 동 등으로 분리하는 함수 (번지와 상호명 제거)
def split_address(address):
    if pd.isnull(address):
        return None
    parts = address.split()
    # 번지나 상호명을 포함할 수 있는 마지막 부분을 제거 (정교한 방법은 필요시 조정)
    if len(parts) > 2:
        return parts[:-1]  # 마지막 부분을 제외하고 반환
    return parts

# 세부적으로 주소를 비교하는 함수
def compare_addresses(addr1, addr2):
    if addr1 is None or addr2 is None:
        return 0
    match_count = 0
    total_count = min(len(addr1), len(addr2))
    for part1, part2 in zip(addr1, addr2):
        if fuzz.token_set_ratio(part1, part2) > 60:
            match_count += 1
    return (match_count / total_count) * 100

# 유사도 분석 및 매핑 함수
def find_best_match(row, data1):
    addr2_full = str(row[header2]) + " " + str(row[header3])
    addr2_split = split_address(addr2_full)

    best_match = None
    best_score = 0

    for _, row1 in data1.iterrows():
        addr1_split = split_address(row1[header1])
        score = compare_addresses(addr1_split, addr2_split)
        if score > 60 and score > best_score:
            best_match = row1[header1]
            best_score = score

    return best_match

# 검증대상 파일의 각 행에 대해 유사한 주소 찾기
data2['매핑된주소'] = data2.apply(lambda row: find_best_match(row, data1), axis=1)

# 매핑된 주소를 설치주소와 비교하여 일치할 경우, 값을 적용 (매핑된 결과를 D열부터 적용)
for idx, row in data2.iterrows():
    match = row['매핑된주소']
    if pd.notnull(match):
        for col in data2.columns[3:]:  # D열부터 끝까지
            data2.at[idx, col] = data1[data1[header1] == match][col].values[0]

# 결과 확인
print(data2[['소재지전체주소', '도로명전체주소', '설치주소(매핑)']])

# 매핑된 결과를 새로운 Excel 파일로 저장
result_file_path_csv = zip_file_path + '검증대상_매핑결과.csv'
result_file_path_excel = zip_file_path + '검증대상_매핑결과.xlsx'
data2.to_csv(result_file_path_csv, index=False, encoding='utf-8')
data2.to_excel(result_file_path_excel, index=False, encoding='utf-8')
print(f'매핑된 결과가 {result_file_path_csv} 및 {result_file_path_excel} 파일로 저장되었습니다.')
```
