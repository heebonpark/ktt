```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
from ipywidgets import Layout, GridspecLayout
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'D:\\명\\폴더의 명0717.cs'  # 파일 경로와 확장자를 확인하고 수정하세요.
data = pd.read_csv(file_path)

# Set matplotlib to display Korean characters
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우
plt.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용 시, 마이너스 기호 깨짐 방지

# EDA Functions
def plot_histogram(column, filtered_data):
    plt.figure(figsize=(10, 6))
    sns.histplot(filtered_data[column], kde=True)
    plt.title(f'{column} 분포')
    plt.xlabel(column)
    plt.ylabel('빈도수')
    plt.show()

def plot_bar(column, filtered_data):
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(y=column, data=filtered_data, order=filtered_data[column].value_counts().index)
    plt.title(f'{column} 분포')
    plt.xlabel('빈도수')
    plt.ylabel(column)
    for p in ax.patches:
        width = p.get_width()
        plt.text(width + 1, p.get_y() + p.get_height() / 2, int(width), ha='center', va='center')
    plt.show()

def plot_correlation(filtered_data):
    plt.figure(figsize=(12, 10))
    correlation_matrix = filtered_data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('상관관계 히트맵')
    plt.show()

def plot_pairplot(filtered_data):
    sns.pairplot(filtered_data)
    plt.title('페어플롯')
    plt.show()

def plot_boxplot(column, filtered_data):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=column, data=filtered_data)
    plt.title(f'{column} 박스플롯')
    plt.xlabel(column)
    plt.show()

def plot_violinplot(column, filtered_data):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=column, data=filtered_data)
    plt.title(f'{column} 바이올린 플롯')
    plt.xlabel(column)
    plt.show()

# Create input widgets for EDA
eda_options = ['분포도 (히스토그램)', '분포도 (막대그래프)', '상관관계 히트맵', '페어플롯', '박스플롯', '바이올린 플롯']
eda_select = widgets.Dropdown(options=eda_options, description='EDA 선택:', layout=Layout(width='300px'))

columns_options = data.columns.tolist()
columns_dropdown = widgets.Dropdown(options=columns_options, description='열 선택:', layout=Layout(width='300px'))

eda_button = widgets.Button(description='시각화 실행', layout=Layout(width='100px'))

# Create input widgets for filtering
name_input = widgets.Text(placeholder='성명 입력 (콤마로 구분)', description='성명:', layout=Layout(width='300px'))
department_input = widgets.Text(placeholder='부서 입력 (키워드 포함)', description='부서:', layout=Layout(width='300px'))
rank_input = widgets.Text(placeholder='직급 입력 (키워드 포함)', description='직급:', layout=Layout(width='300px'))
address_input = widgets.Text(placeholder='주소 입력 (키워드 포함)', description='주소:', layout=Layout(width='300px'))

# Define the button click event for EDA
def on_eda_button_clicked(b):
    clear_output(wait=True)
    display(ui)  # 유지 위젯 인터페이스
   
    eda_type = eda_select.value
    column = columns_dropdown.value
    names = [name.strip() for name in name_input.value.split(',')]
    department_keyword = department_input.value
    rank_keyword = rank_input.value
    address_keyword = address_input.value
   
    filtered_data = data
    if names:
        filtered_data = filtered_data[filtered_data['성명'].isin(names)]
    if department_keyword:
        filtered_data = filtered_data[filtered_data['부서1'].str.contains(department_keyword, na=False)]
    if rank_keyword:
        filtered_data = filtered_data[filtered_data['직급'].str.contains(rank_keyword, na=False)]
    if address_keyword:
        filtered_data = filtered_data[filtered_data['주소'].str.contains(address_keyword, na=False)]
   
    if eda_type == '분포도 (히스토그램)':
        plot_histogram(column, filtered_data)
    elif eda_type == '분포도 (막대그래프)':
        plot_bar(column, filtered_data)
    elif eda_type == '상관관계 히트맵':
        plot_correlation(filtered_data)
    elif eda_type == '페어플롯':
        plot_pairplot(filtered_data)
    elif eda_type == '박스플롯':
        plot_boxplot(column, filtered_data)
    elif eda_type == '바이올린 플롯':
        plot_violinplot(column, filtered_data)

eda_button.on_click(on_eda_button_clicked)

# Create a button to collect selected columns
collect_button = widgets.Button(description='열 선택', layout=Layout(width='100px'))

def collect_columns(b):
    selected_columns = [cb.description for cb in checkboxes if cb.value]
    search_data(name_input.value, department_input.value, rank_input.value, address_input.value, selected_columns)

collect_button.on_click(collect_columns)

# Create checkboxes for columns in a grid layout
def create_grid(checkboxes, num_columns):
    num_rows = (len(checkboxes) + num_columns - 1) // num_columns
    grid = GridspecLayout(num_rows, num_columns)
    for i, cb in enumerate(checkboxes):
        grid[i // num_columns, i % num_columns] = cb
    return grid

columns_options = data.columns.tolist()
num_columns = 3  # Number of columns in the grid
checkboxes = [widgets.Checkbox(value=False, description=col) for col in columns_options]
all_columns_checkbox = widgets.Checkbox(value=False, description='전체 선택')

grid = create_grid(checkboxes, num_columns)

def update_checkboxes(change):
    for cb in checkboxes:
        cb.value = all_columns_checkbox.value

all_columns_checkbox.observe(update_checkboxes, names='value')

# Visualization functions
def visualize_department_distribution(filtered_data):
    dept_counts = filtered_data['부서1'].value_counts().reset_index()
    dept_counts.columns = ['부서', '인원 수']
    display(dept_counts)

def visualize_rank_distribution(filtered_data):
    rank_counts = filtered_data['직급'].value_counts()
    plt.figure(figsize=(8, 8))
    rank_counts.plot(kind='pie', autopct='%1.1f%%')
    plt.title('직급별 인력 현황')
    plt.ylabel('')
    plt.show()

# Visualization buttons
dept_button = widgets.Button(description='부서별 인력 분포도 보기', layout=Layout(width='200px'))
rank_button = widgets.Button(description='직급별 인력 현황 시각화', layout=Layout(width='200px'))

def on_dept_button_clicked(b):
    clear_output(wait=True)
    display(ui)
    names = [name.strip() for name in name_input.value.split(',')]
    department_keyword = department_input.value
    rank_keyword = rank_input.value
    address_keyword = address_input.value
   
    filtered_data = data
    if names:
        filtered_data = filtered_data[filtered_data['성명'].isin(names)]
    if department_keyword:
        filtered_data = filtered_data[filtered_data['부서1'].str.contains(department_keyword, na=False)]
    if rank_keyword:
        filtered_data = filtered_data[filtered_data['직급'].str.contains(rank_keyword, na=False)]
    if address_keyword:
        filtered_data = filtered_data[filtered_data['주소'].str.contains(address_keyword, na=False)]
   
    visualize_department_distribution(filtered_data)

def on_rank_button_clicked(b):
    clear_output(wait=True)
    display(ui)
    names = [name.strip() for name in name_input.value.split(',')]
    department_keyword = department_input.value
    rank_keyword = rank_input.value
    address_keyword = address_input.value
   
    filtered_data = data
    if names:
        filtered_data = filtered_data[filtered_data['성명'].isin(names)]
    if department_keyword:
        filtered_data = filtered_data[filtered_data['부서1'].str.contains(department_keyword, na=False)]
    if rank_keyword:
        filtered_data = filtered_data[filtered_data['직급'].str.contains(rank_keyword, na=False)]
    if address_keyword:
        filtered_data = filtered_data[filtered_data['주소'].str.contains(address_keyword, na=False)]
   
    visualize_rank_distribution(filtered_data)

dept_button.on_click(on_dept_button_clicked)
rank_button.on_click(on_rank_button_clicked)

# Organize widgets in a box layout
ui = widgets.VBox([
    widgets.HBox([name_input, department_input]),
    widgets.HBox([rank_input, address_input]),
    all_columns_checkbox,
    grid,
    collect_button,
    widgets.HBox([dept_button, rank_button

]),
    widgets.HBox([eda_select, columns_dropdown]),
    eda_button
])

# Display the interface
display(ui)

# Ensure matplotlib displays inline in Jupyter
%matplotlib inline
```
