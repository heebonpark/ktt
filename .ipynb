```python
import gspread
from google.auth.transport.requests import Request
from google.oauth2.service_account import Credentials

# 구글 스프레드시트와 연동을 위한 인증 정보 설정
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]

# 업로드된 credentials.json 파일 경로 확인 및 설정
creds = Credentials.from_service_account_file("/content/credentials.json", scopes=scope)

# Google Sheets와 연결
client = gspread.authorize(creds)

# 새 스프레드시트 생성
spreadsheet_name = "유통망정산"
spreadsheet = client.create(spreadsheet_name)

# 스프레드시트 URL 출력
print(f"스프레드시트가 생성되었습니다: {spreadsheet.url}")
```
