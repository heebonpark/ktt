```python
# ✅ 라이브러리 불러오기
import win32com.client as win32

# ✅ Excel 열려 있는 인스턴스에 연결
excel = win32.GetActiveObject("Excel.Application")
wb = excel.ActiveWorkbook
ws = wb.ActiveSheet

# ✅ 데이터 범위 파악
used_range = ws.UsedRange
row_count = used_range.Rows.Count

# ✅ 셀에서 LV1 ~ 주차수 정보 추출 함수
def extract_column_values(row_index):
    result = {
        'LV1': None, 'LV2': None, 'LV3': None,
        'N1': None, 'N2': None,
        '일자': None, '할인율': None, '주차수': None
    }

    for col in range(7, 18):  # G(7) ~ Q(17)
        val = ws.Cells(row_index, col).Value
        if val is None:
            continue
        val_str = str(val)
        if 'LV1' in val_str:
            result['LV1'] = val_str
        elif 'LV2' in val_str:
            result['LV2'] = val_str
        elif 'LV3' in val_str:
            result['LV3'] = val_str
        elif 'N1' in val_str:
            result['N1'] = val_str
        elif 'N2' in val_str:
            result['N2'] = val_str
        elif '일' in val_str and any(k in val_str for k in ['월', '일']):
            result['일자'] = val_str
        elif '할인' in val_str or '%' in val_str:
            result['할인율'] = val_str
        elif '주차' in val_str or '주' in val_str:
            result['주차수'] = val_str
    return result

# ✅ 결과 적용: W~AD열
for i in range(2, row_count + 1):  # 1행은 헤더로 제외
    info = extract_column_values(i)
    ws.Cells(i, 23).Value = info['LV1']      # W
    ws.Cells(i, 24).Value = info['LV2']      # X
    ws.Cells(i, 25).Value = info['LV3']      # Y
    ws.Cells(i, 26).Value = info['N1']       # Z
    ws.Cells(i, 27).Value = info['N2']       # AA
    ws.Cells(i, 28).Value = info['일자']     # AB
    ws.Cells(i, 29).Value = info['할인율']   # AC
    ws.Cells(i, 30).Value = info['주차수']   # AD

print("✅ W~AD열 데이터 자동 정리 완료!")
```

```python
# ✅ 필수 라이브러리 설치 (한 번만 실행)
# !pip install pywin32

# ✅ 라이브러리 불러오기
import win32com.client as win32
import pandas as pd

# ✅ 엑셀 인스턴스 연결
excel = win32.GetActiveObject("Excel.Application")
wb = excel.ActiveWorkbook
ws = wb.ActiveSheet

# ✅ 사용된 범위 추출
used_range = ws.UsedRange
row_count = used_range.Rows.Count

# ✅ 키워드 기반 추출 함수
def extract_column_values(row_index):
    values = {'LV1': None, 'LV2': None, 'LV3': None, 'N1': None, 'N2': None, '일자': None, '할인율': None, '주차수': None}
    for col in range(7, 18):  # G(7) ~ Q(17)
        cell_value = ws.Cells(row_index, col).Value
        if not cell_value:
            continue
        cell_str = str(cell_value)
        if 'LV1' in cell_str:
            values['LV1'] = cell_str
        elif 'LV2' in cell_str:
            values['LV2'] = cell_str
        elif 'LV3' in cell_str:
            values['LV3'] = cell_str
        elif 'N1' in cell_str:
            values['N1'] = cell_str
        elif 'N2' in cell_str:
            values['N2'] = cell_str
        elif '일' in cell_str and '자' in cell_str or '월' in cell_str or '일' in cell_str:
            values['일자'] = cell_str
        elif '할인' in cell_str or '%' in cell_str:
            values['할인율'] = cell_str
        elif '주차' in cell_str or '주' in cell_str:
            values['주차수'] = cell_str
    return values

# ✅ 각 행마다 분석 후 W열부터 값 채우기
for i in range(2, row_count + 1):  # 1행은 헤더
    extracted = extract_column_values(i)
    ws.Cells(i, 23).Value = extracted['LV1']     # W열
    ws.Cells(i, 24).Value = extracted['LV2']     # X열
    ws.Cells(i, 25).Value = extracted['LV3']     # Y열
    ws.Cells(i, 26).Value = extracted['N1']      # Z열
    ws.Cells(i, 27).Value = extracted['N2']      # AA열
    ws.Cells(i, 28).Value = extracted['일자']    # AB열
    ws.Cells(i, 29).Value = extracted['할인율']  # AC열
    ws.Cells(i, 30).Value = extracted['주차수']  # AD열

print("✅ 모든 데이터 자동 정리 완료되었습니다!")
```
