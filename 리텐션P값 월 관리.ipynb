```python
import pandas as pd
import glob
import os
import tkinter as tk
from tkinter import ttk

# 폴더 경로 설정
folder_path = r'D:\ㅁ 리텐션실적\2024년 5월\취합'

# 디렉토리가 존재하지 않으면 생성
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 모든 cs 파일 경로 가져오기
all_files = glob.glob(folder_path + "/*.cs")

# 모든 파일을 읽어 데이터프레임으로 병합
all_data = pd.DataFrame()
for file in all_files:
    df = pd.read_csv(file, encoding='cp949')
    all_data = pd.concat([all_data, df], ignore_index=True)

# '계약번호'와 '리텐션 유형'을 기준으로 중복 제거
all_data = all_data.drop_duplicates(subset=['계약번호', '리텐션 유형'])

# 모든 열이 비어있는 행 제거
all_data = all_data.dropna(how='all')

# '리텐션 유형' 열이 있는 경우 지정된 순서로 정렬
if '리텐션 유형' in all_data.columns:
    category_order = ['면제', '인하', '할인']
    all_data['리텐션 유형'] = pd.Categorical(all_data['리텐션 유형'], categories=category_order, ordered=True)
    all_data = all_data.sort_values('리텐션 유형')

# '관리자사' 필터에 사용할 고유 값 추출
admin_categories = all_data['관리지사'].dropna().unique().tolist()

# 병합된 데이터를 cs 파일로 저장 (CP949 인코딩 적용)
output_path = os.path.join(folder_path, "리텐션P값 5월 병합.cs")
all_data.to_csv(output_path, index=False, encoding='cp949')

print(f"병합된 파일이 {output_path}에 저장되었습니다.")

# GUI 생성
class RetentionApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Retention Data Viewer")

        # 스타일 설정
        style = ttk.Style()
        style.configure("Treeview", font=("Helvetica", 10), rowheight=25)
        style.configure("Treeview.Heading", font=("Helvetica", 12, "bold"))

        self.label1 = ttk.Label(root, text="리텐션 유형을 선택하세요:")
        self.label1.pack(pady=10)

        self.combo1 = ttk.Combobox(root, values=["전체"] + category_order)
        self.combo1.current(0)
        self.combo1.pack(pady=10)

        self.label2 = ttk.Label(root, text="관리지사를 선택하세요:")
        self.label2.pack(pady=10)

        self.combo2 = ttk.Combobox(root, values=["전체"] + admin_categories)
        self.combo2.current(0)
        self.combo2.pack(pady=10)

        self.button = ttk.Button(root, text="미리보기", command=self.preview_data)
        self.button.pack(pady=10)

        self.tree = ttk.Treeview(root, columns=list(all_data.columns), show='headings')
        for col in all_data.columns:
            self.tree.heading(col, text=col)
            self.tree.column(col, anchor="w")
       
        self.tree.pack(fill='both', expand=True)

        self.scroll_y = ttk.Scrollbar(root, orient='vertical', command=self.tree.yview)
        self.scroll_y.pack(side='right', fill='y')
        self.tree.configure(yscrollcommand=self.scroll_y.set)

        self.scroll_x = ttk.Scrollbar(root, orient='horizontal', command=self.tree.xview)
        self.scroll_x.pack(side='bottom', fill='x')
        self.tree.configure(xscrollcommand=self.scroll_x.set)

        # 하단 요약 표시
        self.summary_frame = ttk.Frame(root)
        self.summary_frame.pack(pady=10)

        self.summary_label = ttk.Label(self.summary_frame, text="", font=("Helvetica", 12))
        self.summary_label.pack()

    def preview_data(self):
        for row in self.tree.get_children():
            self.tree.delete(row)
       
        selected_category = self.combo1.get()
        selected_admin = self.combo2.get()

        if selected_category == "전체" and selected_admin == "전체":
            filtered_data = all_data
        elif selected_category == "전체":
            filtered_data = all_data[all_data['관리지사'] == selected_admin]
        elif selected_admin == "전체":
            filtered_data = all_data[all_data['리텐션 유형'] == selected_category]
        else:
            filtered_data = all_data[(all_data['리텐션 유형'] == selected_category) & (all_data['관리지사'] == selected_admin)]

        for index, row in filtered_data.iterrows():
            self.tree.insert("", "end", values=list(row))

        # 요약 정보 계산
        total_contracts = filtered_data['계약번호'].nunique()
        total_exemption = filtered_data['총 면제금액'].sum()
        total_before = filtered_data['변경 전 월정료'].sum()
        total_after = filtered_data['변경 후 월정료'].sum()
        total_discount = filtered_data['인하/할인액'].sum()

        summary_text = f"계약번호 기준 건수: {total_contracts}\n총면제금액: {total_exemption}\n변경 전 월정료: {total_before}\n변경 후 월정료: {total_after}\n인하/할인액 합계: {total_discount}"
        self.summary_label.config(text=summary_text)

# Tkinter 루프 시작
root = tk.Tk()
app = RetentionApp(root)
root.mainloop()
```

```python
import pandas as pd
import glob
import os
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 폴더 경로 설정
folder_path = r'D:\ㅁ 리텐션실적\2024년 5월\취합'

# 디렉토리가 존재하지 않으면 생성
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 모든 cs 파일 경로 가져오기
all_files = glob.glob(folder_path + "/*.cs")

# 모든 파일을 읽어 데이터프레임으로 병합
all_data = pd.DataFrame()
for file in all_files:
    df = pd.read_csv(file, encoding='cp949')
    all_data = pd.concat([all_data, df], ignore_index=True)

# '계약번호'와 '리텐션 유형'을 기준으로 중복 제거
all_data = all_data.drop_duplicates(subset=['계약번호', '리텐션 유형'])

# 모든 열이 비어있는 행 제거
all_data = all_data.dropna(how='all')

# NaN 값이 있는 열 제거
all_data = all_data.dropna(axis=1, how='all')

# '리텐션 유형' 열이 있는 경우 지정된 순서로 정렬
if '리텐션 유형' in all_data.columns:
    category_order = ['면제', '인하', '할인']
    all_data['리텐션 유형'] = pd.Categorical(all_data['리텐션 유형'], categories=category_order, ordered=True)
    all_data = all_data.sort_values('리텐션 유형')

# '관리지사' 필터에 사용할 고유 값 추출
if '관리지사' in all_data.columns:
    admin_categories = ['전체'] + all_data['관리지사'].dropna().unique().tolist()
else:
    admin_categories = []

# 날짜 형식 수정
if '날짜' in all_data.columns:
    all_data['날짜'] = pd.to_datetime(all_data['날짜']).dt.strftime('%Y-%m-%d')

# 병합된 데이터를 cs 파일로 저장 (CP949 인코딩 적용)
output_path = os.path.join(folder_path, "리텐션P값 5월 병합.cs")
all_data.to_csv(output_path, index=False, encoding='cp949')

print(f"병합된 파일이 {output_path}에 저장되었습니다.")

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# GUI 생성
class RetentionApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Retention Data Viewer")

        # 스타일 설정
        style = ttk.Style()
        style.configure("Treeview", font=("Helvetica", 10), rowheight=25, borderwidth=1, relief="solid")
        style.configure("Treeview.Heading", font=("Helvetica", 12, "bold"), borderwidth=1, relief="solid")
        style.configure("TButton", font=("Helvetica", 10))

        self.label1 = ttk.Label(root, text="리텐션 유형을 선택하세요:", font=("Helvetica", 12))
        self.label1.pack(pady=10)

        self.combo1 = ttk.Combobox(root, values=["전체"] + category_order, font=("Helvetica", 10))
        self.combo1.current(0)
        self.combo1.pack(pady=10)

        self.label2 = ttk.Label(root, text="관리지사를 선택하세요:", font=("Helvetica", 12))
        self.label2.pack(pady=10)

        self.combo2 = ttk.Combobox(root, values=admin_categories, font=("Helvetica", 10))
        self.combo2.current(0)
        self.combo2.pack(pady=10)

        self.button = ttk.Button(root, text="미리보기", command=self.preview_data, style="TButton")
        self.button.pack(pady=10)

        # 체크박스를 위한 프레임
        self.check_frame = ttk.Frame(root)
        self.check_frame.pack(pady=10)

        self.check_vars = {}
        for col in all_data.columns:
            var = tk.BooleanVar(value=True)
            chk = tk.Checkbutton(self.check_frame, text=col, variable=var, font=("Helvetica", 10))
            chk.pack(side=tk.LEFT)
            self.check_vars[col] = var

        self.tree = ttk.Treeview(root, columns=list(all_data.columns), show='headings')
        self.update_treeview_columns(list(all_data.columns))
       
        self.tree.pack(fill='both', expand=True)

        self.scroll_y = ttk.Scrollbar(root, orient='vertical', command=self.tree.yview)
        self.scroll_y.pack(side='right', fill='y')
        self.tree.configure(yscrollcommand=self.scroll_y.set)

        self.scroll_x = ttk.Scrollbar(root, orient='horizontal', command=self.tree.xview)
        self.scroll_x.pack(side='bottom', fill='x')
        self.tree.configure(xscrollcommand=self.scroll_x.set)

        # 하단 요약 표시
        self.summary_frame = ttk.Frame(root)
        self.summary_frame.pack(pady=10)

        self.summary_label = ttk.Label(self.summary_frame, text="", font=("Helvetica", 12), borderwidth=1, relief="solid")
        self.summary_label.pack()

        # 시각화 유형 선택
        self.chart_type_label = ttk.Label(root, text="차트 유형을 선택하세요:", font=("Helvetica", 12))
        self.chart_type_label.pack(pady=10)

        self.chart_type = ttk.Combobox(root, values=["막대 그래프", "원형 그래프", "선 그래프"], font=("Helvetica", 10))
        self.chart_type.current(0)
        self.chart_type.pack(pady=10)

        # 시각화 영역
        self.fig, self.ax = plt.subplots(figsize=(10, 5))
        self.canvas = FigureCanvasTkAgg(self.fig, master=root)
        self.canvas.get_tk_widget().pack(fill='both', expand=True)

    def update_treeview_columns(self, columns):
        self.tree["columns"] = columns
        for col in columns:
            self.tree.heading(col, text=col)
            self.tree.column(col, anchor="w")

    def preview_data(self):
        selected_columns = [col for col in all_data.columns if self.check_vars[col].get()]
        self.update_treeview_columns(selected_columns)

        for row in self.tree.get_children():
            self.tree.delete(row)
       
        selected_category = self.combo1.get()
        selected_admin = self.combo2.get()

        if selected_category == "전체" and selected_admin == "전체":
            filtered_data = all_data[selected_columns]
        elif selected_category == "전체":
            filtered_data = all_data[all_data['관리지사'] == selected_admin][selected_columns]
        elif selected_admin == "전체":
            filtered_data = all_data[all_data['리텐션 유형'] == selected_category][selected_columns]
        else:
            filtered_data = all_data[(all_data['리텐션 유형'] == selected_category) & (all_data['관리지사'] == selected_admin)][selected_columns]

        for index, row in filtered_data.iterrows():
            self.tree.insert("", "end", values=list(row))

        # 요약 정보 계산
        total_contracts = filtered_data['계약번호'].nunique() if '계약번호' in filtered_data.columns else 0
        total_exemption = filtered_data['총면제금액'].sum() if '총면제금액' in filtered_data.columns else 0
        total_before = filtered_data['변경 전 월정료'].sum() if '변경 전 월정료' in filtered_data.columns else 0
        total_after = filtered_data['변경 후 월정료'].sum() if '변경 후 월정료' in filtered_data.columns else 0
        total_discount = filtered_data['인하/할인액'].sum() if '인하/할인액' in filtered_data.columns else 0

        summary_text = f"계약번호 기준 건수: {total_contracts}\n총면제금액: {total_exemption}\n변경 전 월정료 합계: {total_before}\n변경 후 월정료 합계: {total_after}\n인하/할인액 합계: {total_discount}"
        self.summary_label.config(text=summary_text)

        # 시각화 업데이트
        self.ax.clear()
        if '관리지사' in filtered_data.columns and '리텐션 유형' in filtered_data.columns:
            summary = filtered_data.groupby(['관리지사', '리텐션 유형']).agg(
                계약번호기준건수=('계약번호', 'nunique'),
                총면제금액=('총면제금액', 'sum'),
                인하할인액=('인하/할인액', 'sum')
            ).reset_index()

            chart_type = self.chart_type.get()

            if chart_type == "막대 그래프":
                summary.pivot(index='관리지사', columns='리텐션 유형', values='계약번호기준건수').plot(kind='bar',

 ax=self.ax)
                self.ax.set_ylabel('건수')
            elif chart_type == "원형 그래프":
                pie_data = summary.groupby('관리지사')['계약번호기준건수'].sum()
                pie_data.plot(kind='pie', ax=self.ax, autopct='%1.1f%%')
                self.ax.set_ylabel('')
            elif chart_type == "선 그래프":
                summary.pivot(index='관리지사', columns='리텐션 유형', values='계약번호기준건수').plot(kind='line', marker='o', ax=self.ax)
                self.ax.set_ylabel('건수')

            self.ax.set_title('관리지사별 리텐션 유형별 계약번호 기준 건수')
            self.ax.set_xlabel('관리지사')
            self.fig.tight_layout()
            self.canvas.draw()

# Tkinter 루프 시작
root = tk.Tk()
app = RetentionApp(root)
root.mainloop()
```
