```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle

# 1. Malgun Gothic   (Windows    )
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  #    

#     
data = {
    '': ['/', '/', '/', '/', '/', '/'],
    '': [8, 16, 10, 6, 16, 6],
    '': [94, 46, 75, 67, 76, 124],
    '': [11, 6, 23, 75, 67, 85],
    '': [113, 68, 108, 148, 159, 215],
    '': [14, 8, 13, 18, 20, 27]
}
df = pd.DataFrame(data)

# 2. :   +  
fig, ax1 = plt.subplots(figsize=(14, 8))

#     ( : edgecolor='none')
df.set_index('')[['', '', '']].plot(kind='bar', stacked=True, ax=ax1, color=['#1f77b4', '#2ca02c', '#ff7f0e'], edgecolor='none')
ax1.set_xlabel('', fontsize=14)
ax1.set_ylabel(' ', fontsize=14)
ax1.set_title('  ', fontsize=18)

#  0   
ax1.set_xticklabels(df[''], rotation=0, fontsize=12, ha='center')

#       (  )
for container in ax1.containers:
    ax1.bar_label(container, label_type='center', fontsize=14, color='white', weight='bold')  #    

#   y    ()
ax2 = ax1.twinx()
sns.lineplot(x='', y='', data=df, marker='o', color='black', ax=ax2, linewidth=2)

#   y       
max_bar_height = df[''].max() + 10  #    +   
ax2.set_ylim(0, max_bar_height)  # y     
ax2.set_ylabel(' (%)', fontsize=14)

#      
for i, value in enumerate(df['']):
    bar_height = df[''][i]  #     ( )
    label_y_position = bar_height  #     
    ax2.text(i, label_y_position, f'{value}%', ha='center', va='bottom', fontsize=14, color='black', weight='bold')

#   3    ( )
lowest_regions = df.nsmallest(3, '')[''].tolist()  #   3  
print(f"  3 : {lowest_regions}")

#     (       )
for i, region in enumerate(df['']):
    if region in lowest_regions:
        bar_height = df[''][i]  #     ( )
        #    
        box_x = i - 0.4  #  x  (    )
        box_y = 0        #  y  ( )
        box_width = 0.8  #   (  )
        box_height = bar_height  #   (   )

        #  
        ax1.add_patch(Rectangle((box_x, box_y), box_width, box_height, fill=False, edgecolor='red', linewidth=2, linestyle='--'))

#  
# , ,   
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
#    
ax1.spines['bottom'].set_visible(True)
ax1.spines['bottom'].set_linewidth(1.5)  #    

# ax2  
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['bottom'].set_visible(True)
ax2.spines['bottom'].set_linewidth(1.5)  #    

#  
ax1.grid(False)
ax2.grid(False)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle

# 1. Seaborn    
sns.set_style("whitegrid")
sns.set_palette("Set2")

# 2. Malgun Gothic   (Windows    )
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  #    

#     
data = {
    '': ['/', '/', '/', '/', '/', '/'],
    '': [8, 16, 10, 6, 16, 6],
    '': [94, 46, 75, 67, 76, 124],
    '': [11, 6, 23, 75, 67, 85],
    '': [113, 68, 108, 148, 159, 215],
    '': [14, 8, 13, 18, 20, 27]
}
df = pd.DataFrame(data)

# 3. :   +  
fig, ax1 = plt.subplots(figsize=(14, 8))

#     ( : edgecolor='none',   )
hatches = ['/', '\\', 'x']
for i, col in enumerate(['', '', '']):
    ax1.bar(df[''], df[col], label=col, hatch=hatches[i], edgecolor='white')

ax1.set_xlabel('', fontsize=14)
ax1.set_ylabel(' ', fontsize=14)
ax1.set_title('    ', fontsize=18)

#  0   
ax1.set_xticklabels(df[''], rotation=0, fontsize=12, ha='center')

#       (  )
for container in ax1.containers:
    ax1.bar_label(container, label_type='center', fontsize=14, color='black', weight='bold')  #    

#   y    ()
ax2 = ax1.twinx()
sns.lineplot(x='', y='', data=df, marker='o', color='#d62728', ax=ax2, linewidth=2.5, markersize=10, markerfacecolor='white')

#   y       
max_bar_height = df[''].max() + 20  #    +   
ax2.set_ylim(0, max_bar_height)  # y     
ax2.set_ylabel(' (%)', fontsize=14)

#       ( )
for i, value in enumerate(df['']):
    bar_height = df[''][i]  #     ( )
    #    5    
    label_y_position = bar_height + 5 if value < bar_height else value + 2
    ax2.text(i, label_y_position, f'{value}%', ha='center', va='bottom', fontsize=14, color='#d62728', weight='bold')  #    

#   3    ( )
lowest_regions = df.nsmallest(3, '')[''].tolist()  #   3  
print(f"  3 : {lowest_regions}")

#     (       )
for i, region in enumerate(df['']):
    if region in lowest_regions:
        bar_height = df[''][i]  #     ( )
        #    
        box_x = i - 0.4  #  x  (    )
        box_y = 0        #  y  ( )
        box_width = 0.8  #   (  )
        box_height = bar_height  #   (   )

        #  
        ax1.add_patch(Rectangle((box_x, box_y), box_width, box_height, fill=False, edgecolor='red', linewidth=2, linestyle='--'))

#     
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_visible(True)
ax1.spines['bottom'].set_linewidth(1.5)  #    

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['bottom'].set_visible(True)
ax2.spines['bottom'].set_linewidth(1.5)  #    

#  
ax1.grid(False)
ax2.grid(False)

#  
ax1.legend(loc='upper left', fontsize=12)

plt.tight_layout()
plt.show()
```
