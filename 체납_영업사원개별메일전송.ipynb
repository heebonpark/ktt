```python
import pandas as pd
import os
import chardet
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication
from email.mime.image import MIMEImage
import openpyxl
from openpyxl.styles import PatternFill, Font, Border, Side
import tkinter as tk
from tkinter import simpledialog, Listbox, MULTIPLE, messagebox, Scrollbar, RIGHT, Y, END, Toplevel, StringVar, Entry, Label, Button, Text, WORD, Scale
from tkinter.ttk import Progressbar
import threading
import time

# Sample data to represent the initial agent information
agent_data = [
    {"": "", "": "G000101", "": "7864", "": "", "1": "7864", " SP": "", "": "seong.js@kt.com"},
    {"": "", "": "G000102", "": "85669", "": "", "1": "85669", " SP": "", "": "#N/A"},
    {"": "", "": "G000103", "": "4159", "": "", "1": "4159", " SP": "", "": "youngtaek.park@kt.com"},
    # ... (other entries)
]

email_addresses = {
    "": "82005897@kt.com",
    "": "82003769@kt.com",
    "": "82004038@kt.com",
    "": "kimin.nam@kt.com",
    "": "sj2.moon@kt.com",
    "": "82004734@kt.com",
    "": "jae.yong@kt.com",
    "": "82004907@kt.com",
    "": "uikyum.kim@kt.com",
    "": "youngtaek.park@kt.com",
    "": "ruga.sg@kt.com",
    "": "seong.js@kt.com",
    "": "82004091@kt.com",
    "": "eunseok.seo@kt.com",
    "": "82004793@kt.com",
    "": "jungdo.lee@kt.com",
    "": "82004653@kt.com",
    "": "82002883@kt.com",
    "": "82004316@kt.com",
    "": "hyeongrae.seong@kt.com",
    "": "changjun.yoo@kt.com",
    "": "useok.choi@kt.com",
    "": "jinhoe.kim@kt.com",
    "": "kyoungsuk.lim@kt.com",
    "": "82004661@kt.com",
    "": "soung.bae@kt.com",
    "": "hyundon.kim@kt.com",
    "": "82004150@kt.com",
    "": "82141316@kt.com",
    "": "hyounbae.lee@kt.com",
    "": "wanyong.kim@kt.com",
    "": "82107343@kt.com",
    "": "82077078@kt.com",
    "": "82096643@kt.com",
    "": "kwangho12.kim@kt.com",
    "": "sang_tae.kim@kt.com",
    "": "sys7779.son@kt.com"
}

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

def read_file_with_encoding(file_path, header_row, **kwargs):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{file_path}   .")
    encoding = detect_encoding(file_path)
    return pd.read_csv(file_path, encoding=encoding, header=header_row, **kwargs)

def send_emails(selected_agents, email_addresses, email_body_template, sender_email, sender_password, progress, root, font_size):
    smtp_server = "smtp.naver.com"
    smtp_port = 587

    header_fill = PatternFill(start_color="4F4F4F", end_color="4F4F4F", fill_type="solid")
    header_font = Font(color="FFFFFF", bold=True)
    thin_border = Border(left=Side(style='thin', color='000000'),
                         right=Side(style='thin', color='000000'),
                         top=Side(style='thin', color='000000'),
                         bottom=Side(style='thin', color='000000'))

    summary_path = 'D:\\\\\\ 2 .cs'
    new_detail_path = 'D:\\\\\\ 2 .cs'
   
    header_row = 0

    df_summary = read_file_with_encoding(summary_path, header_row)
    df_new_detail = read_file_with_encoding(new_detail_path, header_row)

    df_summary = df_summary.loc[:, ~df_summary.columns.str.contains('Unnamed')]
    df_summary.columns = df_summary.columns.str.strip().str.replace('\n', '')

    df_new_detail.columns = df_new_detail.columns.str.strip()

    df_summary = df_summary.fillna('')
    df_new_detail = df_new_detail.fillna('')

    missing_emails = []
    total_agents = len(selected_agents)
   
    for idx, agent in enumerate(selected_agents):
        agent_email = email_addresses.get(agent)
        if not agent_email or agent_email == "#N/A":
            missing_emails.append(agent)
            continue

        summary_filtered = df_summary[df_summary[''] == agent].copy()
        new_detail_filtered = df_new_detail[df_new_detail[''] == agent].copy()

        if summary_filtered.empty or new_detail_filtered.empty:
            print(f"{agent}   .")
            continue

        summary_filtered[''] = ''
        summary_filtered['  '] = ''
       
        summary_html = summary_filtered.to_html(index=False)
        new_detail_html = new_detail_filtered.to_html(index=False)

        excel_output_path = f"D:\\\\\\{agent}_.xlsx"
       
        with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
            summary_filtered.to_excel(writer, sheet_name="", index=False)
            new_detail_filtered.to_excel(writer, sheet_name="", index=False)
       
        wb = openpyxl.load_workbook(excel_output_path)
       
        for sheet in wb.sheetnames:
            ws = wb[sheet]
            for cell in ws[1]:
                cell.fill = header_fill
                cell.font = header_font
                cell.border = thin_border
       
        wb.save(excel_output_path)
       
        email_body = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <style>
                table {{
                    width: 100%;
                    border-collapse: collapse;
                }}
                th, td {{
                    border: 1px solid black;
                    padding: 5px;
                    text-align: left;
                    font-size: {font_size}px;
                }}
                th {{
                    background-color: #f2f2f2;
                }}
            </style>
        </head>
        <body>
            <p style="font-size:{font_size}px;">{agent}, .</p>
            <p style="font-size:{font_size}px;">  .</p>
            <p style="font-size:{font_size}px;">  (   )   </p>
            <p style="font-size:{font_size}px;">    , ,    .</p>
            <p style="font-size:{font_size}px;"> (620 )     .</p>
            <div style="max-height: 400px; overflow-y: auto;">
                {summary_html}
                <br>
                {new_detail_html}
            </div>
            <br>
            <img src="cid:image1">
        </body>
        </html>
        """
       
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = agent_email
        msg['Subject'] = f"{agent}   "
        msg.attach(MIMEText(email_body, 'html', 'utf-8'))

        attach_files = [
            excel_output_path
        ]

        for file in attach_files:
            with open(file, 'rb') as f:
                part = MIMEApplication(f.read(), Name=os.path.basename(file))
            part['Content-Disposition'] = f'attachment; filename="{os.path.basename(file)}"'
            msg.attach(part)

        # Attach the resized image
        try:
            img_path = 'D:\\\\\\IMG_7573.jpeg'
            with Image.open(img_path) as img:
                img = img.resize((img.width // 4, img.height // 4))  # Resize image to 1/4 size
                img.save('D:\\\\\\resized_image.jpeg')

            with open('D:\\\\\\resized_image.jpeg', 'rb') as img:
                mime_image = MIMEImage(img.read())
                mime_image.add_header('Content-ID', '<image1>')
                msg.attach(mime_image)
        except FileNotFoundError:
            print("    .    .")
       
        try:
            with smtplib.SMTP(smtp_server, smtp_port) as server:
                server.starttls()
                time.sleep(1)  #    
                server.login(sender_email, sender_password)
                time.sleep(1)  #    
                server.sendmail(sender_email, agent_email, msg.as_string())
                print(f"{agent}   .")
        except smtplib.SMTPAuthenticationError as e:
            print(f"   .    : {e}")
            break  #     
        except Exception as e:
            print(f"{agent}   : {e}")

        # Update progress bar
        progress['value'] = ((idx + 1) / total_agents) * 100
        root.update_idletasks()

    if missing_emails:
        print("     :")
        for agent in missing_emails:
            print(f" - {agent}")

def start_sending(selected_agents, email_addresses, email_body_template, sender_email, sender_password, root, font_size):
    progress_window = Toplevel(root)
    progress_window.title("Progress")
    tk.Label(progress_window, text="Sending emails, please wait...").pack(pady=10)
    progress = Progressbar(progress_window, orient=tk.HORIZONTAL, length=300, mode='determinate')
    progress.pack(pady=20)
   
    # Start sending emails in a separate thread
    threading.Thread(target=send_emails, args=(selected_agents, email_addresses, email_body_template, sender_email, sender_password, progress, root, font_size)).start()

def main():
    root = tk.Tk()
    root.withdraw()

    sender_email = "bough38@naver.com"
    sender_password = simpledialog.askstring("Input", "  :", show='*')

    agents = [agent[""] for agent in agent_data]

    def add_agent():
        new_window = Toplevel(root)
        new_window.title("   ")

        def save_agent():
            agent_info = {
                "": entry_office.get(),
                "": entry_zone.get(),
                "": entry_id.get(),
                "": entry_name.get(),
                "1": entry_id1.get(),
                " SP": entry_sp.get(),
                "": entry_email.get()
            }
            agent_data.append(agent_info)
            agents.append(agent_info[""])
            email_addresses[agent_info[""]] = agent_info[""]
            checklist.insert(tk.END, agent_info[""])
            new_window.destroy()
            update_selected_agents()

        Label(new_window, text="").grid(row=0, column=0)
        entry_office = Entry(new_window)
        entry_office.grid(row=0, column=1)

        Label(new_window, text="").grid(row=1, column=0)
        entry_zone = Entry(new_window)
        entry_zone.grid(row=1, column=1)

        Label(new_window, text="").grid(row=2, column=0)
        entry_id = Entry(new_window)
        entry_id.grid(row=2, column=1)

        Label(new_window, text="").grid(row=3, column=0)
        entry_name = Entry(new_window)
        entry_name.grid(row=3, column=1)

        Label(new_window, text="1").grid(row=4, column=0)
        entry_id1 = Entry(new_window)
        entry_id1.grid(row=4, column=1)

        Label(new_window, text=" SP").grid(row=5, column=0)
        entry_sp = Entry(new_window)
        entry_sp.grid(row=5, column=1)

        Label(new_window, text="").grid(row=6, column=0)
        entry_email = Entry(new_window)
        entry_email.grid(row=6, column=1)

        Button(new_window, text="", command=save_agent).grid(row=7, column=0, columnspan=2)

    def edit_agent():
        selected_indices = checklist.curselection()
        if not selected_indices:
            messagebox.showwarning("Warning", "  .")
            return
        index = selected_indices[0]
        agent_name = checklist.get(index)
        agent_info = next(agent for agent in agent_data if agent[""] == agent_name)

        edit_window = Toplevel(root)
        edit_window.title("   ")

        def save_changes():
            agent_info[""] = entry_office.get()
            agent_info[""] = entry_zone.get()
            agent_info[""] = entry_id.get()
            agent_info[""] = entry_name.get()
            agent_info["1"] = entry_id1.get()
            agent_info[" SP"] = entry_sp.get()
            agent_info[""] = entry_email.get()

            agents[index] = agent_info[""]
            email_addresses[agent_info[""]] = agent_info[""]
            checklist.delete(index)
            checklist.insert(index, agent_info[""])
            edit_window.destroy()
            update_selected_agents()

        Label(edit_window, text="").grid(row=0, column=0)
        entry_office = Entry(edit_window)
        entry_office.insert(0, agent_info[""])
        entry_office.grid(row=0, column=1)

        Label(edit_window, text="").grid(row=1, column=0)
        entry_zone = Entry(edit_window)
        entry_zone.insert(0, agent_info[""])
        entry_zone.grid(row=1, column=1)

        Label(edit_window, text="").grid(row=2, column=0)
        entry_id = Entry(edit_window)
        entry_id.insert(0, agent_info[""])
        entry_id.grid(row=2, column=1)

        Label(edit_window, text="").grid(row=3, column=0)
        entry_name = Entry(edit_window)
        entry_name.insert(0, agent_info[""])
        entry_name.grid(row=3, column=1)

        Label(edit_window, text="1").grid(row=4, column=0)
        entry_id1 = Entry(edit_window)
        entry_id1.insert(0, agent_info["1"])
        entry_id1.grid(row=4, column=1)

        Label(edit_window, text=" SP").grid(row=5, column=0)
        entry_sp = Entry(edit_window)
        entry_sp.insert(0, agent_info[" SP"])
        entry_sp.grid(row=5, column=1)

        Label(edit_window, text="").grid(row=6, column=0)
        entry_email = Entry(edit_window)
        entry_email.insert(0, agent_info[""])
        entry_email.grid(row=6, column=1)

        Button(edit_window, text="", command=save_changes).grid(row=7, column=0, columnspan=2)

    def delete_agent():
        selected_indices = checklist.curselection()
        for index in reversed(selected_indices):
            agent_name = checklist.get(index)
            agents.remove(agent_name)
            email_addresses.pop(agent_name, None)
            agent_data[:] = [agent for agent in agent_data if agent[""] != agent_name]
            checklist.delete(index)
        update_selected_agents()

    def update_selected_agents():
        selected_agents_listbox.delete(0, tk.END)
        selected_indices = checklist.curselection()
        for i in selected_indices:
            selected_agents_listbox.insert(tk.END, f"{checklist.get(i)} ({email_addresses.get(checklist.get(i), 'N/A')})")

    def select_all():
        checklist.select_set(0, tk.END)
        update_selected_agents()

    def deselect_all():
        checklist.select_clear(0, tk.END)
        update_selected_agents()

    def send_selected():
        selected_indices = checklist.curselection()
        selected_agents = [agents[i] for i in selected_indices]
        if selected_agents:
            email_body_template = email_body_var.get()
            font_size = font_size_var.get()
            start_sending(selected_agents, email_addresses, email_body_template, sender_email, sender_password, root, font_size)
        else:
            messagebox.showwarning("Warning", "    .")

    def preview_email():
        selected_indices = checklist.curselection()
        selected_agents = [agents[i] for i in selected_indices]
        if selected_agents:
            email_body_template = email_body_var.get()
            font_size = font_size_var.get()
            preview_window = Toplevel(root)
            preview_window.title("Email Preview")
            email_preview = tk.Text(preview_window, wrap=WORD)
            email_preview.pack(expand=True, fill=tk.BOTH)
            for agent in selected_agents:
                email_body = email_body_template.replace("{agent}", agent).replace("{font_size}", str(font_size))
                email_preview.insert(tk.END, f"To: {email_addresses.get(agent, 'N/A')}\n\n{email_body}\n\n{'-'*50}\n\n")
        else:
            messagebox.showwarning("Warning", "    .")

    root.deiconify()
    root.title("Select Agents to Send Emails")
    root.geometry("1000x600")
   
    left_frame = tk.Frame(root)
    left_frame.pack(side=tk.LEFT, padx=10, pady=10, fill=tk.BOTH, expand=True)
   
    right_frame = tk.Frame(root)
    right_frame.pack(side=tk.RIGHT, padx=10, pady=10, fill=tk.BOTH, expand=True)
   
    tk.Label(left_frame, text="   :", font=("Helvetica", 12)).pack(pady=10)
   
    left_inner_frame = tk.Frame(left_frame)
    left_inner_frame.pack(pady=10, fill=tk.BOTH, expand=True)

    scrollbar = Scrollbar(left_inner_frame, orient=tk.VERTICAL)
    checklist = Listbox(left_inner_frame, selectmode=MULTIPLE, yscrollcommand=scrollbar.set, font=("Helvetica", 10))
    scrollbar.config(command=checklist.yview)
    scrollbar.pack(side=RIGHT, fill=Y)
    checklist.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
   
    for agent in agents:
        checklist.insert(tk.END, agent)

    button_frame = tk.Frame(left_frame)
    button_frame.pack(pady=5)
    tk.Button(button_frame, text=" ", command=select_all, font=("Helvetica", 10)).pack(side=tk.LEFT, padx=5)
    tk.Button(button_frame, text=" ", command=deselect_all, font=("Helvetica", 10)).pack(side=tk.LEFT, padx=5)

    tk.Button(left_frame, text="", command=add_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=edit_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=delete_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=send_selected, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=preview_email, font=("Helvetica", 10)).pack(pady=5)

    tk.Label(right_frame, text="   :", font=("Helvetica", 12)).pack(pady=10)
   
    selected_agents_listbox = Listbox(right_frame, font=("Helvetica", 10))
    selected_agents_listbox.pack(fill=tk.BOTH, expand=True)
   
    tk.Label(right_frame, text=" :", font=("Helvetica", 12)).pack(pady=10)
   
    email_body_var = StringVar()
    email_body_var.set("""<!DOCTYPE html>
<html>
<body>
    <p style="font-size:{font_size}px;">{agent}, .</p>
    <p style="font-size:{font_size}px;">  .</p>
    <p style="font-size:{font_size}px;">  (   )   </p>
    <p style="font-size:{font_size}px;">    , ,    .</p>
    <p style="font-size:{font_size}px;"> (620 )     .</p>
</body>
</html>
""")
    email_body_entry = tk.Entry(right_frame, textvariable=email_body_var, font=("Helvetica", 10), width=50)
    email_body_entry.pack(pady=5)

    tk.Label(right_frame, text=" :", font=("Helvetica", 12)).pack(pady=10)
    font_size_var = tk.IntVar(value=12)
    font_size_scale = Scale(right_frame, from_=8, to=20, orient=tk.HORIZONTAL, variable=font_size_var)
    font_size_scale.pack(pady=5)

    checklist.bind("<<ListboxSelect>>", lambda event: update_selected_agents())

    # Exit after 10 minutes (600000 ms)
    root.after(600000, lambda: root.quit())
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
import chardet
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication
from email.mime.image import MIMEImage
from PIL import Image
import openpyxl
from openpyxl.styles import PatternFill, Font, Border, Side
import tkinter as tk
from tkinter import simpledialog, Listbox, MULTIPLE, messagebox, Scrollbar, RIGHT, Y, END, Toplevel, StringVar, Entry, Label, Button, WORD, Scale
from tkinter.ttk import Progressbar
import threading
import time
import numpy as np

# Sample data to represent the initial agent information
agent_data = [
    {"": "", "": "G000101", "": "7864", "": "", "1": "7864", " SP": "", "": "heebon.park@kt.com"},
    {"": "", "": "G000102", "": "85669", "": "", "1": "85669", " SP": "", "": "bough38@naver.com"},
    {"": "", "": "G000103", "": "4159", "": "", "1": "4159", " SP": "", "": "youngtaek.park@kt.com"},
    # ... (other entries)
]

email_addresses = {
    "": "82005897@kt.com",
    "": "82003769@kt.com",
    "": "82004038@kt.com",
    "": "kimin.nam@kt.com",
    "": "sj2.moon@kt.com",
    "": "82004734@kt.com",
    "": "jae.yong@kt.com",
    "": "82004907@kt.com",
    "": "uikyum.kim@kt.com",
    "": "youngtaek.park@kt.com",
    "": "ruga.sg@kt.com",
    "": "seong.js@kt.com",
    "": "82004091@kt.com",
    "": "eunseok.seo@kt.com",
    "": "82004793@kt.com",
    "": "jungdo.lee@kt.com",
    "": "82004653@kt.com",
    "": "82002883@kt.com",
    "": "82004316@kt.com",
    "": "hyeongrae.seong@kt.com",
    "": "changjun.yoo@kt.com",
    "": "useok.choi@kt.com",
    "": "jinhoe.kim@kt.com",
    "": "kyoungsuk.lim@kt.com",
    "": "82004661@kt.com",
    "": "soung.bae@kt.com",
    "": "hyundon.kim@kt.com",
    "": "82004150@kt.com",
    "": "82141316@kt.com",
    "": "hyounbae.lee@kt.com",
    "": "wanyong.kim@kt.com",
    "": "82107343@kt.com",
    "": "82077078@kt.com",
    "": "82096643@kt.com",
    "": "kwangho12.kim@kt.com",
    "": "sang_tae.kim@kt.com",
    "": "sys7779.son@kt.com",
    "": "82224305@kt.com",
    "": "82077070@kt.com",
    "": "82277431@kt.com",
    "": "sung.rok@kt.com",
    "": "hyungkeun.cho@kt.com"
    
}

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

def read_file_with_encoding(file_path, header_row, **kwargs):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{file_path}   .")
    encoding = detect_encoding(file_path)
    return pd.read_csv(file_path, encoding=encoding, header=header_row, **kwargs)

def format_numeric(value):
    if pd.isna(value):
        return ''
    if isinstance(value, (int, float)):
        return f"{value:,.0f}"
    return value

def send_emails(selected_agents, email_addresses, email_body_template, sender_email, sender_password, progress, root, font_size):
    smtp_server = "smtp.naver.com"
    smtp_port = 587

    header_fill = PatternFill(start_color="4F4F4F", end_color="4F4F4F", fill_type="solid")
    header_font = Font(color="FFFFFF", bold=True)
    thin_border = Border(left=Side(style='thin', color='000000'),
                         right=Side(style='thin', color='000000'),
                         top=Side(style='thin', color='000000'),
                         bottom=Side(style='thin', color='000000'))

    summary_path = 'D:\\\\\\6.11\\ _0613.cs'
    new_detail_path = 'D:\\\\\\6.11\\ .cs'
   
    header_row = 0

    df_summary = read_file_with_encoding(summary_path, header_row)
    df_new_detail = read_file_with_encoding(new_detail_path, header_row)

    df_summary = df_summary.loc[:, ~df_summary.columns.str.contains('Unnamed')]
    df_summary.columns = df_summary.columns.str.strip().str.replace('\n', '')

    df_new_detail.columns = df_new_detail.columns.str.strip()

    df_summary = df_summary.fillna('')
    df_new_detail = df_new_detail.fillna('')

    if '' in df_summary.columns:
        df_summary[''] = pd.to_datetime(df_summary[''], errors='coerce').dt.strftime('%Y-%m-%d').fillna('')

    df_summary = df_summary.applymap(format_numeric)
    df_new_detail = df_new_detail.applymap(format_numeric)

    def format_row(row):
        if row[''] != '':
            return ['color: red;'] * len(row)
        return [''] * len(row)

    def remove_empty_columns(df):
        return df.dropna(axis=1, how='all')

    missing_emails = []
    total_agents = len(selected_agents)
   
    for idx, agent in enumerate(selected_agents):
        agent_email = email_addresses.get(agent)
        if not agent_email or agent_email == "#N/A":
            missing_emails.append(agent)
            continue

        summary_filtered = df_summary[df_summary[''] == agent].copy()
        new_detail_filtered = df_new_detail[df_new_detail[''] == agent].copy()

        if summary_filtered.empty or new_detail_filtered.empty:
            print(f"{agent}   .")
            continue

        summary_filtered[''] = ''
        summary_filtered['  '] = ''
       
        summary_filtered = remove_empty_columns(summary_filtered)
        new_detail_filtered = remove_empty_columns(new_detail_filtered)

        summary_html = summary_filtered.style.apply(format_row, axis=1).set_table_styles(
            [{'selector': 'th', 'props': [('background-color', '#4F4F4F'), ('color', 'white'), ('font-weight', 'bold')]}]
        ).hide_index().render()

        new_detail_html = new_detail_filtered.style.set_table_styles(
            [{'selector': 'th', 'props': [('background-color', '#4F4F4F'), ('color', 'white'), ('font-weight', 'bold')]}]
        ).hide_index().render()

        excel_output_path = f"D:\\\\\\{agent}_.xlsx"
       
        with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
            summary_filtered.to_excel(writer, sheet_name="", index=False)
            new_detail_filtered.to_excel(writer, sheet_name="", index=False)
       
        wb = openpyxl.load_workbook(excel_output_path)
       
        for sheet in wb.sheetnames:
            ws = wb[sheet]
            for cell in ws[1]:
                cell.fill = header_fill
                cell.font = header_font
                cell.border = thin_border
       
        wb.save(excel_output_path)
       
        email_body = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <style>
                table {{
                    width: 100%;
                    border-collapse: collapse;
                }}
                th, td {{
                    border: 1px solid black;
                    padding: 5px;
                    text-align: left;
                    font-size: {font_size}px;
                }}
                th {{
                    background-color: #4F4F4F;
                    color: white;
                    font-weight: bold;
                }}
                td {{
                    height: 20px;
                }}
            </style>
        </head>
        <body>
            <p style="font-size:{font_size}px;">{agent}, .</p>
            <p style="font-size:{font_size}px;">  .</p>
            <p style="font-size:{font_size}px;">  (   )   </p>
            <p style="font-size:{font_size}px;">    , ,    .</p>
            <p style="font-size:{font_size}px;"> (620 )     .</p>
            <div style="max-height: 400px; overflow-y: auto;">
                {summary_html}
                <br>
                {new_detail_html}
            </div>
            <br>
            <img src="cid:image1">
        </body>
        </html>
        """
       
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = agent_email
        msg['Subject'] = f"{agent}   "
        msg.attach(MIMEText(email_body, 'html', 'utf-8'))

        # Attach the excel file
        with open(excel_output_path, 'rb') as f:
            part = MIMEApplication(f.read(), Name="_.xlsx")
            part['Content-Disposition'] = 'attachment; filename="_.xlsx"'
            msg.attach(part)

        # Attach the resized image
        try:
            img_path = 'D:\\\\\\IMG_7573.jpeg'
            with Image.open(img_path) as img:
                img = img.resize((img.width // 4, img.height // 4))  # Resize image to 1/4 size
                img.save('D:\\\\\\resized_image.jpeg')

            with open('D:\\\\\\resized_image.jpeg', 'rb') as img:
                mime_image = MIMEImage(img.read())
                mime_image.add_header('Content-ID', '<image1>')
                msg.attach(mime_image)
        except FileNotFoundError:
            print("    .    .")
       
        try:
            with smtplib.SMTP(smtp_server, smtp_port) as server:
                server.starttls()
                time.sleep(1)  #    
                server.login(sender_email, sender_password)
                time.sleep(1)  #    
                server.sendmail(sender_email, agent_email, msg.as_string())
                print(f"{agent}   .")
        except smtplib.SMTPAuthenticationError as e:
            print(f"   .    : {e}")
            break  #     
        except Exception as e:
            print(f"{agent}   : {e}")

        # Update progress bar
        progress['value'] = ((idx + 1) / total_agents) * 100
        root.update_idletasks()

    if missing_emails:
        print("     :")
        for agent in missing_emails:
            print(f" - {agent}")

    # Close the root window after email sending is complete
    root.quit()

def start_sending(selected_agents, email_addresses, email_body_template, sender_email, sender_password, root, font_size):
    progress_window = Toplevel(root)
    progress_window.title("Progress")
    tk.Label(progress_window, text="Sending emails, please wait...").pack(pady=10)
    progress = Progressbar(progress_window, orient=tk.HORIZONTAL, length=300, mode='determinate')
    progress.pack(pady=20)
   
    # Start sending emails in a separate thread
    threading.Thread(target=send_emails, args=(selected_agents, email_addresses, email_body_template, sender_email, sender_password, progress, root, font_size)).start()

def main():
    root = tk.Tk()
    root.withdraw()

    sender_email = "bough38@naver.com"
    sender_password = simpledialog.askstring("Input", "  :", show='*')

    agents = [agent[""] for agent in agent_data]

    def add_agent():
        new_window = Toplevel(root)
        new_window.title("   ")

        def save_agent():
            agent_info = {
                "": entry_office.get(),
                "": entry_zone.get(),
                "": entry_id.get(),
                "": entry_name.get(),
                "1": entry_id1.get(),
                " SP": entry_sp.get(),
                "": entry_email.get()
            }
            agent_data.append(agent_info)
            agents.append(agent_info[""])
            email_addresses[agent_info[""]] = agent_info[""]
            checklist.insert(tk.END, agent_info[""])
            new_window.destroy()
            update_selected_agents()

        Label(new_window, text="").grid(row=0, column=0)
        entry_office = Entry(new_window)
        entry_office.grid(row=0, column=1)

        Label(new_window, text="").grid(row=1, column=0)
        entry_zone = Entry(new_window)
        entry_zone.grid(row=1, column=1)

        Label(new_window, text="").grid(row=2, column=0)
        entry_id = Entry(new_window)
        entry_id.grid(row=2, column=1)

        Label(new_window, text="").grid(row=3, column=0)
        entry_name = Entry(new_window)
        entry_name.grid(row=3, column=1)

        Label(new_window, text="1").grid(row=4, column=0)
        entry_id1 = Entry(new_window)
        entry_id1.grid(row=4, column=1)

        Label(new_window, text=" SP").grid(row=5, column=0)
        entry_sp = Entry(new_window)
        entry_sp.grid(row=5, column=1)

        Label(new_window, text="").grid(row=6, column=0)
        entry_email = Entry(new_window)
        entry_email.grid(row=6, column=1)

        Button(new_window, text="", command=save_agent).grid(row=7, column=0, columnspan=2)

    def edit_agent():
        selected_indices = checklist.curselection()
        if not selected_indices:
            messagebox.showwarning("Warning", "  .")
            return
        index = selected_indices[0]
        agent_name = checklist.get(index)
        agent_info = next(agent for agent in agent_data if agent[""] == agent_name)

        edit_window = Toplevel(root)
        edit_window.title("   ")

        def save_changes():
            agent_info[""] = entry_office.get()
            agent_info[""] = entry_zone.get()
            agent_info[""] = entry_id.get()
            agent_info[""] = entry_name.get()
            agent_info["1"] = entry_id1.get()
            agent_info[" SP"] = entry_sp.get()
            agent_info[""] = entry_email.get()

            agents[index] = agent_info[""]
            email_addresses[agent_info[""]] = agent_info[""]
            checklist.delete(index)
            checklist.insert(index, agent_info[""])
            edit_window.destroy()
            update_selected_agents()

        Label(edit_window, text="").grid(row=0, column=0)
        entry_office = Entry(edit_window)
        entry_office.insert(0, agent_info[""])
        entry_office.grid(row=0, column=1)

        Label(edit_window, text="").grid(row=1, column=0)
        entry_zone = Entry(edit_window)
        entry_zone.insert(0, agent_info[""])
        entry_zone.grid(row=1, column=1)

        Label(edit_window, text="").grid(row=2, column=0)
        entry_id = Entry(edit_window)
        entry_id.insert(0, agent_info[""])
        entry_id.grid(row=2, column=1)

        Label(edit_window, text="").grid(row=3, column=0)
        entry_name = Entry(edit_window)
        entry_name.insert(0, agent_info[""])
        entry_name.grid(row=3, column=1)

        Label(edit_window, text="1").grid(row=4, column=0)
        entry_id1 = Entry(edit_window)
        entry_id1.insert(0, agent_info["1"])
        entry_id1.grid(row=4, column=1)

        Label(edit_window, text=" SP").grid(row=5, column=0)
        entry_sp = Entry(edit_window)
        entry_sp.insert(0, agent_info[" SP"])
        entry_sp.grid(row=5, column=1)

        Label(edit_window, text="").grid(row=6, column=0)
        entry_email = Entry(edit_window)
        entry_email.insert(0, agent_info[""])
        entry_email.grid(row=6, column=1)

        Button(edit_window, text="", command=save_changes).grid(row=7, column=0, columnspan=2)

    def delete_agent():
        selected_indices = checklist.curselection()
        for index in reversed(selected_indices):
            agent_name = checklist.get(index)
            agents.remove(agent_name)
            email_addresses.pop(agent_name, None)
            agent_data[:] = [agent for agent in agent_data if agent[""] != agent_name]
            checklist.delete(index)
        update_selected_agents()

    def update_selected_agents():
        selected_agents_listbox.delete(0, tk.END)
        selected_indices = checklist.curselection()
        for i in selected_indices:
            selected_agents_listbox.insert(tk.END, f"{checklist.get(i)} ({email_addresses.get(checklist.get(i), 'N/A')})")

    def select_all():
        checklist.select_set(0, tk.END)
        update_selected_agents()

    def deselect_all():
        checklist.select_clear(0, tk.END)
        update_selected_agents()

    def send_selected():
        selected_indices = checklist.curselection()
        selected_agents = [agents[i] for i in selected_indices]
        if selected_agents:
            email_body_template = email_body_var.get()
            font_size = font_size_var.get()
            start_sending(selected_agents, email_addresses, email_body_template, sender_email, sender_password, root, font_size)
        else:
            messagebox.showwarning("Warning", "    .")

    def preview_email():
        selected_indices = checklist.curselection()
        selected_agents = [agents[i] for i in selected_indices]
        if selected_agents:
            email_body_template = email_body_var.get()
            font_size = font_size_var.get()
            preview_window = Toplevel(root)
            preview_window.title("Email Preview")
            email_preview = tk.Text(preview_window, wrap=WORD)
            email_preview.pack(expand=True, fill=tk.BOTH)
            for agent in selected_agents:
                email_body = email_body_template.replace("{agent}", agent).replace("{font_size}", str(font_size))
                email_preview.insert(tk.END, f"To: {email_addresses.get(agent, 'N/A')}\n\n{email_body}\n\n{'-'*50}\n\n")
        else:
            messagebox.showwarning("Warning", "    .")

    root.deiconify()
    root.title("Select Agents to Send Emails")
    root.geometry("1000x600")
   
    left_frame = tk.Frame(root)
    left_frame.pack(side=tk.LEFT, padx=10, pady=10, fill=tk.BOTH, expand=True)
   
    right_frame = tk.Frame(root)
    right_frame.pack(side=tk.RIGHT, padx=10, pady=10, fill=tk.BOTH, expand=True)
   
    tk.Label(left_frame, text="   :", font=("Helvetica", 12)).pack(pady=10)
   
    left_inner_frame = tk.Frame(left_frame)
    left_inner_frame.pack(pady=10, fill=tk.BOTH, expand=True)

    scrollbar = Scrollbar(left_inner_frame, orient=tk.VERTICAL)
    checklist = Listbox(left_inner_frame, selectmode=MULTIPLE, yscrollcommand=scrollbar.set, font=("Helvetica", 10))
    scrollbar.config(command=checklist.yview)
    scrollbar.pack(side=RIGHT, fill=Y)
    checklist.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
   
    for agent in agents:
        checklist.insert(tk.END, agent)

    button_frame = tk.Frame(left_frame)
    button_frame.pack(pady=5)
    tk.Button(button_frame, text=" ", command=select_all, font=("Helvetica", 10)).pack(side=tk.LEFT, padx=5)
    tk.Button(button_frame, text=" ", command=deselect_all, font=("Helvetica", 10)).pack(side=tk.LEFT, padx=5)

    tk.Button(left_frame, text="", command=add_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=edit_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=delete_agent, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=send_selected, font=("Helvetica", 10)).pack(pady=5)
    tk.Button(left_frame, text="", command=preview_email, font=("Helvetica", 10)).pack(pady=5)

    tk.Label(right_frame, text="   :", font=("Helvetica", 12)).pack(pady=10)
   
    selected_agents_listbox = Listbox(right_frame, font=("Helvetica", 10))
    selected_agents_listbox.pack(fill=tk.BOTH, expand=True)
   
    tk.Label(right_frame, text=" :", font=("Helvetica", 12)).pack(pady=10)
   
    email_body_var = StringVar()
    email_body_var.set("""<!DOCTYPE html>
<html>
<body>
    <p style="font-size:{font_size}px;">{agent}, .</p>
    <p style="font-size:{font_size}px;">  .</p>
    <p style="font-size:{font_size}px;">  (   )   </p>
    <p style="font-size:{font_size}px;">    , ,    .</p>
    <p style="font-size:{font_size}px;"> (620 )     .</p>
</body>
</html>
""")
    email_body_entry = tk.Entry(right_frame, textvariable=email_body_var, font=("Helvetica", 10), width=50)
    email_body_entry.pack(pady=5)

    tk.Label(right_frame, text=" :", font=("Helvetica", 12)).pack(pady=10)
    font_size_var = tk.IntVar(value=12)
    font_size_scale = Scale(right_frame, from_=8, to=20, orient=tk.HORIZONTAL, variable=font_size_var)
    font_size_scale.pack(pady=5)

    checklist.bind("<<ListboxSelect>>", lambda event: update_selected_agents())

    # Exit after 10 minutes (600000 ms)
    root.after(600000, lambda: root.quit())
    root.mainloop()

if __name__ == "__main__":
    main()
```
