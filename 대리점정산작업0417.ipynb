```python
import pandas as pd

# 파일 경로 설정
xlsx_path = 'D:\\대리점자동화연습\\2024년 3월 대리점.xlsx'
csv_path = 'D:\\대리점자동화연습\\2024년 3월 대리점.csv'

# 관심 있는 상호명
companies = ["(주)미레넷", "구정례", "케이티텔보안네트워크", "애드라인", "아이티인프라"]

# Excel 파일 로드
with pd.ExcelFile(xlsx_path) as xls:
    # 각 시트를 DataFrame으로 읽기
    summary_df = pd.read_excel(xls, '표(요약)', skiprows=7)  # 8번 행부터 읽기
    new_detail_df = pd.read_excel(xls, '신규상세')
    termination_detail_df = pd.read_excel(xls, '해지상세')

    # 필터링 조건 적용
    summary_df = summary_df[summary_df['G'].isin(companies)]
    new_detail_df = new_detail_df[new_detail_df['유통망대분류'].isin(companies)]
    termination_detail_df = termination_detail_df[termination_detail_df['유통망대분류'].isin(companies)]

# 필터링된 데이터를 CSV 파일로 저장
with pd.ExcelWriter(csv_path, mode='w', engine='openpyxl') as writer:
    summary_df.to_excel(writer, index=False, sheet_name='표(요약)')
    new_detail_df.to_excel(writer, index=False, sheet_name='신규상세')
    termination_detail_df.to_excel(writer, index=False, sheet_name='해지상세')

# 다시 xlsx 파일로 저장
with pd.ExcelWriter(xlsx_path, mode='w', engine='openpyxl') as writer:
    summary_df.to_excel(writer, index=False, sheet_name='표(요약)')
    new_detail_df.to_excel(writer, index=False, sheet_name='신규상세')
    termination_detail_df.to_excel(writer, index=False, sheet_name='해지상세')
```
