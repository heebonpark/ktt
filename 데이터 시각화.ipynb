```python
Sub CreatePivotTable()
    Dim wb As Workbook
    Dim wsSource As Worksheet
    Dim wsPivot As Worksheet
    Dim pt As PivotTable
    Dim ptField As PivotField
    Dim lastRow As Long
   
    ' 엑셀 파일 열기
    Set wb = Workbooks.Open("D:\G000_G000_00086.csv")
    ' 원본 데이터 시트 선택
    Set wsSource = wb.Sheets(1)
   
    ' 새 시트 생성 및 이름 설정
    Set wsPivot = wb.Sheets.Add
    wsPivot.Name = "PivotTable"
   
    ' 데이터 전체 범위 찾기
    lastRow = wsSource.Cells(wsSource.Rows.Count, "A").End(xlUp).Row
   
    ' 강원본부를 강북/강원본부로 변경
    For i = 2 To lastRow
        If wsSource.Cells(i, "A").Value = "강원본부" Then
            wsSource.Cells(i, "A").Value = "강북/강원본부"
        End If
    Next i
   
    ' 피벗 테이블 생성
    Set pt = wb.PivotTableWizard(TableDestination:=wsPivot.Cells(3, 1), SourceType:=xlDatabase, SourceData:=wsSource.UsedRange)
   
    ' 피벗 필드 설정
    With pt
        ' 행 필드 설정
        .PivotFields("관리지사명").Orientation = xlRowField
        .PivotFields("관리지사명").AutoSort xlDescending, "중앙지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "강북지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "서대문지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "고양지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "의정부지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "남양주지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "강릉지사"
        .PivotFields("관리지사명").AutoSort xlDescending, "원주지사"
       
        ' 열 필드 설정
        .PivotFields("계약상태(대)").Orientation = xlColumnField
       
        ' 값 필드 설정
        With .PivotFields("계약번호")
            .Orientation = xlDataField
            .Function = xlCount
            .NumberFormat = "0"
            .Name = "계약번호 (개수)"
        End With
       
        ' 필터 설정
        .PivotFields("서비스(대)").PivotFilters.Add Type:=xlCaptionEquals, Value1:="기본서비스"
    End With
   
    ' 피벗 테이블 레이아웃 조정
    wsPivot.Cells(3, 1).Select
   
    ' 피벗 테이블 이름 변경
    wsPivot.PivotTables(1).Name = "Sales_PivotTable"
   
    ' 엑셀 파일 저장
    wb.SaveAs "D:\청약_0222.xlsx"
   
    ' 메시지 박스 표시
    MsgBox "청약_0222 파일로 저장되었습니다.", vbInformation
   
    ' 파일 닫기
    wb.Close
End Sub
```

```python
import pandas as pd

# 엑셀 파일 열기
file_path = r"D:\본파이썬\G000_G000_00086.csv"
data = pd.read_csv(file_path, encoding='cp949')

# 강원본부를 강북/강원본부로 변경
data['관리본부명'] = data['관리본부명'].replace('강원본부', '강북/강원본부')

# 필터 조건을 적용한 데이터프레임 생성
filtered_data = data[data['서비스(대)'] == '기본서비스']

# 새로운 시트로 피벗을 만들어줘
pivot_data = pd.pivot_table(filtered_data, index='관리지사명', columns='계약상태(대)', values='계약번호', aggfunc='count')

# 조건에 맞게 행 순서 변경
pivot_data = pivot_data.reindex(["중앙지사", "강북지사", "서대문지사", "고양지사", "의정부지사", "남양주지사", "강릉지사", "원주지사"])

# 합계 행과 열 추가
pivot_data.loc['합계'] = pivot_data.sum()
pivot_data['합계'] = pivot_data.sum(axis=1)

# 엑셀 파일로 저장
output_file_path = r"D:\본파이썬\청약_0222.xlsx"
with pd.ExcelWriter(output_file_path) as writer:
    pivot_data.to_excel(writer, sheet_name='청약_0222', index=True)

# 필터 조건 설정
workbook = writer.book
worksheet = writer.sheets['청약_0222']
worksheet.autofilter('A1:ZZ1')  # 필터 설정
worksheet.filter_column('J', '기본서비스')  # 서비스(대) 필터 설정

print("청약_0222 파일로 저장되었습니다.")
```

```python
import os
import openpyxl

# 입력 폴더 경로와 취합시트 파일명 설정
input_folder = r"D:\ㅁ 리텐션실적\2024년 2월 폴더"
output_file = r"D:\ㅁ 리텐션실적\★리텐션 P값 일괄등록 참고양식_로그인필요_강북강원본부 승인요청건 관리_20240129.xlsm"
sheet_name = "취합시트"

# 취합시트 열 이름 설정 (예: A열에 복사)
target_column = "A"

# 입력 폴더 내의 엑셀 파일들을 순회하며 취합시트에 복사
for filename in os.listdir(input_folder):
    if filename.endswith(".xlsx"):
        filepath = os.path.join(input_folder, filename)
        wb = openpyxl.load_workbook(filepath, data_only=True)  # data_only=True로 설정하여 수식이 아닌 값만 읽어옴
        ws = wb.active
        for row in ws.iter_rows():
            for cell in row:
                ws[target_column + str(cell.row)].value = cell.value
        wb.save(output_file)

print("엑셀 파일 내용이 취합시트에 복사되었습니다.")
```
