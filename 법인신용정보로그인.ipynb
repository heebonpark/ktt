```python
import os
import subprocess

# AutoIt 경로 설정
autoit_path = r"C:\Program Files (x86)\AutoIt3\AutoIt3.exe"
aut2exe_path = r"C:\Program Files (x86)\AutoIt3\Aut2Exe\Aut2exe.exe"

# AutoIt 스크립트 및 EXE 파일 경로
script_path = r"C:\Users\Public\Documents\login_script.au3"
exe_path = r"C:\Users\Public\Documents\login_script.exe"
log_file_path = r"C:\Users\Public\Documents\log.txt"

# AutoIt 스크립트 내용
autoit_script_content = '''
Func LogMessage($message)
    FileWrite("C:\\Users\\Public\\Documents\\log.txt", $message & @CRLF)
    ConsoleWrite($message & @CRLF)
EndFunc

MouseClick("left", 25, 1419)
Sleep(1000)
MouseClick("left", 374, 862)
Sleep(1000)
'''

# 🔥 **1단계: AutoIt 스크립트 파일 생성**
try:
    with open(script_path, 'w', encoding='utf-8') as script_file:
        script_file.write(autoit_script_content)
    print(f"✅ AutoIt 스크립트가 {script_path}에 저장되었습니다.")
except Exception as e:
    print(f"❌ AutoIt 스크립트 파일 생성 실패: {e}")

# 🔥 **2단계: AutoIt 스크립트를 EXE로 변환**
try:
    if os.path.exists(aut2exe_path):
        result = subprocess.run(
            [
                aut2exe_path,
                '/in', script_path,
                '/out', exe_path,
                '/console'  # 최소화 실행
            ],
            capture_output=True,
            text=True,
            check=True
        )
        print("✅ AutoIt 스크립트가 EXE로 변환되었습니다.")
        print("출력 메시지:", result.stdout)
    else:
        print(f"❌ Aut2Exe 실행 파일을 찾을 수 없습니다: {aut2exe_path}")
except subprocess.CalledProcessError as e:
    print(f"❌ EXE 변환 실패: {e}")
    print(f"에러 메시지:", e.stderr)
except Exception as e:
    print(f"❌ 예외가 발생했습니다: {e}")

# 🔥 **3단계: EXE 파일 실행**
try:
    if os.path.exists(exe_path):
        subprocess.Popen(exe_path, creationflags=subprocess.CREATE_NO_WINDOW)
        print(f"✅ EXE 파일이 실행되었습니다: {exe_path}")
    else:
        print(f"❌ EXE 파일을 찾을 수 없습니다: {exe_path}")
except Exception as e:
    print(f"❌ EXE 실행 중 예외가 발생했습니다: {e}")
```
