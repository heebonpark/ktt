```python
import pandas as pd
import xlwings as xw
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import plotly.express as px
import socket

# 사용 가능한 포트를 찾는 함수
def find_free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind(('', 0))  # 시스템이 사용 가능한 포트를 자동 할당
        return s.getsockname()[1]

# 엑셀 데이터 읽기 함수
def get_excel_data():
    wb = xw.books.active
    sheet = wb.sheets.active
    data = sheet.range("A1:J53").options(pd.DataFrame, header=1, index=False).value
    return data

# 데이터 읽기 및 포맷팅
df = get_excel_data()

# 전체 합계 행 제거
df = df[~df["지사"].str.contains("전체", na=False)]

# 데이터 포맷팅
df["월정료"] = df["월정료"].apply(lambda x: f"{x:,.0f}" if pd.notnull(x) else x)
df["P.P 공헌이익금"] = df["P.P 공헌이익금"].apply(lambda x: f"{x:,.0f}" if pd.notnull(x) else x)

# 달성율 포맷 적용
df["달성율 연간 목표(가중치 70%)"] = df["달성율 연간 목표(가중치 70%)"].apply(lambda x: f"{x * 100:.2f}%" if pd.notnull(x) else "")
df["달성율 개선 목표(가중치 30%)"] = df["달성율 개선 목표(가중치 30%)"].apply(lambda x: f"{x * 100:.2f}%" if pd.notnull(x) else "")

# 아이콘 생성 함수
def create_icon(value):
    if pd.isnull(value):
        return ""
    value = float(value.replace("%", ""))
    if value >= 100:
        return "🟢"
    elif 90 <= value < 100:
        return "🟡"
    else:
        return "🔴"

# 값 + 아이콘 추가
df["연간 목표 (값 + 아이콘)"] = df["달성율 연간 목표(가중치 70%)"].apply(lambda x: f"{x} {create_icon(x)}")
df["개선 목표 (값 + 아이콘)"] = df["달성율 개선 목표(가중치 30%)"].apply(lambda x: f"{x} {create_icon(x)}")

# 소계 데이터 식별
is_summary = df["성명"].str.contains("소계", na=False)

# RANK 처리 및 색상 지정
df["RANK (264명)"] = pd.to_numeric(df["RANK (264명)"], errors="coerce")
df["RANK (264명)"] = df["RANK (264명)"].fillna(df["RANK (264명)"].max() + 1)
df["RANK (264명)"] = df["RANK (264명)"].astype(int)

def rank_color(rank):
    if rank <= 50:
        return "blue"
    elif rank <= 100:
        return "green"
    elif rank <= 200:
        return "yellow"
    else:
        return "red"

df["RANK 색상"] = df["RANK (264명)"].apply(rank_color)

# 소계 데이터 및 개별 데이터 분리
summary_df = df[is_summary].copy()
individual_df = df[~is_summary].copy()

# RANK 내림차순 정렬
individual_df = individual_df.sort_values(by="RANK (264명)", ascending=False)

# Dash 앱 초기화
app = Dash(__name__)

app.layout = html.Div([
    html.H1("지사별 및 성명별 데이터 대시보드", style={"textAlign": "center"}),

    # 체크박스
    html.Div([
        html.Label("시각화 선택:", style={"fontSize": "18px"}),
        dcc.Checklist(
            id="visualization-checklist",
            options=[
                {"label": "P.P 공헌이익금 및 월정료 분석", "value": "pp_analysis"},
                {"label": "RANK 순위 분석", "value": "rank_analysis"},
                {"label": "소계 데이터 분석", "value": "summary_analysis"},
                {"label": "EDA 다양한 시각화", "value": "eda_analysis"}
            ],
            value=["pp_analysis", "rank_analysis"],  # 기본 선택
            inline=True,
            style={"fontSize": "16px"}
        ),
    ], style={"marginBottom": "20px"}),

    # 필터 드롭다운
    html.Div([
        html.Label("지사 선택:", style={"fontSize": "18px"}),
        dcc.Dropdown(
            id="branch-filter",
            options=[
                {"label": branch, "value": branch}
                for branch in individual_df["지사"].unique()
            ],
            placeholder="지사를 선택하세요",
            style={"fontSize": "16px"}
        ),
    ], style={"width": "45%", "display": "inline-block"}),

    # 그래프 및 테이블
    dcc.Graph(id="pp-bar-chart", style={"display": "none"}),  # 기본 숨김 처리
    dcc.Graph(id="rank-bar-chart", style={"display": "none"}),  # 기본 숨김 처리
    dcc.Graph(id="eda-graph", style={"display": "none"}),  # EDA 그래프
    html.Div(id="data-table", style={"marginTop": "20px"})
])

@app.callback(
    [Output("pp-bar-chart", "figure"),
     Output("pp-bar-chart", "style"),
     Output("rank-bar-chart", "figure"),
     Output("rank-bar-chart", "style"),
     Output("eda-graph", "figure"),
     Output("eda-graph", "style"),
     Output("data-table", "children")],
    [Input("branch-filter", "value"), Input("visualization-checklist", "value")]
)
def update_dashboard(selected_branch, selected_visualizations):
    # 기본 데이터프레임 필터링
    filtered_df = individual_df.copy()
    if selected_branch:
        filtered_df = filtered_df[filtered_df["지사"] == selected_branch]

    # 기본 스타일
    pp_style, rank_style, eda_style = {"display": "none"}, {"display": "none"}, {"display": "none"}

    # P.P 공헌이익금 및 월정료 그래프 (누적 세로 막대)
    pp_fig = go.Figure()
    if "pp_analysis" in selected_visualizations:
        pp_fig.add_trace(go.Bar(
            x=filtered_df["성명"],
            y=filtered_df["P.P 공헌이익금"].str.replace(",", "").astype(float),
            name="P.P 공헌이익금",
            marker=dict(color="blue")
        ))
        pp_fig.add_trace(go.Bar(
            x=filtered_df["성명"],
            y=filtered_df["월정료"].str.replace(",", "").astype(float),
            name="월정료",
            marker=dict(color="red")
        ))
        pp_fig.update_layout(
            title="P.P 공헌이익금 및 월정료 분석 (누적)",
            xaxis_title="성명",
            yaxis_title="값",
            barmode="stack",
            legend_title="항목",
            title_font_size=20,
            title_x=0.5
        )
        pp_style = {"display": "block"}

    # RANK 순위 그래프
    rank_fig = go.Figure()
    if "rank_analysis" in selected_visualizations:
        rank_fig.add_trace(go.Bar(
            x=filtered_df["성명"],
            y=filtered_df["RANK (264명)"],
            name="RANK 순위",
            text=filtered_df["RANK (264명)"],
            textposition="auto",
            marker=dict(color=filtered_df["RANK 색상"])
        ))
        rank_fig.update_layout(
            title="RANK 순위 분석",
            xaxis_title="성명",
            yaxis_title="순위",
            legend_title="항목",
            title_font_size=20,
            title_x=0.5
        )
        rank_style = {"display": "block"}

    # EDA 시각화
    eda_fig = go.Figure()
    if "eda_analysis" in selected_visualizations:
        eda_fig = px.box(filtered_df, x="지사", y="P.P 공헌이익금",
                         color="지사", points="all", title="P.P 공헌이익금 분포")
        eda_style = {"display": "block"}

    # 데이터 테이블 생성
    table = html.Table(
        style={"borderCollapse": "collapse", "width": "100%", "fontSize": "14px", "textAlign": "center"},
        children=[
            html.Thead(
                html.Tr([
                    html.Th(col, style={"border": "1px solid gray", "padding": "8px", "backgroundColor": "#f2f2f2"})
                    for col in filtered_df[[
                        "지사", "성명", "월정료", "P.P 공헌이익금", "연간 목표 (값 + 아이콘)", "개선 목표 (값 + 아이콘)", "RANK (264명)"
                    ]].columns
                ])
            ),
            html.Tbody([
                html.Tr([
                    html.Td(filtered_df.iloc[i][col], style={"border": "1px solid gray", "padding": "8px"})
                    for col in filtered_df[[
                        "지사", "성명", "월정료", "P.P 공헌이익금", "연간 목표 (값 + 아이콘)", "개선 목표 (값 + 아이콘)", "RANK (264명)"
                    ]].columns
                ]) for i in range(len(filtered_df))
            ])
        ]
    )

    return pp_fig, pp_style, rank_fig, rank_style, eda_fig, eda_style, table

# 실행
if __name__ == "__main__":
    free_port = find_free_port()
    app.run_server(debug=True, port=free_port)
```
