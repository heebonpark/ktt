# 2단계(날짜형식변환)

```python
import pandas as pd
import glob
import os

# 폴더 경로 설정
folder_path = r'd:\시설\해지\0829'

# 모든 파일 경로 가져오기
all_files = glob.glob(os.path.join(folder_path, "*.xls"))

# 모든 파일을 읽어 데이터프레임으로 병합
all_data = pd.DataFrame()

# 각 파일의 데이터를 확인하면서 병합
for file in all_files:
    try:
        # HTML 형식으로 파일 읽기, dtype을 str로 지정하여 모든 데이터를 문자열로 읽기
        df_list = pd.read_html(file, header=0, converters={i: str for i in range(100)})
        df = df_list[0]  # 첫 번째 테이블만 사용

        # NaN 값은 공백으로 변환
        df.fillna('', inplace=True)

        print(f"{file} 데이터 미리보기:\n", df.head())  # 각 파일의 데이터 확인
        all_data = pd.concat([all_data, df], ignore_index=True)
    except Exception as e:
        print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")

# 병합된 데이터를 csv 파일로 저장 (CP949 인코딩 적용)
output_path = os.path.join(folder_path, "0829.cs")
if not all_data.empty:
    all_data.to_csv(output_path, index=False, encoding='cp949')
    print(f"병합된 파일이 {output_path}에 저장되었습니다.")
else:
    print("병합된 데이터가 비어 있습니다. 엑셀 파일들을 확인하세요.")

# 날짜 형식을 변환하는 로직
# 날짜 열을 정의하세요 (예: '실적최초등록일')
date_columns = ['실적최초등록일']  # 변환이 필요한 날짜 열 지정

# CSV 파일을 다시 읽고 날짜 형식을 변환하여 저장
df_final = pd.read_csv(output_path, encoding='cp949')

# 날짜 형식 변환
for col in date_columns:
    df_final[col] = pd.to_datetime(df_final[col], errors='coerce').dt.strftime('%Y-%m-%d')

# 변환된 날짜 형식으로 다시 저장
df_final.to_csv(output_path, index=False, encoding='cp949')
print("날짜 형식이 변환된 CSV 파일이 저장되었습니다.")
```

```python
import pandas as pd

# 엑셀 파일 경로 설정
input_file_path = 'd:/시설/해지/8월해지.xls'
output_html_path = 'd:/시설/해지/해지 조건추출 및 집계 결과0829.html'

# 필요한 컬럼 리스트 정의
columns_to_extract = [
    '관리본부명', '관리지사명', '계약번호', 'KTT월정료(조정)'
]

# 엑셀 파일의 Sheet1에서 데이터 읽기
df = pd.read_excel(input_file_path, sheet_name='Sheet1')

# 필요한 컬럼만 추출
df_extracted = df[columns_to_extract]

# KTT월정료(조정)을 천 단위 콤마로 포맷하고 숫자로 변환
df_extracted['KTT월정료(조정)'] = df_extracted['KTT월정료(조정)'].apply(lambda x: f"{x:,.0f}")

# 관리본부명과 관리지사명 기준으로 계약번호 건수 및 KTT월정료(조정) 합계 계산
grouped = df_extracted.groupby(['관리본부명', '관리지사명']).agg(
    계약번호_건수=('계약번호', 'count'),
    KTT월정료_합계=('KTT월정료(조정)', lambda x: f"{sum(int(v.replace(',', '')) for v in x):,}")
).reset_index()

# 하단에 합계 행 추가
total_row = pd.DataFrame({
    '관리본부명': ['합계'],
    '관리지사명': [''],
    '계약번호_건수': [grouped['계약번호_건수'].sum()],
    'KTT월정료_합계': [f"{grouped['KTT월정료_합계'].str.replace(',', '').astype(int).sum():,}"]
})
grouped = pd.concat([grouped, total_row], ignore_index=True)

# HTML 파일로 저장
html_content = grouped.to_html(index=False)

# EDA를 위해 간단한 시각화 추가
import matplotlib.pyplot as plt
import seaborn as sns

# 계약번호 건수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='계약번호_건수', y='관리본부명', data=grouped[:-1])  # 마지막 행(합계)을 제외하고 시각화
plt.title('관리본부명별 계약번호 건수')
plt.tight_layout()

# 그래프를 이미지로 저장
plt.savefig('d:/시설/해지/contract_count.png')
plt.close()

# KTT월정료 합계 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='KTT월정료_합계', y='관리본부명', data=grouped[:-1])  # 마지막 행(합계)을 제외하고 시각화
plt.title('관리본부명별 KTT월정료 합계')
plt.tight_layout()

# 그래프를 이미지로 저장
plt.savefig('d:/시설/해지/ktt_fee_sum.png')
plt.close()

# HTML에 시각화 이미지 추가
html_content += '''
<h2>EDA 시각화</h2>
<img src="contract_count.png" alt="계약번호 건수 시각화">
<img src="ktt_fee_sum.png" alt="KTT월정료 합계 시각화">
'''

# 최종 HTML 파일로 저장
with open(output_html_path, 'w', encoding='utf-8') as f:
    f.write(html_content)

print("결과가 HTML 파일로 저장되었습니다.")
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk
import matplotlib.pyplot as plt

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("데이터 필터링 및 시각화")
       
        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")

        self.create_widgets()
        self.configure_grid()
   
    def create_widgets(self):
        tk.Label(self.root, text="파일 경로:").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="파일 선택", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        self.column_listbox = tk.Listbox(self.root, selectmode='single', exportselection=0)
        self.column_listbox.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        self.condition_frame = tk.Frame(self.root)
        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        tk.Label(self.root, text="값으로 표시할 열 헤더 선택:").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[]).grid(row=3, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="시각화 유형 선택:").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")
       
        tk.Label(self.root, text="KTT월정료(조정) 단위:").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="저장 인코딩 선택:").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="조건 적용", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 추가", command=self.add_condition).grid(row=7, column=1, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 취소", command=self.clear_conditions).grid(row=7, column=2, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)
   
    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV and CS files", "*.csv;*.cs")])
        if file_path:
            self.file_path.set(file_path)
            try:
                self.df = pd.read_csv(self.file_path.get(), encoding=self.selected_encoding.get())
                self.df.fillna('', inplace=True)
                self.modify_columns()
                self.columns = self.df.columns.tolist()
                self.populate_columns()
            except Exception as e:
                messagebox.showerror("오류 발생", f"파일을 읽는 중 오류 발생: {e}")
   
    def modify_columns(self):
        """강원본부와 서부본부 이름 변경"""
        self.df['관리본부명'] = self.df['관리본부명'].replace({
            '강원본부': '강북/강원본부',
            '서부본부': '강남/서부본부'
        })
   
    def populate_columns(self):
        self.column_listbox.delete(0, tk.END)
        for col in self.columns:
            self.column_listbox.insert(tk.END, col)
       
        values_combobox = self.root.children["!combobox"]
        values_combobox['values'] = ["계약번호(행건수)", "ktt월정료(조정)"] + self.columns
        values_combobox.current(0)

        self.column_listbox.bind('<<ListboxSelect>>', self.populate_conditions)
   
    def populate_conditions(self, event):
        selected_column = self.column_listbox.get(self.column_listbox.curselection())
        if selected_column == '관리본부명':
            self.populate_gisa_condition()
        else:
            self.populate_generic_condition(selected_column)

        # 조건 프레임을 업데이트하여 새 조건을 반영
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

        for col, val_dict in self.conditions.items():
            condition_frame = tk.LabelFrame(self.condition_frame, text=col)
            condition_frame.grid(row=list(self.conditions.keys()).index(col), column=0, padx=5, pady=5, sticky='ew')
            for val, var in val_dict.items():
                chk = tk.Checkbutton(condition_frame, text=val, variable=var)
                chk.grid(sticky='w')

    def populate_generic_condition(self, col):
        if col in self.conditions:
            unique_values = self.df[col].unique().tolist()
            for val in unique_values:
                if val not in self.conditions[col]:
                    self.conditions[col][val] = tk.BooleanVar()

    def populate_gisa_condition(self):
        selected_bonbu = [val for val, var in self.conditions.get('관리본부명', {}).items() if var.get()]
        if '강북/강원본부' in selected_bonbu:
            gisa_values = ['중앙지사', '강북지사', '서대문지사', '고양지사', '의정부지사', '남양주지사', '강릉지사', '원주지사']
        elif '강남/서부본부' in selected_bonbu:
            gisa_values = ['서부지사', '서초지사', '관악지사', '강남지사', '송파지사']
        else:
            gisa_values = []

        self.conditions['관리지사'] = {}
        for gisa in gisa_values:
            self.conditions['관리지사'][gisa] = tk.BooleanVar()
   
    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()
                for col, val_dict in self.conditions.items():
                    selected_values = [val for val, var in val_dict.items() if var.get()]
                    if selected_values:
                        filtered_data = filtered_data[filtered_data[col].isin(selected_values)]
               
                # 관리본부명이 '강북/강원본부'인 경우에 따른 관리지사 정렬
                if '관리본부명' in filtered_data.columns and '관리지사' in filtered_data.columns:
                    if '강북/강원본부' in filtered_data['관리본부명'].unique():
                        sort_order = ['중앙지사', '강북지사', '서대문지사', '고양지사', '의정부지사', '남양주지사', '강릉지사', '원주지사']
                        filtered_data['관리지사'] = pd.Categorical(filtered_data['관리지사'], categories=sort_order, ordered=True)
                        filtered_data = filtered_data.sort_values(['관리본부명', '관리지사'])

                # 단위 설정
                unit = self.selected_unit.get()
                unit_factor = 1
                if unit == "천원":
                    unit_factor = 1000
                elif unit == "백만원":
                    unit_factor = 1000000
               
                # 값 표시 열 선택
                selected_values_column = self.selected_values_column.get()
                if selected_values_column == "계약번호(행건수)":
                    summary = filtered_data.groupby('관리본부명').agg(
                        계약번호_건수=('계약번호', 'count')
                    ).reset_index()
                elif selected_values_column == "ktt월정료(조정)":
                    summary = filtered_data.groupby('관리본부명').agg(
                        ktt월정료_합계=('ktt월정료(조정)', lambda x: pd.to_numeric(x, errors='coerce').sum() / unit_factor)
                    ).reset_index()
                    summary['ktt월정료_합계'] = summary['ktt월정료_합계'].apply(lambda x: "{:,.0f}".format(x) + " " + unit)
                else:
                    summary = filtered_data.groupby('관리본부명').agg(
                        계약번호_건수=('계약번호', 'count'),
                        ktt월정료_합계=('ktt월정료(조정)', lambda x: pd.to_numeric(x, errors='coerce').sum() / unit_factor)
                    ).reset_index()
                    summary['ktt월정료_합계'] = summary['ktt월정료_합계'].apply(lambda x: "{:,.0f}".format(x) + " " + unit)

                # 필터링된 결과를 CSV로 저장
                output_filtered_csv = self.file_path.get().replace(".csv", "_filtered_summary.csv").replace(".cs", "_filtered_summary.csv")
                summary.to_csv(output_filtered_csv, index=False, encoding=self.selected_encoding.get())
                messagebox.showinfo("저장 완료", f"필터링된 결과가 {output_filtered_csv}에 저장되었습니다.")
               
                # 시각화 선택 및 생성
                chart_type = self.selected_chart_type.get()
                plt.figure(figsize=(10, 6))
                if chart_type == "Bar":
                    plt.bar(summary['관리본부명'], summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), color='skyblue')
                elif chart_type == "Line":
                    plt.plot(summary['관리본부명'], summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), marker='o', linestyle='-', color='skyblue')
                elif chart_type == "Pie":
                    plt.pie(summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), labels=summary['관리본부명'], autopct='%1.1f%%', startangle=140)

                plt.ylabel(f'{selected_values_column} ({unit})')
                plt.xlabel('관리본부명')
                plt.title(f'관리본부명별 {selected_values_column} - {chart_type} 차트')
                plt.xticks(rotation=45)
                plt.tight_layout()
               
                # 시각화 결과를 이미지 파일로 저장
                chart_image_output_path = self.file_path.get().replace(".csv", f"_{chart_type}_chart.png").replace(".cs", f"_{chart_type}_chart.png")
                plt.savefig(chart_image_output_path)
                plt.show()
                messagebox.showinfo("시각화 완료", f"시각화된 이미지가 {chart_image_output_path}에 저장되었습니다.")
       
        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")

    def add_condition(self):
        # 조건을 추가할 수 있는 로직 구현 (예: 새로운 조건을 추가할 때 사용)
        pass

    def clear_conditions(self):
        self.conditions.clear()
        self.populate_columns()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk
import matplotlib.pyplot as plt

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("데이터 필터링 및 시각화")
       
        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")

        self.bonbu_gisa_map = {
            '강남/서부본부': ['강남지사', '강동지사', '강서지사', '관악지사', '부천지사', '분당지사', '수원지사', '안산지사', '용인지사', '인천지사', '안양지사','평택지사'],
            '강북/강원본부': ['중앙지사', '강북지사', '고양지사', '서대문지사', '의정부지사', '남양주지사', '원주지사', '강릉지사'],
            '대구/경북본부': ['대구지사', '동대구지사', '수성지사', '구미지사', '포항지사'],
            '부산/경남본부': ['동부산지사', '남부산지사', '서부산지사', '울산지사', '창원지사', '진주지사'],
            '전남/전북본부': ['광주지사', '북광주지사', '목포지사', '순천지사', '전주지사', '익산지사', '제주지사'],
            '충남/충북본부': ['대전지사', '서대전지사', '전주지사', '천안지사', '충남서부지사', '충북지사']
        }

        self.create_widgets()
        self.configure_grid()
   
    def create_widgets(self):
        tk.Label(self.root, text="파일 경로:").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="파일 선택", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        self.column_listbox = tk.Listbox(self.root, selectmode='single', exportselection=0)
        self.column_listbox.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        self.condition_frame = tk.Frame(self.root)
        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        tk.Label(self.root, text="값으로 표시할 열 헤더 선택:").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[]).grid(row=3, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="시각화 유형 선택:").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")
       
        tk.Label(self.root, text="KTT월정료(조정) 단위:").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="저장 인코딩 선택:").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="조건 적용", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 추가", command=self.add_condition).grid(row=7, column=1, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 취소", command=self.clear_conditions).grid(row=7, column=2, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)
   
    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV and CS files", "*.csv;*.cs")])
        if file_path:
            self.file_path.set(file_path)
            try:
                self.df = pd.read_csv(self.file_path.get(), encoding=self.selected_encoding.get())
                self.df.fillna('', inplace=True)
                self.modify_columns()
                self.columns = self.df.columns.tolist()
                self.populate_columns()
            except Exception as e:
                messagebox.showerror("오류 발생", f"파일을 읽는 중 오류 발생: {e}")
   
    def modify_columns(self):
        """강원본부와 서부본부 이름 변경"""
        self.df['관리본부명'] = self.df['관리본부명'].replace({
            '강원본부': '강북/강원본부',
            '서부본부': '강남/서부본부'
        })
   
    def populate_columns(self):
        self.column_listbox.delete(0, tk.END)
        for col in self.columns:
            self.column_listbox.insert(tk.END, col)
       
        values_combobox = self.root.children["!combobox"]
        values_combobox['values'] = ["계약번호(행건수)", "ktt월정료(조정)"] + self.columns
        values_combobox.current(0)

        self.column_listbox.bind('<<ListboxSelect>>', self.populate_conditions)
   
    def populate_conditions(self, event):
        selected_column = self.column_listbox.get(self.column_listbox.curselection())
        if selected_column == '관리본부명':
            self.populate_bonbu_condition(selected_column)
        else:
            self.populate_generic_condition(selected_column)

        # 조건 프레임을 업데이트하여 새 조건을 반영
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

        for col, val_dict in self.conditions.items():
            condition_frame = tk.LabelFrame(self.condition_frame, text=f"조건: {col}")
            condition_frame.grid(row=list(self.conditions.keys()).index(col), column=0, padx=5, pady=5, sticky='ew')
            for val, var in val_dict.items():
                chk = tk.Checkbutton(condition_frame, text=val, variable=var)
                chk.grid(sticky='w')

    def populate_generic_condition(self, col):
        if col in self.conditions:
            unique_values = self.df[col].unique().tolist()
            for val in unique_values:
                if val not in self.conditions[col]:
                    self.conditions[col][val] = tk.BooleanVar()

    def populate_bonbu_condition(self, col):
        bonbu_list = self.df[col].unique().tolist()
        if col not in self.conditions:
            self.conditions[col] = {}
        for bonbu in bonbu_list:
            if bonbu not in self.conditions[col]:
                self.conditions[col][bonbu] = tk.BooleanVar()

    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()
                for col, val_dict in self.conditions.items():
                    selected_values = [val for val, var in val_dict.items() if var.get()]
                    if selected_values:
                        filtered_data = filtered_data[filtered_data[col].isin(selected_values)]
               
                # 관리본부명에 따른 관리지사 정렬
                if '관리본부명' in filtered_data.columns and '관리지사' in filtered_data.columns:
                    for bonbu, gisa_list in self.bonbu_gisa_map.items():
                        if bonbu in filtered_data['관리본부명'].unique():
                            filtered_data.loc[filtered_data['관리본부명'] == bonbu, '관리지사'] = pd.Categorical(
                                filtered_data.loc[filtered_data['관리본부명'] == bonbu, '관리지사'],
                                categories=gisa_list,
                                ordered=True
                            )
                    filtered_data = filtered_data.sort_values(['관리본부명', '관리지사'])

                # 단위 설정
                unit = self.selected_unit.get()
                unit_factor = 1
                if unit == "천원":
                    unit_factor = 1000
                elif unit == "백만원":
                    unit_factor = 1000000
               
                # 값 표시 열 선택
                selected_values_column = self.selected_values_column.get()
                if selected_values_column == "계약번호(행건수)":
                    summary = filtered_data.groupby('관리본부명').agg(
                        계약번호_건수=('계약번호', 'count')
                    ).reset_index()
                elif selected_values_column == "ktt월정료(조정)":
                    summary = filtered_data.groupby('관리본부명').agg(
                        ktt월정료_합계=('ktt월정료(조정)', lambda x: pd.to_numeric(x, errors='coerce').sum() / unit_factor)
                    ).reset_index()
                    summary['ktt월정료_합계'] = summary['ktt월정료_합계'].apply(lambda x: "{:,.0f}".format(x) + " " + unit)
                else:
                    summary = filtered_data.groupby('관리본부명').agg(
                        계약번호_건수=('계약번호', 'count'),
                        ktt월정료_합계=('ktt월정료(조정)', lambda x: pd.to_numeric(x, errors='coerce').sum() / unit_factor)
                    ).reset_index()
                    summary['ktt월정료_합계'] = summary['ktt월정료_합계'].apply(lambda x: "{:,.0f}".format(x) + " " + unit)

                # 필터링된 결과를 CSV로 저장
                output_filtered_csv = self.file_path.get().replace(".csv", "_filtered_summary.csv").replace(".cs", "_filtered_summary.csv")
                summary.to_csv(output_filtered_csv, index=False, encoding=self.selected_encoding.get())
                messagebox.showinfo("저장 완료", f"필터링된 결과가 {output_filtered_csv}에 저장되었습니다.")
               
                # 시각화 선택 및 생성
                chart_type = self.selected_chart_type.get()
                plt.figure(figsize=(10, 6))
                if chart_type == "Bar":
                    plt.bar(summary['관리본부명'], summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), color='skyblue')
                elif chart_type == "Line":
                    plt.plot(summary['관리본부명'], summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), marker='o', linestyle='-', color='skyblue')
                elif chart_type == "Pie":
                    plt.pie(summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), labels=summary['관리본부명'], autopct='%1.1f%%', startangle=140)

                plt.ylabel(f'{selected_values_column} ({unit})')
                plt.xlabel('관리본부명')
                plt.title(f'관리본부명별 {selected_values_column} - {chart_type} 차트')
                plt.xticks(rotation=45)
                plt.tight_layout()
               
                # 시각화 결과를 이미지 파일로 저장
                chart_image_output_path = self.file_path.get().replace(".csv", f"_{chart_type}_chart.png").replace(".cs", f"_{chart_type}_chart.png")
                plt.savefig(chart_image_output_path)
                plt.show()
                messagebox.showinfo("시각화 완료", f"시각화된 이미지가 {chart_image_output_path}에 저장되었습니다.")
       
        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")

    def add_condition(self):
        # 조건을 추가할 수 있는 로직 구현 (예: 새로운 조건을 추가할 때 사용)
        pass

    def clear_conditions(self):
        self.conditions.clear()
        self.populate_columns()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk
import matplotlib.pyplot as plt
import os

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Data Filtering & Visualization (데이터 필터링 및 시각화)")

        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")

        # 마지막 업로드된 파일 자동으로 불러오기
        self.last_uploaded_file = "last_uploaded_file.txt"
        if os.path.exists(self.last_uploaded_file):
            with open(self.last_uploaded_file, "r") as f:
                last_file = f.read().strip()
                if os.path.exists(last_file):
                    self.file_path.set(last_file)
                    self.load_file(auto_load=True)

        self.bonbu_gisa_map = {
            '강남/서부본부': ['강남지사', '강동지사', '강서지사', '관악지사', '부천지사', '분당지사', '수원지사', '안산지사', '용인지사', '인천지사', '안양지사', '평택지사'],
            '강북/강원본부': ['중앙지사', '강북지사', '고양지사', '서대문지사', '의정부지사', '남양주지사', '원주지사', '강릉지사'],
            '대구/경북본부': ['대구지사', '동대구지사', '수성지사', '구미지사', '포항지사'],
            '부산/경남본부': ['동부산지사', '남부산지사', '서부산지사', '울산지사', '창원지사', '진주지사'],
            '전남/전북본부': ['광주지사', '북광주지사', '목포지사', '순천지사', '전주지사', '익산지사', '제주지사'],
            '충남/충북본부': ['대전지사', '서대전지사', '전주지사', '천안지사', '충남서부지사', '충북지사']
        }

        self.create_widgets()
        self.configure_grid()
   
    def create_widgets(self):
        tk.Label(self.root, text="File Path (파일 경로):").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="Select File (파일 선택)", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        self.column_listbox = tk.Listbox(self.root, selectmode='single', exportselection=0)
        self.column_listbox.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        # 조건 영역을 가로로 배치하도록 수정
        self.condition_frame = tk.Frame(self.root)
        self.condition_canvas = tk.Canvas(self.condition_frame)
        self.scrollbar = tk.Scrollbar(self.condition_frame, orient="horizontal", command=self.condition_canvas.xview)  # 가로 스크롤바로 변경
        self.condition_frame_inner = tk.Frame(self.condition_canvas)

        self.condition_canvas.create_window((0, 0), window=self.condition_frame_inner, anchor="nw")
        self.condition_canvas.configure(xscrollcommand=self.scrollbar.set)

        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")
        self.scrollbar.pack(side="bottom", fill="x")
        self.condition_canvas.pack(side="top", fill="both", expand=True)

        self.condition_frame_inner.bind("<Configure>", lambda e: self.condition_canvas.configure(scrollregion=self.condition_canvas.bbox("all")))

        tk.Label(self.root, text="Select Column to Display Values (표시할 열 선택):").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[]).grid(row=3, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="Select Chart Type (차트 유형 선택):").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="KTT Monthly Charge Unit (KTT 월정료 단위):").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="Select Encoding (인코딩 선택):").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="Apply Conditions (조건 적용)", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="Add Condition (조건 추가)", command=self.add_condition).grid(row=7, column=1, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="Clear Conditions (조건 초기화)", command=self.clear_conditions).grid(row=7, column=2, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)

    def load_file(self, auto_load=False):
        if not auto_load:
            file_path = filedialog.askopenfilename(filetypes=[("CSV, CS, and Excel Files", "*.csv;*.cs;*.xls;*.xlsx")])
            if not file_path:
                return
            self.file_path.set(file_path)

        file_path = self.file_path.get()

        try:
            if file_path.endswith(".csv") or file_path.endswith(".cs"):
                self.df = self.load_csv_file(file_path)
            elif file_path.endswith(".xls") or file_path.endswith(".xlsx"):
                self.df = self.load_xls_file(file_path)

            # 마지막으로 불러온 파일 경로 저장
            with open(self.last_uploaded_file, "w") as f:
                f.write(file_path)

            self.df.fillna('', inplace=True)
            self.modify_columns()
            self.columns = self.df.columns.tolist()
            self.populate_columns()
            messagebox.showinfo("Success (성공)", "File loaded successfully (파일이 성공적으로 로드되었습니다).")

        except Exception as e:
            messagebox.showerror("Error (오류 발생)", f"Failed to read file: {e}. Please ensure the file format is correct. (파일을 읽는 데 실패했습니다. 파일 형식을 확인하세요.)")
            self.suggest_alternative_solutions(file_path, e)

    def load_csv_file(self, file_path):
        """CSV 또는 CS 파일을 로드하는 함수"""
        try:
            return pd.read_csv(file_path, encoding=self.selected_encoding.get())
        except UnicodeDecodeError:
            messagebox.showwarning("Encoding Error (인코딩 오류)", "Failed to load with selected encoding. Trying 'CP949'. (선택한 인코딩으로 파일을 로드하는 데 실패했습니다. 'CP949'로 시도합니다.)")
            return pd.read_csv(file_path, encoding='CP949')

    def load_xls_file(self, file_path):
        """XLS 또는 XLSX 파일을 로드하는 함수"""
        try:
            return pd.read_html(file_path)[0]
       
        except ValueError as e:
            messagebox.showwarning("File Format Error (파일 형식 오류)", "Failed to load Excel file as HTML. Please check if the file is a valid Excel file. (엑셀 파일을 HTML로 로드하는 데 실패했습니다. 파일 형식을 확인하세요.)")
            raise e

    def suggest_alternative_solutions(self, file_path, error):
        """파일 처리 실패 시 대안 제시 및 안내 제공"""
        if 'encoding' in str(error):
            messagebox.showinfo("Solution (해결책)", "Try changing the encoding to 'CP949' or check if the file contains non-UTF8 characters. (인코딩을 'CP949'로 변경하거나 파일에 비-UTF8 문자가 포함되어 있는지 확인하세요.)")
        elif 'file format' in str(error).lower():
            messagebox.showinfo("Solution (해결책)", f"The file '{file_path}' might be in an unsupported format. Please ensure it's a CSV, CS, or XLS file. (파일 '{file_path}'이 지원되지 않는 형식일 수 있습니다. CSV, CS 또는 XLS 파일인지 확인하세요.)")
        else:
            messagebox.showinfo("Solution (해결책)", "If the problem persists, ensure the file format and structure are correct. (문제가 지속되면 파일 형식과 구조가 올바른지 확인하세요.)")

    def modify_columns(self):
        """강원본부와 서부본부 이름 변경"""
        if '관리본부명' in self.df.columns:
            self.df['관리본부명'] = self.df['관리본부명'].replace({
                '강원본부': '강북/강원본부',
                '서부본부': '강남/서부본부'
            })
   
    def populate_columns(self):
        self.column_listbox.delete(0, tk.END)
        for col in self.columns:
            self.column_listbox.insert(tk.END, col)

        values_combobox = self.root.children["!combobox"]
        values_combobox['values'] = ["계약번호(행건수)", "ktt월정료(조정)"] + self.columns
        values_combobox.current(0)

        self.column_listbox.bind('<<ListboxSelect>>', self.populate_conditions)

    def populate_conditions(self, event):
        selected_column = self.column_listbox.get(self.column_listbox.curselection())
        if selected_column == '관리본부명':
            self.populate_bonbu_condition(selected_column)
        elif '월정료' in selected_column:
            self.populate_amount_condition(selected_column)
        elif selected_column == '계약번호':
            self.populate_count_condition(selected_column)
        else:
            self.populate_generic_condition(selected_column)

        for widget in self.condition_frame_inner.winfo_children():
            widget.destroy()

        # 조건을 가로로 배치하도록 수정
        for col, val_dict in self.conditions.items():
            condition_frame = tk.LabelFrame(self.condition_frame_inner, text=f"Condition: {col}")
            condition_frame.grid(row=0, column=list(self.conditions.keys()).index(col), padx=5, pady=5, sticky='nsew')

            for val, var in val_dict.items():
                if '월정료' in col:  # 금액 조건은 별도로 처리
                    entry = tk.Entry(condition_frame)
                    entry.insert(0, val)
                    entry.grid(sticky='w')
                    var.trace_add("write", lambda *args, var=var: var.set(entry.get()))
                else:
                    chk = tk.Checkbutton(condition_frame, text=val, variable=var)
                    chk.grid(sticky='w')

    def populate_generic_condition(self, col):
        if col not in self.conditions:
            self.conditions[col] = {}
        unique_values = self.df[col].unique().tolist()
        for val in unique_values:
            if val not in self.conditions[col]:
                self.conditions[col][val] = tk.BooleanVar()

    def populate_bonbu_condition(self, col):
        bonbu_list = self.df[col].unique().tolist()
        if col not in self.conditions:
            self.conditions[col] = {}
        for bonbu in bonbu_list:
            if bonbu not in self.conditions[col]:
                self.conditions[col][bonbu] = tk.BooleanVar()

    def populate_amount_condition(self, col):
        if col not in self.conditions:
            self.conditions[col] = {}
        self.conditions[col] = {
            '5만원 이상': tk.BooleanVar(),
            '5만원 이하': tk.BooleanVar(),
            '1원 이상': tk.BooleanVar()
        }

    def populate_count_condition(self, col):
        if col not in self.conditions:
            self.conditions[col] = {}
        self.conditions[col] = {
            '계약 건수 카운터': tk.BooleanVar()
        }

    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()

                for col, val_dict in self.conditions.items():
                    if col not in self.df.columns:
                        messagebox.showerror("Column Error", f"The column '{col}' does not exist in the file. (해당 파일에 '{col}' 열이 없습니다.)")
                        return

                    if '월정료' in col:  # 금액 필터링 적용
                        for condition, var in val_dict.items():
                            if var.get():
                                if condition == '5만원 이상':
                                    filtered_data = filtered_data[pd.to_numeric(filtered_data[col], errors='coerce') >= 50000]
                                elif condition == '5만원 이하':
                                    filtered_data = filtered_data[pd.to_numeric(filtered_data[col], errors='coerce') <= 50000]
                                elif condition == '1원 이상':
                                    filtered_data = filtered_data[pd.to_numeric(filtered_data[col], errors='coerce') > 0]
                    else:
                        selected_values = [val for val, var in val_dict.items() if var.get()]
                        if selected_values:
                            filtered_data = filtered_data[filtered_data[col].isin(selected_values)]

                # 관리본부명과 관리지사에 따른 정렬 적용
                if '관리본부명' in filtered_data.columns and '관리지사' in filtered_data.columns:
                    for bonbu, gisa_list in self.bonbu_gisa_map.items():
                        if bonbu in filtered_data['관리본부명'].unique():
                            filtered_data.loc[filtered_data['관리본부명'] == bonbu, '관리지사'] = pd.Categorical(
                                filtered_data.loc[filtered_data['관리본부명'] == bonbu, '관리지사'],
                                categories=gisa_list,
                                ordered=True
                            )
                    filtered_data = filtered_data.sort_values(['관리본부명', '관리지사'])

                unit = self.selected_unit.get()
                unit_factor = 1
                if unit == "천원":
                    unit_factor = 1000
                elif unit == "백만원":
                    unit_factor = 1000000

                # 값 표시 열 선택
                selected_values_column = self.selected_values_column.get()
                if selected_values_column not in self.df.columns:
                    messagebox.showerror("Column Error", f"The column '{selected_values_column}' does not exist in the file. (해당 파일에 '{selected_values_column}' 열이 없습니다.)")
                    return

                if selected_values_column == "계약번호(행건수)":
                    summary = filtered_data.groupby('관리본부명').agg(
                        계약번호_건수=('계약번호', 'count')
                    ).reset_index()
                elif selected_values_column == "ktt월정료":
                    summary = filtered_data.groupby('관리본부명').agg(
                        ktt월정료_합계=('ktt월정료', lambda x: pd.to_numeric(x, errors='coerce').sum() / unit_factor)
                    ).reset_index()
                    summary['ktt월정료_합계'] = summary['ktt월정료_합계'].apply(lambda x: "{:,.0f}".format(x) + " " + unit)
                else:
                    summary = filtered_data.groupby('관리본부명').agg(
                        계약번호_건수=('계약번호', 'count'),
                        ktt월정료_합계=('ktt월정료', lambda x: pd.to_numeric(x, errors='coerce').sum() / unit_factor)
                    ).reset_index()
                    summary['ktt월정료_합계'] = summary['ktt월정료_합계'].apply(lambda x: "{:,.0f}".format(x) + " " + unit)

                # 필터링된 결과를 CSV로 저장
                output_filtered_csv = self.file_path.get().replace(".csv", "_filtered_summary.csv").replace(".cs", "_filtered_summary.csv").replace(".xls", "_filtered_summary.csv")
                summary.to_csv(output_filtered_csv, index=False, encoding=self.selected_encoding.get())
                messagebox.showinfo("저장 완료", f"필터링된 결과가 {output_filtered_csv}에 저장되었습니다.")

                # 시각화 선택 및 생성
                self.create_chart(summary, selected_values_column)

        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")
 
    def create_chart(self, summary, selected_values_column):
        chart_type = self.selected_chart_type.get()
        plt.figure(figsize=(10, 6))
       
        # 차트 유형에 따른 시각화 생성
        if chart_type == "Bar":
            plt.bar(summary['관리본부명'], summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), color='skyblue')
        elif chart_type == "Line":
            plt.plot(summary['관리본부명'], summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), marker='o', linestyle='-', color='skyblue')
        elif chart_type == "Pie":
            plt.pie(summary[selected_values_column].apply(lambda x: int(x.replace(",", "").split()[0])), labels=summary['관리본부명'], autopct='%1.1f%%', startangle=140)

        # 차트 스타일 및 저장 설정
        plt.ylabel(f'{selected_values_column} ({self.selected_unit.get()})')
        plt.xlabel('관리본부명')
        plt.title(f'관리본부명별 {selected_values_column} - {chart_type} 차트')
        plt.xticks(rotation=45)
        plt.tight_layout()

        # 시각화 결과를 이미지 파일로 저장
        chart_image_output_path = self.file_path.get().replace(".csv", f"_{chart_type}_chart.png").replace(".cs", f"_{chart_type}_chart.png").replace(".xls", f"_{chart_type}_chart.png")
        plt.savefig(chart_image_output_path)
        plt.show()
        messagebox.showinfo("시각화 완료", f"시각화된 이미지가 {chart_image_output_path}에 저장되었습니다.")

    def add_condition(self):
        # 조건을 추가할 때 사용할 메서드 (필요시 구현)
        pass

    def clear_conditions(self):
        """조건 초기화 메서드"""
        self.conditions.clear()
        self.populate_columns()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Data Filtering & Visualization (데이터 필터링 및 시각화)")

        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")
        self.include_conditions = []  # 포함할 조건
        self.exclude_conditions = []  # 제외할 조건

        self.create_widgets()
        self.configure_grid()

    def create_widgets(self):
        tk.Label(self.root, text="File Path (파일 경로):").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="Select File (파일 선택)", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        self.column_listbox = tk.Listbox(self.root, selectmode='single', exportselection=0)
        self.column_listbox.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        self.condition_frame = tk.Frame(self.root)
        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        tk.Label(self.root, text="Select Column to Display Values (표시할 열 선택):").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[]).grid(row=3, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="Select Chart Type (차트 유형 선택):").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="KTT Monthly Charge Unit (KTT 월정료 단위):").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        tk.Label(self.root, text="Select Encoding (인코딩 선택):").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949", "EUC-KR"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="Apply Conditions (조건 적용)", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="Add Condition (조건 추가)", command=self.add_condition).grid(row=7, column=1, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="Clear Conditions (조건 초기화)", command=self.clear_conditions).grid(row=7, column=2, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)

    def load_file(self):
        # .csv 및 .cs 확장자 파일 모두 허용
        file_path = filedialog.askopenfilename(filetypes=[("CSV, CS, and Excel Files", "*.csv;*.cs;*.xls;*.xlsx")])
        if file_path:
            self.file_path.set(file_path)
            try:
                # 여러 인코딩 방식 시도
                try:
                    if file_path.endswith(".csv") or file_path.endswith(".cs"):
                        # 기본적으로 UTF-8 시도
                        self.df = pd.read_csv(self.file_path.get(), encoding=self.selected_encoding.get())
                    elif file_path.endswith(".xls") or file_path.endswith(".xlsx"):
                        self.df = pd.read_excel(self.file_path.get())
                except UnicodeDecodeError:
                    # UTF-8 인코딩 실패 시 CP949 시도
                    try:
                        self.df = pd.read_csv(self.file_path.get(), encoding="CP949")
                    except UnicodeDecodeError:
                        # CP949도 실패할 경우 EUC-KR 시도
                        self.df = pd.read_csv(self.file_path.get(), encoding="EUC-KR")

                self.df.fillna('', inplace=True)
                self.columns = self.df.columns.tolist()
                self.populate_columns()

            except Exception as e:
                messagebox.showerror("오류 발생", f"파일을 읽는 중 오류 발생: {e}")
   
    def populate_columns(self):
        self.column_listbox.delete(0, tk.END)
        for col in self.columns:
            self.column_listbox.insert(tk.END, col)

        values_combobox = self.root.children["!combobox"]
        values_combobox['values'] = ["계약번호", "ktt월정료(조정)"] + self.columns
        values_combobox.current(0)

    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()

                # ktt월정료(조정)이 없을 경우 대체할 열 확인
                target_column = self.selected_values_column.get()
                if target_column not in self.df.columns:
                    # 가장 비슷한 열 이름을 찾아 대체
                    similar_columns = [col for col in self.df.columns if 'ktt월정료' in col]
                    if similar_columns:
                        target_column = similar_columns[0]
                        self.selected_values_column.set(target_column)  # 대체 열 설정
                        messagebox.showwarning("Column Warning", f"'{self.selected_values_column.get()}' does not exist. Using '{target_column}' instead.")
                    else:
                        messagebox.showerror("Column Error", f"The column '{self.selected_values_column.get()}' does not exist in the file. (해당 파일에 '{self.selected_values_column.get()}' 열이 없습니다.)")
                        return

                # 필터링 조건 적용
                for condition in self.include_conditions:
                    filtered_data = filtered_data[filtered_data[condition[0]] == condition[1]]

                for condition in self.exclude_conditions:
                    filtered_data = filtered_data[filtered_data[condition[0]] != condition[1]]

                # 조건에 따른 필터링 적용
                summary = filtered_data.groupby('관리본부명').agg(
                    계약번호_건수=('계약번호', 'count'),
                    ktt월정료_합계=(target_column, lambda x: pd.to_numeric(x, errors='coerce').sum())
                ).reset_index()

                # 결과 테이블 출력
                self.display_summary_table(summary)

        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")

    def add_condition(self):
        """조건 추가할 때 사용할 메서드"""
        condition_window = tk.Toplevel(self.root)
        condition_window.title("Add Condition")

        tk.Label(condition_window, text="Select Condition Column:").grid(row=0, column=0, padx=10, pady=10)
        condition_column = ttk.Combobox(condition_window, values=self.columns)
        condition_column.grid(row=0, column=1, padx=10, pady=10)

        tk.Label(condition_window, text="Condition Value:").grid(row=1, column=0, padx=10, pady=10)
        condition_value = tk.Entry(condition_window)
        condition_value.grid(row=1, column=1, padx=10, pady=10)

        def add_include():
            self.include_conditions.append((condition_column.get(), condition_value.get()))
            messagebox.showinfo("Condition Added", f"Included Condition: {condition_column.get()} = {condition_value.get()}")
            condition_window.destroy()

       
        def add_exclude():
            self.exclude_conditions.append((condition_column.get(), condition_value.get()))
            messagebox.showinfo("Condition Added", f"Excluded Condition: {condition_column.get()} != {condition_value.get()}")
            condition_window.destroy()

        tk.Button(condition_window, text="Include Condition", command=add_include).grid(row=2, column=0, padx=10, pady=10)
        tk.Button(condition_window, text="Exclude Condition", command=add_exclude).grid(row=2, column=1, padx=10, pady=10)

    def display_summary_table(self, summary):
        """계약번호 건수와 월정료 금액을 표로 표시"""
        table_window = tk.Toplevel(self.root)
        table_window.title("Summary Table (요약 테이블)")

        # 테이블 생성
        cols = list(summary.columns)
        tree = ttk.Treeview(table_window, columns=cols, show="headings")

        for col in cols:
            tree.heading(col, text=col)
            tree.column(col, width=100)

        for index, row in summary.iterrows():
            tree.insert("", "end", values=list(row))

        tree.pack(expand=True, fill="both")

    def clear_conditions(self):
        """조건 초기화"""
        self.include_conditions.clear()
        self.exclude_conditions.clear()
        self.populate_columns()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("데이터 필터링 및 시각화")

        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")
        self.include_conditions = []  # 포함할 조건
        self.exclude_conditions = []  # 제외할 조건

        # 관리본부명 매핑
        self.bonbu_map = {
            '강원본부': '강북/강원본부',
            '서부본부': '강남/서부본부'
        }

        self.create_widgets()
        self.configure_grid()

    def create_widgets(self):
        tk.Label(self.root, text="파일 경로 (File Path):").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="파일 선택 (Select File)", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        self.column_listbox = tk.Listbox(self.root, selectmode='single', exportselection=0)
        self.column_listbox.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        # 조건 적용 프레임
        self.condition_frame = tk.Frame(self.root)
        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        # 표시할 열 선택 콤보박스
        tk.Label(self.root, text="표시할 열 선택 (Select Column to Display):").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        self.value_combobox = ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[])
        self.value_combobox.grid(row=3, column=1, padx=10, pady=10, sticky="ew")
        self.value_combobox.bind("<<ComboboxSelected>>", self.populate_conditions)

        # 차트 유형 선택 콤보박스
        tk.Label(self.root, text="차트 유형 선택 (Select Chart Type):").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")

        # 단위 선택
        tk.Label(self.root, text="월정료 단위 (Unit of KTT Monthly Charge):").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        # 인코딩 선택
        tk.Label(self.root, text="인코딩 선택 (Select Encoding):").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949", "EUC-KR"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="조건 적용 (Apply Conditions)", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 초기화 (Clear Conditions)", command=self.clear_conditions).grid(row=7, column=1, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)

    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV, CS, and Excel Files", "*.csv;*.cs;*.xls;*.xlsx")])
        if file_path:
            self.file_path.set(file_path)
            try:
                # 여러 인코딩 방식 시도
                try:
                    if file_path.endswith(".csv") or file_path.endswith(".cs"):
                        # UTF-8로 시도
                        self.df = pd.read_csv(self.file_path.get(), encoding=self.selected_encoding.get())
                    elif file_path.endswith(".xls") or file_path.endswith(".xlsx"):
                        self.df = pd.read_excel(self.file_path.get())
                except UnicodeDecodeError:
                    # UTF-8 인코딩 실패 시 CP949 시도
                    try:
                        self.df = pd.read_csv(self.file_path.get(), encoding="CP949")
                    except UnicodeDecodeError:
                        # CP949도 실패할 경우 EUC-KR 시도
                        self.df = pd.read_csv(self.file_path.get(), encoding="EUC-KR")

                self.df.fillna('', inplace=True)
                self.columns = self.df.columns.tolist()
                self.populate_columns()

            except Exception as e:
                messagebox.showerror("오류 발생", f"파일을 읽는 중 오류 발생: {e}")
   
    def populate_columns(self):
        self.column_listbox.delete(0, tk.END)
        for col in self.columns:
            self.column_listbox.insert(tk.END, col)

        self.value_combobox['values'] = ["계약번호", "ktt월정료(조정)"] + self.columns
        self.value_combobox.current(0)

    def populate_conditions(self, event):
        selected_column = self.value_combobox.get()
        if selected_column == '관리본부명':
            self.populate_bonbu_condition()
        else:
            self.populate_generic_condition(selected_column)

    def populate_bonbu_condition(self):
        """관리본부명을 체크박스로 보여주기"""
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

        bonbu_list = ['강북/강원본부', '강남/서부본부', '대구/경북본부', '부산/경남본부', '전남/전북본부', '충남/충북본부']
        self.bonbu_vars = {}

        for bonbu in bonbu_list:
            var = tk.BooleanVar()
            chk = tk.Checkbutton(self.condition_frame, text=bonbu, variable=var)
            chk.grid(sticky='w')
            self.bonbu_vars[bonbu] = var

    def populate_generic_condition(self, column):
        """선택한 열에 대한 조건 표시 (예: 서비스, 계약번호 등)"""
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

        unique_values = self.df[column].unique().tolist()
        self.condition_vars = {}

        for value in unique_values:
            var = tk.BooleanVar()
            chk = tk.Checkbutton(self.condition_frame, text=value, variable=var)
            chk.grid(sticky='w')
            self.condition_vars[value] = var

    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()

                # 관리본부명에 대한 체크된 값 필터링
                if '관리본부명' in filtered_data.columns:
                    selected_bonbu = [bonbu for bonbu, var in self.bonbu_vars.items() if var.get()]
                    if selected_bonbu:
                        filtered_data = filtered_data[filtered_data['관리본부명'].isin(selected_bonbu)]

                # 다른 선택한 조건들에 대한 필터링
                selected_column = self.value_combobox.get()
                if selected_column != '관리본부명':
                    selected_values = [val for val, var in self.condition_vars.items() if var.get()]
                    if selected_values:
                       
                        filtered_data = filtered_data[filtered_data[selected_column].isin(selected_values)]

                # 필터링 결과를 그룹화하여 계산
                summary = filtered_data.groupby('관리본부명').agg(
                    계약번호_건수=('계약번호', 'count'),
                    ktt월정료_합계=('ktt월정료(조정)', lambda x: pd.to_numeric(x, errors='coerce').sum())
                ).reset_index()

                # 결과 테이블 출력
                self.display_summary_table(summary)

        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")

    def display_summary_table(self, summary):
        """계약번호 건수와 월정료 금액을 표로 표시"""
        table_window = tk.Toplevel(self.root)
        table_window.title("Summary Table (요약 테이블)")

        # 테이블 생성
        cols = list(summary.columns)
        tree = ttk.Treeview(table_window, columns=cols, show="headings")

        for col in cols:
            tree.heading(col, text=col)
            tree.column(col, width=100)

        for index, row in summary.iterrows():
            tree.insert("", "end", values=list(row))

        tree.pack(expand=True, fill="both")

    def clear_conditions(self):
        """조건 초기화"""
        self.include_conditions.clear()
        self.exclude_conditions.clear()
        self.populate_columns()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("데이터 필터링 및 시각화")

        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")
        self.include_conditions = []  # 포함할 조건
        self.exclude_conditions = []  # 제외할 조건

        # 관리본부명 매핑
        self.bonbu_map = {
            '강원본부': '강북/강원본부',
            '서부본부': '강남/서부본부'
        }

        self.create_widgets()
        self.configure_grid()

    def create_widgets(self):
        tk.Label(self.root, text="파일 경로 (File Path):").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="파일 선택 (Select File)", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        # 헤더 표시 리스트박스
        self.column_frame = tk.Frame(self.root)
        self.column_frame.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        # 조건 적용 프레임
        self.condition_frame = tk.Frame(self.root)
        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        # 표시할 열 선택 콤보박스
        tk.Label(self.root, text="표시할 열 선택 (Select Column to Display):").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        self.value_combobox = ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[])
        self.value_combobox.grid(row=3, column=1, padx=10, pady=10, sticky="ew")
        self.value_combobox.bind("<<ComboboxSelected>>", self.populate_conditions)

        # 차트 유형 선택 콤보박스
        tk.Label(self.root, text="차트 유형 선택 (Select Chart Type):").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")

        # 단위 선택
        tk.Label(self.root, text="월정료 단위 (Unit of KTT Monthly Charge):").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        # 인코딩 선택
        tk.Label(self.root, text="인코딩 선택 (Select Encoding):").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949", "EUC-KR"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="조건 적용 (Apply Conditions)", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 초기화 (Clear Conditions)", command=self.clear_conditions).grid(row=7, column=1, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)

    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV, CS, and Excel Files", "*.csv;*.cs;*.xls;*.xlsx")])
        if file_path:
            self.file_path.set(file_path)
            try:
                # 여러 인코딩 방식 시도
                try:
                    if file_path.endswith(".csv") or file_path.endswith(".cs"):
                        # UTF-8로 시도
                        self.df = pd.read_csv(self.file_path.get(), encoding=self.selected_encoding.get())
                    elif file_path.endswith(".xls") or file_path.endswith(".xlsx"):
                        self.df = pd.read_excel(self.file_path.get())
                except UnicodeDecodeError:
                    # UTF-8 인코딩 실패 시 CP949 시도
                    try:
                        self.df = pd.read_csv(self.file_path.get(), encoding="CP949")
                    except UnicodeDecodeError:
                        # CP949도 실패할 경우 EUC-KR 시도
                        self.df = pd.read_csv(self.file_path.get(), encoding="EUC-KR")

                self.df.fillna('', inplace=True)
                self.columns = self.df.columns.tolist()

                # 파일이 로드되면 헤더를 4줄로 보기 좋게 나열
                self.populate_columns()

            except Exception as e:
                messagebox.showerror("오류 발생", f"파일을 읽는 중 오류 발생: {e}")

    def populate_columns(self):
        # 열 이름을 4줄로 보기 좋게 표시
        for widget in self.column_frame.winfo_children():
            widget.destroy()

        cols_per_row = 4
        for i in range(0, len(self.columns), cols_per_row):
            frame = tk.Frame(self.column_frame)
            frame.pack(side="top", fill="x", pady=5)
            for col in self.columns[i:i+cols_per_row]:
                tk.Button(frame, text=col, command=lambda c=col: self.populate_conditions_by_column(c)).pack(side="left", padx=5)

        # 계약번호 및 KTT월정료(조정)를 자동으로 기본 값으로 설정
        if '계약번호' in self.columns:
            self.selected_values_column.set('계약번호')
        if 'KTT월정료(조정)' in self.columns:
            self.selected_values_column.set('KTT월정료(조정)')
        elif 'KTT월정료' in self.columns:
            self.selected_values_column.set('KTT월정료')

        self.value_combobox['values'] = ["계약번호", "KTT월정료(조정)", "KTT월정료"] + self.columns
        self.value_combobox.current(0)

    def populate_conditions(self, event=None):
        """콤보박스에서 선택한 열의 값을 체크박스로 표시"""
        selected_column = self.value_combobox.get()
        self.populate_conditions_by_column(selected_column)

    def populate_conditions_by_column(self, column):
        """선택한 열의 값을 체크박스로 표시"""
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

        unique_values = self.df[column].unique().tolist()
        self.condition_vars = {}

        for value in unique_values:
            var = tk.BooleanVar()
            chk = tk.Checkbutton(self.condition_frame, text=value, variable=var)
            chk.pack(anchor='w')
            self.condition_vars[value] = var

    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()

                # 관리본부명 매핑 적용
                if '관리본부명' in filtered_data.columns:
                    filtered_data['관리본부명'] = filtered_data['관리본부명'].replace(self.bonbu_map)

                # 계약번호 필터링 기본 적용
                if '계약번호' in filtered_data.columns:
                    filtered_data = filtered_data.dropna(subset=['계약번호'])

                # 월정료 필터링 (KTT월정료(조정), KTT월정료)
                selected_revenue_column = None
                if 'KTT월정료(조정)' in filtered_data.columns:
                 
                    selected_revenue_column = 'KTT월정료(조정)'
                elif 'KTT월정료' in filtered_data.columns:
                    selected_revenue_column = 'KTT월정료'
                else:
                    selected_revenue_column = self.selected_values_column.get()

                # 선택한 조건으로 필터링
                selected_values = [val for val, var in self.condition_vars.items() if var.get()]
                if selected_values:
                    filtered_data = filtered_data[filtered_data[selected_revenue_column].isin(selected_values)]

                # 필터링 결과를 그룹화하여 계산
                summary = filtered_data.groupby('관리본부명').agg(
                    계약번호_건수=('계약번호', 'count'),
                    매출_합계=(selected_revenue_column, lambda x: pd.to_numeric(x, errors='coerce').sum())
                ).reset_index()

                # 결과 테이블 출력
                self.display_summary_table(summary)

        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")

    def display_summary_table(self, summary):
        """계약번호 건수와 월정료 금액을 표로 표시"""
        table_window = tk.Toplevel(self.root)
        table_window.title("Summary Table (요약 테이블)")

        # 테이블 생성
        cols = list(summary.columns)
        tree = ttk.Treeview(table_window, columns=cols, show="headings")

        for col in cols:
            tree.heading(col, text=col)
            tree.column(col, width=100)

        for index, row in summary.iterrows():
            tree.insert("", "end", values=list(row))

        tree.pack(expand=True, fill="both")

    def clear_conditions(self):
        """조건 초기화"""
        self.condition_vars = {}
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog, messagebox, ttk

class DataFilterApp:
    def __init__(self, root):
        self.root = root
        self.root.title("데이터 필터링 및 시각화")

        self.file_path = tk.StringVar()
        self.df = None
        self.columns = []
        self.conditions = {}
        self.selected_values_column = tk.StringVar()
        self.selected_chart_type = tk.StringVar(value="Bar")
        self.selected_encoding = tk.StringVar(value="UTF-8")
        self.selected_unit = tk.StringVar(value="원")
        self.include_conditions = []  # 포함할 조건
        self.exclude_conditions = []  # 제외할 조건

        # 관리본부명 매핑
        self.bonbu_map = {
            '강원본부': '강북/강원본부',
            '서부본부': '강남/서부본부'
        }

        self.create_widgets()
        self.configure_grid()

    def create_widgets(self):
        tk.Label(self.root, text="파일 경로 (File Path):").grid(row=0, column=0, padx=10, pady=10, sticky="w")
        tk.Entry(self.root, textvariable=self.file_path, width=50).grid(row=0, column=1, padx=10, pady=10, sticky="ew")
        tk.Button(self.root, text="파일 선택 (Select File)", command=self.load_file).grid(row=0, column=2, padx=10, pady=10, sticky="e")

        # 스크롤바가 포함된 헤더 표시 프레임
        self.header_frame = tk.Frame(self.root)
        self.header_frame.grid(row=1, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        self.header_canvas = tk.Canvas(self.header_frame)
        self.header_scrollbar = ttk.Scrollbar(self.header_frame, orient="vertical", command=self.header_canvas.yview)
        self.header_inner_frame = tk.Frame(self.header_canvas)

        self.header_inner_frame.bind(
            "<Configure>",
            lambda e: self.header_canvas.configure(scrollregion=self.header_canvas.bbox("all"))
        )
        self.header_canvas.create_window((0, 0), window=self.header_inner_frame, anchor="nw")
        self.header_canvas.configure(yscrollcommand=self.header_scrollbar.set)

        self.header_canvas.pack(side="left", fill="both", expand=True)
        self.header_scrollbar.pack(side="right", fill="y")

        # 조건 적용 프레임
        self.condition_frame = tk.Frame(self.root)
        self.condition_frame.grid(row=2, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

        # 표시할 열 선택 콤보박스
        tk.Label(self.root, text="표시할 열 선택 (Select Column to Display):").grid(row=3, column=0, padx=10, pady=10, sticky="w")
        self.value_combobox = ttk.Combobox(self.root, textvariable=self.selected_values_column, values=[])
        self.value_combobox.grid(row=3, column=1, padx=10, pady=10, sticky="ew")
        self.value_combobox.bind("<<ComboboxSelected>>", self.populate_conditions)

        # 차트 유형 선택 콤보박스
        tk.Label(self.root, text="차트 유형 선택 (Select Chart Type):").grid(row=4, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_chart_type, values=["Bar", "Line", "Pie"]).grid(row=4, column=1, padx=10, pady=10, sticky="ew")

        # 단위 선택
        tk.Label(self.root, text="월정료 단위 (Unit of KTT Monthly Charge):").grid(row=5, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_unit, values=["원", "천원", "백만원"]).grid(row=5, column=1, padx=10, pady=10, sticky="ew")

        # 인코딩 선택
        tk.Label(self.root, text="인코딩 선택 (Select Encoding):").grid(row=6, column=0, padx=10, pady=10, sticky="w")
        ttk.Combobox(self.root, textvariable=self.selected_encoding, values=["UTF-8", "CP949", "EUC-KR"]).grid(row=6, column=1, padx=10, pady=10, sticky="ew")

        tk.Button(self.root, text="조건 적용 (Apply Conditions)", command=self.apply_conditions).grid(row=7, column=0, padx=10, pady=20, sticky="ew")
        tk.Button(self.root, text="조건 초기화 (Clear Conditions)", command=self.clear_conditions).grid(row=7, column=1, padx=10, pady=20, sticky="ew")

    def configure_grid(self):
        self.root.grid_columnconfigure(0, weight=1)
        self.root.grid_columnconfigure(1, weight=3)
        self.root.grid_columnconfigure(2, weight=1)
        self.root.grid_rowconfigure(1, weight=1)
        self.root.grid_rowconfigure(2, weight=1)

    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV, CS, and Excel Files", "*.csv;*.cs;*.xls;*.xlsx")])
        if file_path:
            self.file_path.set(file_path)
            try:
                # 여러 인코딩 방식 시도
                try:
                    if file_path.endswith(".csv") or file_path.endswith(".cs"):
                        self.df = pd.read_csv(self.file_path.get(), encoding=self.selected_encoding.get())
                    elif file_path.endswith(".xls") or file_path.endswith(".xlsx"):
                        self.df = pd.read_excel(self.file_path.get())
                except UnicodeDecodeError:
                    try:
                        self.df = pd.read_csv(self.file_path.get(), encoding="CP949")
                    except UnicodeDecodeError:
                        self.df = pd.read_csv(self.file_path.get(), encoding="EUC-KR")

                self.df.fillna('', inplace=True)
                self.columns = self.df.columns.tolist()

                # 파일이 로드되면 헤더를 보기 좋게 나열
                self.populate_columns()

            except Exception as e:
                messagebox.showerror("오류 발생", f"파일을 읽는 중 오류 발생: {e}")

    def populate_columns(self):
        # 열 이름을 보기 좋게 표시하고 스크롤 가능하도록 설정
        for widget in self.header_inner_frame.winfo_children():
            widget.destroy()

        cols_per_row = 4
        for i in range(0, len(self.columns), cols_per_row):
            frame = tk.Frame(self.header_inner_frame)
            frame.pack(side="top", fill="x", pady=5)
            for col in self.columns[i:i+cols_per_row]:
                button = tk.Button(frame, text=col, command=lambda c=col: self.populate_conditions_by_column(c))
                button.pack(side="left", padx=5)

        # 기본 필드 설정
        if '계약번호' in self.columns:
            self.selected_values_column.set('계약번호')
        if 'KTT월정료(조정)' in self.columns:
            self.selected_values_column.set('KTT월정료(조정)')
        elif 'KTT월정료' in self.columns:
            self.selected_values_column.set('KTT월정료')

        self.value_combobox['values'] = ["계약번호", "KTT월정료(조정)", "KTT월정료"] + self.columns
        self.value_combobox.current(0)

    def populate_conditions(self, event=None):
        """콤보박스에서 선택한 열의 값을 체크박스로 표시"""
        selected_column = self.value_combobox.get()
        self.populate_conditions_by_column(selected_column)

    def populate_conditions_by_column(self, column):
        """선택한 열의 값을 체크박스로 표시"""
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

        unique_values = sorted(set(self.df[column].dropna()))  # 중복 제거 및 정렬
        self.condition_vars = {}

        for value in unique_values:
            var = tk.BooleanVar()
            chk = tk.Checkbutton(self.condition_frame, text=value, variable=var)
            chk.pack(anchor='w')
            self.condition_vars[value] = var

   
    def apply_conditions(self):
        try:
            if not self.df.empty:
                filtered_data = self.df.copy()

                # 관리본부명 매핑 적용 (만약 관리본부명을 선택한 경우)
                if '관리본부명' in filtered_data.columns:
                    filtered_data['관리본부명'] = filtered_data['관리본부명'].replace(self.bonbu_map)

                # 선택된 체크박스에 따라 데이터를 필터링
                selected_column = self.value_combobox.get()
                selected_values = [val for val, var in self.condition_vars.items() if var.get()]

                if selected_values:
                    filtered_data = filtered_data[filtered_data[selected_column].isin(selected_values)]

                # 필터링 결과를 그룹화하여 계산
                summary = filtered_data.groupby('관리본부명').agg(
                    계약번호_건수=('계약번호', 'count'),
                    매출_합계=(selected_column, lambda x: pd.to_numeric(x, errors='coerce').sum())
                ).reset_index()

                # 결과 테이블 출력
                self.display_summary_table(summary)

        except Exception as e:
            messagebox.showerror("오류 발생", f"오류가 발생했습니다: {str(e)}")

    def display_summary_table(self, summary):
        """계약번호 건수와 매출 합계를 표로 표시"""
        table_window = tk.Toplevel(self.root)
        table_window.title("Summary Table (요약 테이블)")

        # 테이블 생성
        cols = list(summary.columns)
        tree = ttk.Treeview(table_window, columns=cols, show="headings")

        for col in cols:
            tree.heading(col, text=col)
            tree.column(col, width=100)

        for index, row in summary.iterrows():
            tree.insert("", "end", values=list(row))

        tree.pack(expand=True, fill="both")

    def clear_conditions(self):
        """조건 초기화"""
        self.condition_vars = {}
        for widget in self.condition_frame.winfo_children():
            widget.destroy()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFilterApp(root)
    root.mainloop()
```
