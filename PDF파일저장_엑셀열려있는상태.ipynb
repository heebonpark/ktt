## 아래는 요청하신 전체 작업을 한 번에 처리하는 Python 코드입니다. 주어진 단계별 작업을 모두 포함하여 효율적으로 처리할 수 있도록 수정했습니다.

### 주요 작업 흐름:
1. `신규P값병합08월_09월_0925_누계.csv`와 `신규P값병합0925_일일.csv` 파일을 병합하고, 각각 엑셀의 `Pv_db` 시트의 지정된 열에 값으로 붙여넣습니다.
2. `M`열에 첫 번째 "본사"로 시작하는 행을 찾아서 지정된 시트에 복사해 붙여넣습니다.
3. 복사 후 `B2~I` 영역에 값이 있는데, `J~O` 열이 비어 있을 경우 바로 위의 값을 서식과 함께 자동으로 붙여넣습니다.
4. 마지막으로 작업한 내용을 PDF로 변환하고 저장 및 엽니다.
### 통합된 Python 코드:

### 설명:
1. **파일 병합 및 데이터 처리**: `merge_and_save_files()` 함수로 일일 및 누계 데이터를 병합하여 CSV 파일로 저장한 후, 이 데이터를 `copy_data_to_excel()`에서 엑셀의 지정된 위치에 복사합니다.
2. **특정 "본사" 키워드 찾기 및 복사**: `Pv_db` 시트의 `M`열에서 "본사" 키워드를 찾고 해당 행의 데이터를 `92_SP사원(개인)(9월 누계)` 시트의 `B2:I2`에 복사합니다.
3. **빈 열 처리**: `B2:I`에 값이 있으나 `J:O`가 비어 있는 경우, 바로 위의 셀에서 서식과 값을 복사합니다.
4. **PDF 변환 및 열기**: `export_sheet_to_pdf()` 함수에서 "메일" 시트를 PDF로 변환하고 저장한 후, PDF 파일을 자동으로 엽니다.

이 코드로 주어진 모든 작업을 한 번에 처리할 수 있습니다.

```python
import os
import pandas as pd
import glob
import win32com.client
from datetime import datetime

# 폴더 경로 설정 (일일 데이터와 누계 데이터 경로)
daily_folder_path = r'd:\신규P값총량제\병합'
cumulative_folder_path = r'd:\신규P값총량제\누계병합'

# 파일명에 따른 정렬 기준
file_order = [
    "0. 채널별, 월", "1.기관별, 월",  # "기관별" 파일을 채널 다음으로 정렬
    "2.강북_월", "2.강남_월", "2.부산",
    "2.전남", "2.대구", "2.충남",
    "사원별"
]

# 파일명 정렬 함수
def get_file_priority(file_name):
    for i, keyword in enumerate(file_order):
        if keyword in file_name:
            return i
    return len(file_order)  # 키워드가 없으면 가장 뒤로 정렬

# "제거할 특정 숫자 + 공백 + 건" 패턴을 포함하는 행을 필터링하기 위한 함수 (A열에만 적용)
def filter_out_specific_gun_rows(df):
    remove_patterns = [r'8\s건$', r'12\s건$', r'6\s건$', r'9\s건$', r'7\s건$']
    pattern = '|'.join(remove_patterns)
    return df[~df.iloc[:, 0].astype(str).str.contains(pattern)]

# 폴더에서 파일을 읽어 병합하고 저장하는 함수
def merge_and_save_files(folder_path, output_filename):
    # 모든 파일 경로 가져오기
    all_files = glob.glob(folder_path + "/*.xls")

    # 파일명에 따라 정렬
    all_files.sort(key=lambda x: get_file_priority(os.path.basename(x)))

    # 전체 데이터프레임 초기화
    all_data = pd.DataFrame()

    # 각 파일의 데이터를 순서대로 병합
    for i, file in enumerate(all_files):
        try:
            # HTML 형식으로 파일 읽기
            df_list = pd.read_html(file)
            df = df_list[0]  # 첫 번째 테이블만 사용

            # 특정 "숫자 + 공백 + 건" 패턴을 포함하는 행 제거 (A열만)
            df = filter_out_specific_gun_rows(df)

            # 첫 번째 파일에서만 헤더를 사용하고, 이후 파일들은 헤더 없이 추가
            if all_data.empty:
                all_data = df
            else:
                # 기존 열들과 현재 파일의 열을 맞추어 병합
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)

        except Exception as e:
            print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")

    # 병합된 데이터를 csv 파일로 저장 (CP949 인코딩 적용)
    if not all_data.empty:
        csv_output_path = os.path.join(folder_path, output_filename)
        all_data.to_csv(csv_output_path, index=False, encoding='cp949')
        print(f"병합된 파일이 {csv_output_path}에 저장되었습니다.")
    else:
        print("병합된 데이터가 비어 있습니다. 엑셀 파일들을 확인하세요.")

# Excel 데이터를 병합하고 붙여넣는 함수
def copy_data_to_excel():
    excel = win32com.client.Dispatch("Excel.Application")
    excel.Visible = True  # Excel을 표시함
    excel.ScreenUpdating = False  # 작업 도중 화면 업데이트 방지

    # 현재 열려있는 엑셀 파일 찾기
    workbook = excel.Workbooks.Item(1)
    pv_db_sheet = workbook.Sheets("Pv_db")
    cumulative_sheet = workbook.Sheets("92_SP사원(개인)(9월 누계)")

    # 일일 데이터 처리
    merge_and_save_files(daily_folder_path, "신규P값병합0925_일일.csv")
    daily_data = pd.read_csv(os.path.join(daily_folder_path, "신규P값병합0926_일일.csv"), encoding='cp949')

    # 누계 데이터 처리
    merge_and_save_files(cumulative_folder_path, "신규P값병합08월_09월_0926_누계.csv")
    cumulative_data = pd.read_csv(os.path.join(cumulative_folder_path, "신규P값병합08월_09월_0926_누계.csv"), encoding='cp949')

    # 엑셀의 B~I열에 누계 데이터 붙여넣기
    print("누계 데이터 붙여넣기 작업 시작...")
    for row_num, row_data in enumerate(cumulative_data.values):
        for col_num, value in enumerate(row_data[:8]):  # B~I 열
            pv_db_sheet.Cells(row_num + 1, col_num + 2).Value = value

    # 엑셀의 M~T열에 일일 데이터 붙여넣기
    print("일일 데이터 붙여넣기 작업 시작...")
    for row_num, row_data in enumerate(daily_data.values):
        for col_num, value in enumerate(row_data[:8]):  # M~T 열
            pv_db_sheet.Cells(row_num + 1, col_num + 13).Value = value

    # "본사"로 시작하는 첫 번째 행 찾기 및 복사
    for row in range(1, pv_db_sheet.UsedRange.Rows.Count + 1):
        if "본사" in str(pv_db_sheet.Cells(row, 13).Value):  # M열에서 "본사" 찾기
            for col in range(13, 21):  # M~T 열 복사
                cumulative_sheet.Cells(2, col - 11).Value = pv_db_sheet.Cells(row, col).Value
            break

    # 빈 셀 서식 복사
    print("빈 셀 서식 복사 작업 시작...")
    for row in range(2, cumulative_sheet.UsedRange.Rows.Count + 1):
        for col in range(2, 9):  # B~I 열 확인
            if cumulative_sheet.Cells(row, col).Value is not None and cumulative_sheet.Cells(row, col + 7).Value is None:
                for shift_col in range(10, 16):  # J~O 열에 서식 복사
                    cumulative_sheet.Cells(row, shift_col).Value = cumulative_sheet.Cells(row - 1, shift_col).Value

    # 엑셀 저장
    workbook.Save()

    # 화면 업데이트 복구
    excel.ScreenUpdating = True

    print("데이터 병합 및 엑셀 처리 완료.")

    return excel, workbook  # 엑셀 객체 반환

# PDF 변환 및 저장 함수
def export_sheet_to_pdf(excel, workbook):
    current_date = datetime.now().strftime("%Y%m%d")
    output_dir = "D:\\신규P값총량제\\공유메일"
    output_pdf = f"메일용_{current_date}.pdf"
    final_pdf = os.path.join(output_dir, output_pdf)

    mail_sheet = workbook.Sheets("메일")
    mail_sheet.Visible = True

    # PDF 변환
    mail_sheet.PageSetup.PrintGridlines = False  # 눈금선 제외
    mail_sheet.ExportAsFixedFormat(0, final_pdf)

    # PDF 파일 열기
    os.startfile(final_pdf)

    print(f"PDF 파일이 {final_pdf}에 저장되고 열렸습니다.")

    # 엑셀 종료
    workbook.Close(SaveChanges=True)
    excel.Quit()

# 메인 함수 실행
if __name__ == "__main__":
    excel_app, open_workbook = copy_data_to_excel()  # Excel 객체와 Workbook을 받음
    export_sheet_to_pdf(excel_app, open_workbook)  # PDF 변환 및 엑셀 종료
```

```python
import os
import win32com.client
from datetime import datetime
import pandas as pd
import glob
import psutil

# 백그라운드의 엑셀 프로세스를 강제로 종료하는 함수
def kill_excel_process():
    for process in psutil.process_iter():
        try:
            if process.name().lower() == 'excel.exe':
                process.kill()
        except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
            pass

# 동일한 파일 이름이 존재할 경우 이름을 변경하는 함수
def get_unique_filename(file_path):
    if not os.path.exists(file_path):
        return file_path
    base, ext = os.path.splitext(file_path)
    counter = 1
    new_file_path = f"{base}_{counter}{ext}"
    while os.path.exists(new_file_path):
        counter += 1
        new_file_path = f"{base}_{counter}{ext}"
    return new_file_path

# 병합된 CSV 파일 생성 경로 설정
daily_folder_path = r'd:\신규P값총량제\자동화연습\병합'
cumulative_folder_path = r'd:\신규P값총량제\자동화연습\누계병합'

# 파일명에 따른 정렬 기준
file_order = [
    "0. 채널별, 월", "1.기관별, 월",  
    "2.강북_월", "2.강남_월", "2.부산",
    "2.전남", "2.대구", "2.충남",
    "사원별"
]

# 파일명 정렬 함수
def get_file_priority(file_name):
    for i, keyword in enumerate(file_order):
        if keyword in file_name:
            return i
    return len(file_order)

# "제거할 특정 숫자 + 공백 + 건" 패턴을 포함하는 행을 필터링하는 함수
def filter_out_specific_gun_rows(df):
    remove_patterns = [r'8\s건$', r'12\s건$', r'6\s건$', r'9\s건$', r'7\s건$']
    pattern = '|'.join(remove_patterns)
    return df[~df.iloc[:, 0].astype(str).str.contains(pattern)]

# 폴더에서 파일을 읽어 병합하고 저장하는 함수 (동일 파일명 처리 포함)
def merge_and_save_files(folder_path, output_filename):
    all_files = glob.glob(folder_path + "/*.xls")
    all_files.sort(key=lambda x: get_file_priority(os.path.basename(x)))

    all_data = pd.DataFrame()

    for file in all_files:
        try:
            df_list = pd.read_html(file)
            df = df_list[0]
            df = filter_out_specific_gun_rows(df)

            if all_data.empty:
                all_data = df
            else:
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)
        except Exception as e:
            print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")

    if not all_data.empty:
        csv_output_path = os.path.join(folder_path, output_filename)
        unique_output_path = get_unique_filename(csv_output_path)
        all_data.to_csv(unique_output_path, index=False, encoding='cp949')
        print(f"병합된 파일이 {unique_output_path}에 저장되었습니다.")
    else:
        print("병합된 데이터가 비어 있습니다. 엑셀 파일들을 확인하세요.")

# 병합된 CSV 파일을 Excel로 열고 붙여넣기 함수
def open_csv_in_excel_and_copy(csv_path, target_sheet, start_col, start_row):
    workbook = excel.Workbooks.Open(csv_path)
    sheet = workbook.Sheets(1)

    last_row = sheet.UsedRange.Rows.Count
    last_col = sheet.UsedRange.Columns.Count

    source_range = sheet.Range(sheet.Cells(1, 1), sheet.Cells(last_row, last_col))
    target_range = target_sheet.Range(target_sheet.Cells(start_row, start_col), target_sheet.Cells(start_row + last_row - 1, start_col + last_col - 1))
    target_range.Value = source_range.Value

    workbook.Close(False)

# 엑셀 데이터를 병합하고 붙여넣는 함수
def copy_data_to_excel():
    global excel
    excel = win32com.client.Dispatch("Excel.Application")
    excel.Visible = True
    excel.ScreenUpdating = False  # 화면 업데이트 비활성화

    try:
        # 엑셀 파일을 명시적으로 열기 (파일 경로를 정확하게 명시)
        excel_path = r"D:\신규P값총량제\자동화연습\작업파일.xlsx"
       
        # 파일이 존재하지 않으면 에러 발생
        if not os.path.exists(excel_path):
            raise Exception(f"파일을 찾을 수 없습니다: {excel_path}")
       
        workbook = excel.Workbooks.Open(excel_path)
        pv_db_sheet = workbook.Sheets("Pv_db")
        cumulative_sheet = workbook.Sheets("92_SP사원(개인)(9월 누계)")

        merge_and_save_files(daily_folder_path, "신규P값병합_일일.csv")
        daily_csv_path = os.path.join(daily_folder_path, "신규P값병합_일일.csv")

        merge_and_save_files(cumulative_folder_path, "신규P값병합08월_09월_누계.csv")
        cumulative_csv_path = os.path.join(cumulative_folder_path, "신규P값병합08월_09월_누계.csv")

        print("누계 데이터 붙여넣기 작업 시작...")
        open_csv_in_excel_and_copy(cumulative_csv_path, pv_db_sheet, start_col=2, start_row=1)

        print("일일 데이터 붙여넣기 작업 시작...")
        open_csv_in_excel_and_copy(daily_csv_path, pv_db_sheet, start_col=13, start_row=1)

        for row in range(1, pv_db_sheet.UsedRange.Rows.Count + 1):
            if "본사" in str(pv_db_sheet.Cells(row, 13).Value):
                for col in range(13, 21):
                    cumulative_sheet.Cells(2, col - 11).Value = pv_db_sheet.Cells(row, col).Value
                break

        print("빈 셀 서식 복사 작업 시작...")
        for row in range(2, cumulative_sheet.UsedRange.Rows.Count + 1):
            for col in range(2, 9):
                if cumulative_sheet.Cells(row, col).Value is not None and cumulative_sheet.Cells(row, col + 7).Value is None:
                    for shift_col in range(10, 16):
                        cumulative_sheet.Cells(row, shift_col).Value = cumulative_sheet.Cells(row - 1, shift_col).Value

        workbook.Save()
    except Exception as e:
        print(f"작업 중 오류 발생: {e}")
    finally:
        excel.ScreenUpdating = True  # 화면 업데이트 다시 활성화
        print("데이터 병합 및 엑셀 처리 완료.")

    return excel, workbook

# PDF 변환 및 저장 함수
def export_sheet_to_pdf(excel, workbook):
    current_date = datetime.now().strftime("%Y%m%d")
    output_dir = "D:\\신규P값총량제\\공유메일"
    output_pdf = f"메일용_{current_date}.pdf"
    final_pdf = os.path.join(output_dir, output_pdf)

    mail_sheet = workbook.Sheets("메일")
    mail_sheet.Visible = True

    mail_sheet.PageSetup.PrintGridlines = False
    mail_sheet.ExportAsFixedFormat(0, final_pdf)

    os.startfile(final_pdf)

    print(f"PDF 파일이 {final_pdf}에 저장되고 열렸습니다.")

    workbook.Close(SaveChanges=True)
    excel.Quit()

# 메인 함수 실행
if __name__ == "__main__":
    try:
        kill_excel_process()  # 백그라운드의 엑셀 프로세스를 강제로 종료
        excel_app, open_workbook = copy_data_to_excel()  # 엑셀 작업 수행 후 닫기 전에 계속 유지
        export_sheet_to_pdf(excel_app, open_workbook)  # PDF 저장 후에만 엑셀 파일을 닫음
    except Exception as e:
        print(f"프로그램 실행 중 오류 발생: {e}")
```

```python
import os
import pandas as pd
import glob
import win32com.client
from datetime import datetime
import numpy as np

# 폴더 경로 설정 (일일 데이터와 누계 데이터 경로)
daily_folder_path = r'd:\신규P값총량제\병합'
cumulative_folder_path = r'd:\신규P값총량제\누계병합'

# 파일명에 따른 정렬 기준
file_order = [
    "0. 채널별, 월", "1.기관별, 월",  # "기관별" 파일을 채널 다음으로 정렬
    "2.강북_월", "2.강남_월", "2.부산",
    "2.전남", "2.대구", "2.충남",
    "사원별"
]

# 파일명 정렬 함수
def get_file_priority(file_name):
    for i, keyword in enumerate(file_order):
        if keyword in file_name:
            return i
    return len(file_order)

# "제거할 특정 숫자 + 공백 + 건" 패턴을 포함하는 행을 필터링하기 위한 함수 (A열에만 적용)
def filter_out_specific_gun_rows(df):
    remove_patterns = [r'8\s건$', r'12\s건$', r'6\s건$', r'9\s건$', r'7\s건$']
    pattern = '|'.join(remove_patterns)
    return df[~df.iloc[:, 0].astype(str).str.contains(pattern)]

# 폴더에서 파일을 읽어 병합하고 저장하는 함수
def merge_and_save_files(folder_path, output_filename):
    all_files = glob.glob(folder_path + "/*.xls")
    all_files.sort(key=lambda x: get_file_priority(os.path.basename(x)))

    all_data = pd.DataFrame()

    for file in all_files:
        try:
            df_list = pd.read_html(file)
            df = df_list[0]
            df = filter_out_specific_gun_rows(df)

            if all_data.empty:
                all_data = df
            else:
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)

        except Exception as e:
            print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")

    if not all_data.empty:
        csv_output_path = os.path.join(folder_path, output_filename)
        all_data.to_csv(csv_output_path, index=False, encoding='cp949')
        print(f"병합된 파일이 {csv_output_path}에 저장되었습니다.")
        return csv_output_path
    else:
        print("병합된 데이터가 비어 있습니다. 엑셀 파일들을 확인하세요.")
        return None

# Excel 데이터 병합 및 복사 최적화
def paste_dataframe_to_excel(sheet, start_cell, df):
    values = df.fillna("").values.tolist()
    start_row, start_col = start_cell
    end_row = start_row + len(values) - 1
    end_col = start_col + len(values[0]) - 1

    excel_range = sheet.Range(sheet.Cells(start_row, start_col), sheet.Cells(end_row, end_col))
    excel_range.Value = values

# Excel 데이터 병합 및 붙여넣기
def copy_data_to_excel():
    try:
        excel = win32com.client.Dispatch("Excel.Application")
        excel.Visible = True
        excel.ScreenUpdating = False
        excel.DisplayAlerts = False
        excel.Calculation = -4135  # xlCalculationManual

        if excel.Workbooks.Count == 0:
            raise Exception("열려있는 엑셀 파일이 없습니다. 엑셀 파일을 열어주세요.")

        workbook = excel.Workbooks(1)

        try:
            pv_db_sheet = workbook.Sheets("Pv_db")
            cumulative_sheet = workbook.Sheets("92_SP사원(개인)(9월 누계)")
        except Exception as e:
            raise Exception(f"필요한 시트를 찾을 수 없습니다. 시트가 존재하는지 확인하세요.")

        daily_csv = merge_and_save_files(daily_folder_path, "신규P값병합_일일.csv")
        cumulative_csv = merge_and_save_files(cumulative_folder_path, "신규P값병합_누계.csv")

        if not daily_csv or not cumulative_csv:
            raise Exception("CSV 병합 파일 생성 중 오류가 발생했습니다. 확인 후 다시 시도하세요.")

        daily_data = pd.read_csv(daily_csv, encoding='cp949')
        cumulative_data = pd.read_csv(cumulative_csv, encoding='cp949')

        print("누계 데이터 복사 중...")
        paste_dataframe_to_excel(pv_db_sheet, (1, 2), cumulative_data)

        print("일일 데이터 복사 중...")
        paste_dataframe_to_excel(pv_db_sheet, (1, 13), daily_data)

        for row in range(1, pv_db_sheet.UsedRange.Rows.Count + 1):
            if "본사" in str(pv_db_sheet.Cells(row, 13).Value):
                for col in range(13, 21):
                    cumulative_sheet.Cells(2, col - 11).Value = pv_db_sheet.Cells(row, col).Value
                break

        for row in range(2, cumulative_sheet.UsedRange.Rows.Count + 1):
            for col in range(2, 9):
                if cumulative_sheet.Cells(row, col).Value is not None and cumulative_sheet.Cells(row, col + 7).Value is None:
                    for shift_col in range(10, 16):
                        cumulative_sheet.Cells(row, shift_col).Value = cumulative_sheet.Cells(row - 1, shift_col).Value

        current_date = datetime.now().strftime("%Y%m%d")
        new_filename = f"총량제 실적현황_{current_date}.xlsx"
        save_path = os.path.join(os.path.dirname(workbook.FullName), new_filename)
        workbook.SaveAs(save_path)

        print(f"데이터 병합 및 엑셀 저장 완료. 파일명: {save_path}")
        return excel, workbook, save_path
    except Exception as e:
        print(f"작업 중 오류 발생: {e}")
        return None, None, None
    finally:
        if excel:
            excel.ScreenUpdating = True
            excel.DisplayAlerts = True
            excel.Calculation = -4105  # xlCalculationAutomatic

# PDF 변환 및 저장 함수
def export_sheet_to_pdf(excel, workbook, excel_file_path):
    try:
        current_date = datetime.now().strftime("%Y%m%d")
        output_dir = os.path.dirname(excel_file_path)
        output_pdf = f"메일용_{current_date}.pdf"
        final_pdf = os.path.join(output_dir, output_pdf)

        mail_sheet = workbook.Sheets("메일")
        mail_sheet.Visible = True

        mail_sheet.PageSetup.PrintGridlines = False
        mail_sheet.ExportAsFixedFormat(0, final_pdf)

        os.startfile(final_pdf)
        print(f"PDF 파일이 {final_pdf}에 저장되고 열렸습니다.")
    except Exception as e:
        print(f"PDF 변환 중 오류 발생: {e}")

# 메인 함수 실행
if __name__ == "__main__":
    excel_app, open_workbook, excel_file_path = copy_data_to_excel()
    if excel_app and open_workbook and excel_file_path:
        export_sheet_to_pdf(excel_app, open_workbook, excel_file_path)
        open_workbook.Close(SaveChanges=True)
        excel_app.Quit()
```

```python
import os
import pandas as pd
import glob
import win32com.client
from datetime import datetime
import time

# 폴더 경로 설정 (일일 데이터와 누계 데이터 경로)
daily_folder_path = r'd:\신규P값총량제\병합'
cumulative_folder_path = r'd:\신규P값총량제\누계병합'
employee_file_path = r'd:\신규P값총량제\병합\3.사원.xls'  # 사원별 데이터를 직접 지정

# 파일명에 따른 정렬 기준
file_order = [
    "0. 채널별, 월", "1.기관별, 월",
    "2.강북_월", "2.강남_월", "2.부산",
    "2.전남", "2.대구", "2.충남"
]

# 파일명 정렬 함수
def get_file_priority(file_name):
    for i, keyword in enumerate(file_order):
        if keyword in file_name:
            return i
    return len(file_order)

# 현재 날짜 폴더 생성 및 경로 반환
def create_date_folder(base_path):
    current_date = datetime.now().strftime("%Y%m%d")
    date_folder = os.path.join(base_path, f"총량제작업_{current_date}")
    if not os.path.exists(date_folder):
        os.makedirs(date_folder)
    return date_folder

# 특정 엑셀 파일을 읽어 CSV로 저장하는 함수 (사원별 데이터)
def save_employee_data_to_csv(file_path, save_folder):
    try:
        # HTML 형식으로 읽기 시도
        df_list = pd.read_html(file_path)
        if not df_list:
            raise Exception(f"파일에서 데이터를 읽지 못했습니다: {file_path}")
       
        # 데이터프레임 첫 번째 테이블 저장 (A3:H열만 사용)
        employee_data = df_list[0].iloc[2:, :8]  # A3부터 H열까지 사용
        employee_csv_path = os.path.join(save_folder, "사원별_데이터.csv")
        employee_data.to_csv(employee_csv_path, index=False, encoding='cp949')
        print(f"사원별 데이터 파일이 {employee_csv_path}에 저장되었습니다.")
        return employee_csv_path
    except Exception as e:
        print(f"사원별 데이터를 읽어오는 중 오류 발생: {e}")
        return None

# 폴더에서 파일을 읽어 병합하고 저장하는 함수
def merge_and_save_files(folder_path, output_filename, save_folder):
    all_files = glob.glob(folder_path + "/*.xls")
    all_files.sort(key=lambda x: get_file_priority(os.path.basename(x)))

    all_data = pd.DataFrame()

    if not all_files:
        print(f"경고: 병합할 파일이 '{folder_path}' 폴더에 없습니다. 경로를 확인하세요.")
        return None

    for file in all_files:
        try:
            df_list = pd.read_html(file)
            if len(df_list) == 0:
                print(f"경고: {file}에서 읽어온 데이터가 비어있습니다.")
                continue
            df = df_list[0]

            if all_data.empty:
                all_data = df
            else:
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)

        except Exception as e:
            print(f"파일을 읽는 중 오류 발생: {file}, 오류: {e}")

    if not all_data.empty:
        csv_output_path = os.path.join(save_folder, output_filename)
        all_data.to_csv(csv_output_path, index=False, encoding='cp949')
        print(f"병합된 파일이 {csv_output_path}에 저장되었습니다.")
        return csv_output_path
    else:
        print("경고: 병합된 데이터가 비어 있습니다. 엑셀 파일들을 확인하세요.")
        return None

# 데이터프레임을 2D 배열로 변환 후 엑셀에 빠르게 붙여넣는 함수
def paste_dataframe_to_excel(sheet, start_cell, df):
    values = df.fillna("").values.tolist()
    start_row, start_col = start_cell
    end_row = start_row + len(values) - 1
    end_col = start_col + len(values[0]) - 1

    excel_range = sheet.Range(sheet.Cells(start_row, start_col), sheet.Cells(end_row, end_col))
    excel_range.Value = values

# Excel 데이터를 병합하고 붙여넣는 함수
def copy_data_to_excel(base_folder):
    try:
        save_folder = create_date_folder(base_folder)

        excel = win32com.client.Dispatch("Excel.Application")
        excel.Visible = True
        excel.ScreenUpdating = False
        excel.DisplayAlerts = False
        excel.Calculation = -4135  # xlCalculationManual

        if excel.Workbooks.Count == 0:
            raise Exception("열려있는 엑셀 파일이 없습니다. 엑셀 파일을 열어주세요.")

        workbook = excel.Workbooks(1)

        # 필요한 시트 설정
        pv_db_sheet = workbook.Sheets("Pv_db")
        cumulative_sheet = workbook.Sheets("92_SP사원(개인)(9월 누계)")
        channel_sheet = workbook.Sheets("91_기관&채널(AM-SC-SP)(9월누계)")

        # 병합 파일 생성 및 저장 경로 설정
        daily_csv = merge_and_save_files(daily_folder_path, "신규P값병합_일일.csv", save_folder)
        cumulative_csv = merge_and_save_files(cumulative_folder_path, "신규P값병합_누계.csv", save_folder)
        employee_csv = save_employee_data_to_csv(employee_file_path, save_folder)

        if not daily_csv or not cumulative_csv or not employee_csv:
            raise Exception("CSV 병합 파일 생성 중 오류가 발생했습니다. 확인 후 다시 시도하세요.")

        daily_data = pd.read_csv(daily_csv, encoding='cp949')
        cumulative_data = pd.read_csv(cumulative_csv, encoding='cp949')
        employee_data = pd.read_csv(employee_csv, encoding='cp949')

        # 누계 데이터 복사
        paste_dataframe_to_excel(pv_db_sheet, (1, 2), cumulative_data)
        # 일일 데이터 복사
        paste_dataframe_to_excel(pv_db_sheet, (1, 13), daily_data)

        # 사원별 데이터를 92_SP사원(개인)(9월 누계) 시트의 B2~I열에 복사
        paste_dataframe_to_excel(cumulative_sheet, (2, 2), employee_data)

        # J열부터 O열까지 비어 있는 경우 위의 서식 복사
        for row in range(2, cumulative_sheet.UsedRange.Rows.Count + 1):
            for col in range(10, 16):  # J~O 열
                if cumulative_sheet.Cells(row, col).Value is None:
                    cumulative_sheet.Cells(row, col).Value = cumulative_sheet.Cells(row - 1, col).Value

        # Pv_db 시트에서 M2:M5의 N~Q 값을 91_기관&채널 시트의 Q3:Q6의 R~U에 복사
        for source_row, target_row in zip(range(2, 6), range(3, 7)):  # M2:M5 -> Q3:Q6
            for col_offset in range(4):  # N, O, P, Q -> R, S, T, U
                channel_sheet.Cells(target_row, 18 + col_offset).Value = pv_db_sheet.Cells(source_row, 14 + col_offset).Value

        # 엑셀 파일을 새로운 이름으로 저장
        current_date = datetime.now().strftime("%Y%m%d")
        new_filename = f"총량제 실적현황_{current_date}.xlsx"
        save_path = os.path.join(save_folder, new_filename)
        workbook.SaveAs(save_path)

        print(f"데이터 병합 및 엑셀 저장 완료. 파일명: {save_path}")
        return excel, workbook, save_path
    except Exception as e:
        print(f"작업 중 오류 발생: {e}")
        return None, None, None
    finally:
        if excel:
            excel.ScreenUpdating = True
            excel.DisplayAlerts = True
            excel.Calculation = -4105  # xlCalculationAutomatic

# 메인 함수 실행
if __name__ == "__main__":
    main()
```

```python
import os
import pandas as pd
import glob
import win32com.client
from datetime import datetime
import time

# 폴더 경로 설정
daily_folder_path = r'd:\신규P값총량제\병합'
cumulative_folder_path = r'd:\신규P값총량제\누계병합'
employee_file_path = r'd:\신규P값총량제\병합\3.사원.xls'  # 사원별 데이터를 직접 지정

# 현재 날짜 폴더 생성 및 경로 반환
def create_date_folder(base_path):
    current_date = datetime.now().strftime("%Y%m%d")
    date_folder = os.path.join(base_path, f"총량제작업_{current_date}")
    if not os.path.exists(date_folder):
        os.makedirs(date_folder)
    return date_folder

# 특정 엑셀 파일을 읽어 CSV로 저장하는 함수 (사원별 데이터)
def save_employee_data_to_csv(file_path, save_folder):
    try:
        df_list = pd.read_html(file_path)
        if not df_list:
            raise Exception(f"파일에서 데이터를 읽지 못했습니다: {file_path}")

        # 데이터프레임 첫 번째 테이블 저장
        employee_data = df_list[0]
        employee_csv_path = os.path.join(save_folder, "사원별_데이터.csv")
        employee_data.to_csv(employee_csv_path, index=False, encoding='cp949')
        print(f"사원별 데이터 파일이 {employee_csv_path}에 저장되었습니다.")
        return employee_csv_path
    except Exception as e:
        print(f"사원별 데이터를 읽어오는 중 오류 발생: {e}")
        return None

# 시트가 있는지 확인하는 함수
def check_sheet_exists(workbook, sheet_name):
    for sheet in workbook.Sheets:
        if sheet.Name == sheet_name:
            return True
    return False

# 데이터프레임을 2D 배열로 변환 후 엑셀에 빠르게 붙여넣는 함수 (공백 제외)
def paste_dataframe_to_excel(sheet, start_cell, df):
    values = df.dropna(how='all').fillna("").values.tolist()
    start_row, start_col = start_cell
    end_row = start_row + len(values) - 1
    end_col = start_col + len(values[0]) - 1
    excel_range = sheet.Range(sheet.Cells(start_row, start_col), sheet.Cells(end_row, end_col))
    excel_range.Value = values

# Excel 데이터를 병합하고 붙여넣는 함수
def copy_data_to_excel(base_folder):
    try:
        save_folder = create_date_folder(base_folder)
        excel = win32com.client.Dispatch("Excel.Application")
        excel.Visible = True
        excel.ScreenUpdating = False
        excel.DisplayAlerts = False
        excel.Calculation = -4135  # xlCalculationManual

        if excel.Workbooks.Count == 0:
            raise Exception("열려있는 엑셀 파일이 없습니다. 엑셀 파일을 열어주세요.")
       
        workbook = excel.Workbooks(1)
        required_sheets = ["Pv_db", "92_SP사원(개인)(9월 누계)", "91_기관&채널(AM-SC-SP)(9월누계)", "11_(기관)신규P값 성과"]

        for sheet_name in required_sheets:
            if not check_sheet_exists(workbook, sheet_name):
                raise Exception(f"필요한 시트 '{sheet_name}'가 없습니다.")

        # 필요한 시트 설정
        pv_db_sheet = workbook.Sheets("Pv_db")
        cumulative_sheet = workbook.Sheets("92_SP사원(개인)(9월 누계)")
        channel_sheet = workbook.Sheets("91_기관&채널(AM-SC-SP)(9월누계)")
        target_sheet = workbook.Sheets("11_(기관)신규P값 성과")

        # 사원별 데이터 파일 읽어와서 붙여넣기
        employee_csv = save_employee_data_to_csv(employee_file_path, save_folder)
        if not employee_csv:
            raise Exception("사원별 CSV 파일 생성 중 오류가 발생했습니다.")

        employee_data = pd.read_csv(employee_csv, encoding='cp949')
        paste_dataframe_to_excel(cumulative_sheet, (2, 2), employee_data)  # B2~I열 값 적용

        # Pv_db 시트의 M열 값 -> 91_기관&채널 시트의 Q열 복사 (공백 제외)
        for row_offset in range(2, 6):
            for col_offset in range(1, 5):
                if pv_db_sheet.Cells(row_offset, 13).Value:
                    channel_sheet.Cells(row_offset + 1, col_offset + 16).Value = pv_db_sheet.Cells(row_offset, col_offset + 13).Value

        # 엑셀 파일을 새로운 이름으로 저장
        current_date = datetime.now().strftime("%Y%m%d")
        new_filename = f"총량제 실적현황_{current_date}.xlsx"
        save_path = os.path.join(save_folder, new_filename)
        workbook.SaveAs(save_path)

        print(f"데이터 병합 및 엑셀 저장 완료. 파일명: {save_path}")
        return excel, workbook, save_path
    except Exception as e:
        print(f"작업 중 오류 발생: {e}")
        return None, None, None
    finally:
        if excel:
            excel.ScreenUpdating = True
            excel.DisplayAlerts = True
            excel.Calculation = -4105  # xlCalculationAutomatic

# 메인 함수 실행
if __name__ == "__main__":
    base_folder = r'd:\신규P값총량제'
    copy_data_to_excel(base_folder)
```
