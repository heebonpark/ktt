```python
import pandas as pd
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import os

# 파일 경로 설정
input_file_path = r'D:\강북강원본부\안심검증\설변_5월마감 안심검증_6.7추출\G000_02753_00069.csv'
output_file_path_cs = os.path.join(os.path.dirname(input_file_path), '5월 설치변경.cs')
output_file_path_xlsx = os.path.join(os.path.dirname(input_file_path), '5월 설치변경.xlsx')

# CSV 파일을 불러와서 DataFrame으로 변환 (인코딩 cp949)
df = pd.read_csv(input_file_path, encoding='cp949')

# DataFrame을 .cs 확장자로 저장 (인코딩 cp949)
df.to_csv(output_file_path_cs, index=False, encoding='cp949')

# .cs 파일을 불러오기 (인코딩 cp949)
df_loaded = pd.read_csv(output_file_path_cs, encoding='cp949')

# DataFrame을 Excel 파일로 저장
df_loaded.to_excel(output_file_path_xlsx, index=False)

# GUI 생성
class DataFrameViewer(tk.Tk):
    def __init__(self, dataframe):
        super().__init__()
       
        self.title('DataFrame Viewer')
        self.geometry('1000x600')
       
        self.dataframe = dataframe
        self.filtered_dataframe = dataframe
       
        self.create_widgets()
       
    def create_widgets(self):
        # 검색 및 조회를 위한 입력 필드와 버튼 생성
        self.label1 = tk.Label(self, text='계약번호:')
        self.label1.pack(anchor='w')
        self.entry_contract = tk.Entry(self)
        self.entry_contract.pack(anchor='w')
       
        self.label2 = tk.Label(self, text='상호 명:')
        self.label2.pack(anchor='w')
        self.entry_name = tk.Entry(self)
        self.entry_name.pack(anchor='w')
       
        self.search_button = tk.Button(self, text='조회', command=self.search)
        self.search_button.pack(anchor='w')
       
        self.show_all_button = tk.Button(self, text='전체 리스트 보기', command=self.show_all)
        self.show_all_button.pack(anchor='w')

        self.save_button = tk.Button(self, text='엑셀로 저장', command=self.save_to_excel)
        self.save_button.pack(anchor='w')
       
        # Treeview와 스크롤바 생성
        self.tree_frame = tk.Frame(self)
        self.tree_frame.pack(expand=True, fill='both')
       
        self.tree_scroll = tk.Scrollbar(self.tree_frame)
        self.tree_scroll.pack(side='right', fill='y')
       
        self.tree = ttk.Treeview(self.tree_frame, yscrollcommand=self.tree_scroll.set, show='headings')
        self.tree.pack(expand=True, fill='both')
       
        self.tree_scroll.config(command=self.tree.yview)
       
        self.update_treeview(self.dataframe)
   
    def update_treeview(self, dataframe):
        # Treeview 초기화
        self.tree.delete(*self.tree.get_children())
        self.tree['columns'] = list(dataframe.columns)
       
        for col in dataframe.columns:
            self.tree.heading(col, text=col)
            self.tree.column(col, anchor='center', width=100)
       
        for index, row in dataframe.iterrows():
            self.tree.insert('', 'end', values=list(row))
   
    def search(self):
        contract_number = self.entry_contract.get()
        name = self.entry_name.get()
       
        filtered_df = self.dataframe
       
        if contract_number:
            filtered_df = filtered_df[filtered_df['계약번호'].astype(str).str.contains(contract_number)]
       
        if name:
            filtered_df = filtered_df[filtered_df['상호 명'].astype(str).str.contains(name)]
       
        self.filtered_dataframe = filtered_df
        self.update_treeview(filtered_df)
   
    def show_all(self):
        self.filtered_dataframe = self.dataframe
        self.update_treeview(self.dataframe)
   
    def save_to_excel(self):
        if self.filtered_dataframe is not None:
            file_path = filedialog.asksaveasfilename(defaultextension=".xlsx", filetypes=[("Excel files", "*.xlsx"), ("All files", "*.*")])
            if file_path:
                self.filtered_dataframe.to_excel(file_path, index=False)
                messagebox.showinfo("저장 완료", "엑셀 파일로 저장되었습니다.")
        else:
            messagebox.showwarning("경고", "저장할 데이터가 없습니다.")

# GUI 실행
if __name__ == '__main__':
    df = pd.read_csv(output_file_path_cs, encoding='cp949')
   
    app = DataFrameViewer(df)
    app.mainloop()
```
