```python
import pandas as pd

import matplotlib.pyplot as plt

import openpyxl as px



# 열려있는 엑셀 파일 접근 (엑셀 파일은 이미 열려 있다고 가정)

file_path = 'your_open_excel_file.xlsx'  # 열려있는 엑셀 파일 경로를 입력하세요.

sheet_name = 'Sheet1'  # 시트 이름



# 엑셀 파일에서 데이터프레임 읽기

# openpyxl을 통해 열려있는 파일을 불러와서 pandas에서 처리

wb = px.load_workbook(file_path, data_only=True)

ws = wb[sheet_name]

data = ws.values



# 첫 번째 행을 열 이름으로 사용하여 데이터프레임 생성

columns = next(data)

df = pd.DataFrame(data, columns=columns)



# 천 단위 구분 기호가 포함된 '누적평균' 열을 숫자 형식으로 변환

df['누적평균'] = df['누적평균'].replace(',', '', regex=True).astype(float)



# 산점도 시각화

plt.figure(figsize=(10, 6))

plt.scatter(df.index, df['누적평균'], color='blue', label='누적평균')



# Y축 자동 조정

min_value = df['누적평균'].min()

max_value = df['누적평균'].max()

plt.ylim(min_value - (max_value - min_value) * 0.1, max_value + (max_value - min_value) * 0.1)



# 제목 및 축 레이블 설정

plt.title('누적평균 산점도')

plt.xlabel('데이터 인덱스')

plt.ylabel('누적평균 (금액)')



# 범례 추가

plt.legend()



# 그래프 출력

plt.show()
```
