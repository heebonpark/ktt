````python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
import matplotlib.pyplot as plt

# Google Sheets API에 대한 접근 권한 설정
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
creds = ServiceAccountCredentials.from_json_keyfile_name('credentials.json', scope)
client = gspread.authorize(creds)

# Google 스프레드시트 불러오기
spreadsheet_url = "https://docs.google.com/spreadsheets/d/11FdXLpoCYsgAcaDoETodiJz93riz-iOKEXvDOMCTVV0/edit"  # 스프레드시트 URL을 여기에 입력
spreadsheet = client.open_by_url(spreadsheet_url)
worksheet = spreadsheet.worksheet("Sheet1")  # 시트 이름을 정확히 입력

# 데이터를 DataFrame으로 불러오기
data = worksheet.get_all_records()
df = pd.DataFrame(data)

# DataFrame 데이터 확인
print(df.head())

# 산점도 시각화: 지사 vs 금액
plt.figure(figsize=(10, 6))
plt.scatter(df['지사'], df['금액'], color='blue', alpha=0.6)
plt.title("지사별 금액 산점도")
plt.xlabel("지사")
plt.ylabel("금액")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
````
