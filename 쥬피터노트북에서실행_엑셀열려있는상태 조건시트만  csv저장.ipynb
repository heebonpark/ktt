```python
# ✅ 1. 라이브러리 설치
!pip install pywin32 pandas openpyxl --quiet

# ✅ 2. 실행 코드
import win32com.client as win32
import pandas as pd
import datetime

# 엑셀 연결
excel = win32.GetActiveObject("Excel.Application")
wb = excel.ActiveWorkbook

# 대상 시트
target_sheets = ["2025년 관리고객 리스트", "(2025) 전담사원 변경요청"]
save_path = r"C:\Users\Public\분리저장_관리고객_리스트.xlsx"

def remove_timezone(val):
    """셀 값이 datetime with timezone이면 tz 제거"""
    if isinstance(val, datetime.datetime) and val.tzinfo is not None:
        return val.replace(tzinfo=None)
    return val

with pd.ExcelWriter(save_path, engine='openpyxl', datetime_format='yyyy-mm-dd') as writer:
    for sheet in wb.Sheets:
        if sheet.Name in target_sheets:
            data = sheet.UsedRange.Value
            if data:
                columns = list(map(str, data[0]))
                if len(set(columns)) != len(columns):
                    columns = [f"{col}_{i}" if columns.count(col) > 1 else col for i, col in enumerate(columns)]
                rows = data[1:]

                # 셀별로 tz 제거
                cleaned_rows = [
                    [remove_timezone(cell) for cell in row]
                    for row in rows
                ]

                df = pd.DataFrame(cleaned_rows, columns=columns)
                df.to_excel(writer, sheet_name=sheet.Name[:31], index=False)

print(f"엑셀 시트별 저장 완료 (타임존 제거, 날짜 유지): {save_path}")
```
