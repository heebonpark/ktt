```python
import pandas as pd
import xlwings as xw
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go

# 1. 엑셀 데이터 읽기 함수
def get_excel_data():
    wb = xw.books.active
    sheet = wb.sheets.active
    data = sheet.range("A1:J53").options(pd.DataFrame, header=1, index=False).value
    return data

# 2. 데이터 읽기 및 포맷팅
df = get_excel_data()

# 소계 행 및 전체 합계 행 제거
df = df[~df["성명"].str.contains("소계", na=False)]  # '성명' 열에서 "소계" 키워드를 포함한 행 제거
df = df[~df["지사"].str.contains("전체합계", na=False)]  # '지사' 열에서 "전체합계" 키워드를 포함한 행 제거

# 데이터 포맷팅
df["월정료"] = df["월정료"].apply(lambda x: f"{x:,.0f}" if pd.notnull(x) else x)
df["P.P 공헌이익금"] = df["P.P 공헌이익금"].apply(lambda x: f"{x:,.0f}" if pd.notnull(x) else x)

# 달성율 데이터 처리
df["달성율 연간 목표(가중치 70%)"] = (
    df["달성율 연간 목표(가중치 70%)"]
    .astype(str)
    .str.replace("%", "")
    .astype(float)
)
df["달성율 개선 목표(가중치 30%)"] = (
    df["달성율 개선 목표(가중치 30%)"]
    .astype(str)
    .str.replace("%", "")
    .astype(float)
)

# KPI 배점 소수점 둘째 자리까지 표시
df["KPI 배점(5점)"] = df["KPI 배점(5점)"].apply(lambda x: f"{x:.2f}" if isinstance(x, float) else x)

# 조건부 아이콘 표시 함수
def get_icon(value):
    if value >= 90:
        return "🟢"
    elif 80 <= value < 90:
        return "🟡"
    else:
        return "🔴"

df["연간 목표 아이콘"] = df["달성율 연간 목표(가중치 70%)"].apply(get_icon)
df["개선 목표 아이콘"] = df["달성율 개선 목표(가중치 30%)"].apply(get_icon)

# RANK (264명) 처리 및 NaN 값 처리
df["RANK (264명)"] = pd.to_numeric(df["RANK (264명)"], errors="coerce")  # NaN 값 처리
df["RANK (264명)"] = df["RANK (264명)"].fillna(df["RANK (264명)"].max() + 1)  # NaN 값을 최대값 + 1로 대체
df["RANK (264명)"] = df["RANK (264명)"].astype(int)  # 정수로 변환

# 지사별 및 성명별 RANK 계산
df["지사별 순위"] = df.groupby("지사")["RANK (264명)"].rank(ascending=True).astype(int)
df["성명별 순위"] = df["RANK (264명)"].rank(ascending=False).astype(int)  # 내림차순 설정

# Dash 앱 초기화
app = Dash(__name__)

# 반응형 레이아웃 스타일 정의
app.layout = html.Div([
    html.H1("지사별 및 성명별 데이터 대시보드", style={"textAlign": "center"}),

    # 필터 드롭다운
    html.Div([
        html.Label("지사 선택:", style={"fontSize": "18px"}),
        dcc.Dropdown(
            id="branch-filter",
            options=[{"label": branch, "value": branch} for branch in df["지사"].unique()],
            placeholder="지사를 선택하세요",
            style={"fontSize": "16px"}
        ),
    ], style={"width": "45%", "display": "inline-block"}),

    html.Div([
        html.Label("성명 선택:", style={"fontSize": "18px"}),
        dcc.Dropdown(
            id="name-filter",
            placeholder="성명을 선택하세요",
            style={"fontSize": "16px"}
        ),
    ], style={"width": "45%", "display": "inline-block", "marginLeft": "5%"}),

    # 그래프 및 테이블
    dcc.Graph(id="pp-bar-chart"),
    dcc.Graph(id="rank-bar-chart"),
    html.Div(id="data-table", style={"marginTop": "20px"})
])

@app.callback(
    Output("name-filter", "options"),
    Input("branch-filter", "value")
)
def update_name_filter(selected_branch):
    if selected_branch:
        filtered_df = df[df["지사"] == selected_branch]
        return [{"label": name, "value": name} for name in filtered_df["성명"].unique()]
    return []

@app.callback(
    [Output("pp-bar-chart", "figure"),
     Output("rank-bar-chart", "figure"),
     Output("data-table", "children")],
    [Input("branch-filter", "value"), Input("name-filter", "value")]
)
def update_dashboard(selected_branch, selected_name):
    filtered_df = df.copy()
    if selected_branch:
        filtered_df = filtered_df[filtered_df["지사"] == selected_branch]
    if selected_name:
        filtered_df = filtered_df[filtered_df["성명"] == selected_name]

    # P.P 공헌이익금 기준 내림차순 정렬
    filtered_df = filtered_df.sort_values(by="P.P 공헌이익금", ascending=False)

    # 그래프 생성 - P.P 공헌이익금 및 월정료
    pp_fig = go.Figure()
    pp_fig.add_trace(go.Bar(
        x=filtered_df["성명"],
        y=filtered_df["P.P 공헌이익금"].str.replace(",", "").astype(float),
        name="P.P 공헌이익금",
        text=filtered_df["P.P 공헌이익금"],
        textposition="auto",
        marker=dict(color="blue")
    ))
    pp_fig.add_trace(go.Bar(
        x=filtered_df["성명"],
        y=filtered_df["월정료"].str.replace(",", "").astype(float),
        name="월정료",
        text=filtered_df["월정료"],
        textposition="auto",
        marker=dict(color="red")
    ))
    pp_fig.update_layout(
        title="P.P 공헌이익금 및 월정료 분석 (내림차순)",
        xaxis_title="성명",
        yaxis_title="값",
        barmode="group",
        legend_title="항목",
        title_font_size=20,
        title_x=0.5
    )

    # 그래프 생성 - 성명별 순위
    rank_fig = go.Figure()
    rank_fig.add_trace(go.Bar(
        x=filtered_df["성명"],
        y=filtered_df["성명별 순위"],
        name="성명별 순위",
        text=filtered_df["성명별 순위"],
        textposition="auto",
        marker=dict(color="green")
    ))
    rank_fig.update_layout(
        title="성명별 순위 (내림차순)",
        xaxis_title="성명",
        yaxis_title="순위",
        legend_title="항목",
        title_font_size=20,
        title_x=0.5
    )

    # 데이터 테이블 생성 (테두리와 RANK 추가)
    table = html.Table(
        style={"borderCollapse": "collapse", "width": "100%", "fontSize": "16px"},
        children=[
            # 테이블 헤더
            html.Thead(
                html.Tr([
                    html.Th(col, style={"border": "1px solid gray", "padding": "8px", "backgroundColor": "#f2f2f2"})
                    for col in filtered_df.columns
                ])
            ),
            # 테이블 본문
            html.Tbody([
                html.Tr([
                    html.Td(filtered_df.iloc[i][col], style={"border": "1px solid gray", "padding": "8px"})
                    for col in filtered_df.columns
                ]) for i in range(len(filtered_df))
            ])
        ]
    )

    return pp_fig, rank_fig, table

if __name__ == "__main__":
    app.run_server(debug=True)
```
